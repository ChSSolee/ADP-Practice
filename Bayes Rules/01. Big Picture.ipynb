#### 질병검사

| | 양성 | 음성 | 총계 |
| -- | -- | -- | -- |
| 질병 | 3 | 1 | 4 |
| 질병없음 | 9 | 87 | 96 |
| 총계 | 12 | 88 | 100 |

# 
- **베이지안의 질문 : 테스트 결과가 양성일때, 실제로 질병에 걸렸을 가능성은?** 
    - 양성 판정을 받은 12명 중 3명만이 질병을 가지고 있기 때문에, 질병에 걸릴 확률은 25%
    - 따라서 질병의 희소성과 상대적으로 높은 위양성률을 고려할 때 실제로 질병에 걸렸을 가능성은 상대적으로 낮음
    
# 

- **빈도주의자의 질문 : 질병 상태는 반복될 수 없기 때문에 질병에 걸렸을 확률은 1 또는 0. 하지만 검사는 반복가능**
    - **실제로 질병이 없는데 양성 판정을 받았을 가능성은?** 
    - 질병이 없는 96명 중 9명만이 양성 반응을 보였기 때문에 질병이 없더라도 양성 반응을 보일 확률은 대략 10%(9/96)
    - 빈도주의자의 확률 계산은 p-값 과 유사
    - 일반적으로 p-값은 실제로 우리의 원래 가설이 잘못된 경우 데이터가 우리보다 극단적이거나 더 극단적일 가능성을 측정
    - p-값은 수십 년 동안 빈도주의 관행에서 두드러졌지만 빈도주의와 베이지안 스펙트럼에서 서서히 강조되지 않음. 
    - 본질적으로 매우 일반적으로 잘못 해석되고 오용

### 1-2 베이지안의 역사

- 베이지안 통계는 18세기 통계학자, 장관, 철학자인 Thomas Bayes에 뿌리
- Bayes는 1740년대에 자신의 철학을 발전시켰지만 이 철학이 대두된 것은 20세기 후반
- 그 사이 200년 이상 동안 빈도주의 철학이 통계 연구와 실천을 지배

# 
*베이지안 철학은 이 기간 동안 인기가 떨어졌을 뿐만 아니라 낙인 찍혔습니다. 1990년에 Marilyn vos Savant는 현재 고전적인 Monty Hall 에 베이지안 솔루션을 제시했을 때 Parade 잡지 칼럼의 일부 독자들에게 조롱을 받았습니다.확률 퍼즐. 분명히 10,000명 이상의 독자들이 그녀의 해결책에 대해 이의를 제기하고 조롱하기 위해 글을 썼습니다. 7 Vos Savant의 Bayesian 솔루션은 나중에 명백히 올바른 것으로 입증되었습니다. 그러나 이러한 수준의 정밀 조사를 통해 많은 연구자들이 베이지안 추구를 비밀로 유지한 것은 놀라운 일이 아닙니다. 당시에는 그렇게 많이 선포되지 않았지만 Alan Turing은 제2차 세계 대전에서 독일의 수수께끼 코드를 풀었고 John Tukey는 베이지안 방법을 사용하여 1960년대에 선거일 예측을 개척했습니다( McGrayne 2012 )*

#### 베이지안이 가지고 있었던 한계점
1. 컴퓨팅의 발전 : 베이지안 철학은 1990년대까지 광범위하게 사용할 수 없었던 정교한 컴퓨팅 리소스가 필요
2. 전통에서 벗어남 : 빈도주의 방법은 통계 커리큘럼에 뿌리 내림 
3. '주관성'의 재평가 : 현대 통계 관행은 계몽주의의 자식. 계몽주의 이상을 반영한 빈도주의적 방법은 주관적인 베이지안 철학에 대한 우월 하고 객관적인 대안 으로 받아들여짐. 
    - "주관적" 레이블은 빈도주의적이든 베이지안이든 모든 통계 분석 에 적용가능
    - 우리의 사전 지식은 우리가 무엇을 측정하고, 왜 측정하고, 어떻게 모델링하는지 자연스럽게 설명. 
    - 계몽주의 이후의 "주관적"은 더 이상 더러운 단어가 아니다.

#### 베이지안 타임라인
- Thomas Bayes가 1740년대에 Bayesian 철학을 개발하기 시작
- 200년 후 1953년 Arianna W. Rosenbluth가 최초의 완전한 Markov 체인 Monte Carlo 알고리즘을 작성, 실제로 베이지안 통계를 수행
- 1969년 David Blackwell이 최초의 베이지안 교과서 중 하나를 도입하여 이러한 방법을 더 많은 청중에게 제공 (Blackwell 1969 )
