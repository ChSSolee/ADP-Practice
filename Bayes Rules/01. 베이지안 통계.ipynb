### 01.01. 베이지안 철학

#### 우리는 경험을 축적하거나 데이터를 수집하면서 세상에 대한 지식을 지속적으로 업데이트

> 예를 들어, 동네에 새로운 이탈리안 레스토랑이 있다고 가정한다면, 레스토랑에 들어서기 전에는 레스토랑이 매우 맛있을 것이라고 예상 
>
> 첫 번째 방문에서 몇 가지 먹을 수 있는 데이터를 수집
>
>파스타 요리가 질척질척하게 엉망으로 도착. 끔찍한 식사(그냥 우연이었을 수도 있음)와 뛰어난 온라인 평가를 비교하면서 지식을 업데이트
>
> 이 레스토랑은 5성급이 아닌 3성급 레스토랑이며, 레스토랑에 또 다른 기회를 주고자 두 번째 방문을 진행. 이번 방문에서 당신은 만족하고 레스토랑 평점을 별 4개로 높임. 
>
> 당신은 계속해서 레스토랑을 방문하여 먹을 수 있는 데이터를 수집하고 매번 지식을 업데이트합니다.

#### 위 그림은 선입견을 인정하고, 데이터를 사용하여 지식을 업데이트하고, 이를 반복하는 자연스러운 베이지안 지식 구축 과정

<br>

#### 이와 동일한 베이지안 프로세스를 엄격한 연구 조사에 적용가능

> 당신이 정치학자라면 투표 패턴의 인구통계학적 요인을 연구
>

> 당신이 환경 과학자라면 기후 변화에서 인간의 역할을 분석

<br>

이전 연구 및 경험을 기반으로 어느 정도 들어오는 정보 또는 사전정보를 가지고 있으며, 당연히, 새로운 데이터를 해석하고 업데이트된 정보나 사후 정보를 개발하는 데 가중치를 두는 것은 이러한 정보에 기반. **새로운 증거를 수집하면서 이 정보를 계속해서 개선**

#### 베이지안 철학은 베이지안 통계의 기초

## 01.02. 베이지안 사고

#### 지식 구축에 대한 베이지안 접근 방식의 자연스러움에도 불구하고, **빈도주의 철학이 전통적으로 통계학를 지배**
* 베이지안 분석과 빈도주의 분석은 우리 주변 세계에 대한 데이터로부터 학습한다는 공통 목표를 공유한다는 점에 유의하는 것이 중요
* 베이지안 분석과 빈도주의 분석은 모두 데이터를 사용하여 모델을 맞추고, 예측하고, 가설을 평가하며, **동일한 데이터로 작업할 때 일반적으로 유사한 결론을 도출**

<br>

#### "베이지안"와 "빈도주의자"라고 부르지만, 실제로 그 차이가 항상 그렇게 명확하지는 않음


<br>

## 01.03. 확률의 의미

#### 베이지안과 빈도주의자는 확률의 의미만큼 근본적인 점에서 서로 다름
- 베이지안과 빈도주의자는 둘 다 공정한 동전 던지기에서 앞면이 관찰될 확률이 1/2이라고 말하지만,

>**베이지안 철학에서 확률은 사건의 상대적 타당성을 측정**
> 
> **빈도주의 철학은 확률을 반복 가능한 사건의 장기적 상대 빈도로 해석**


#### 따라서 동전 뒤집기 예에서 베이지안은 앞면과 뒷면이 동일할 가능성이 있다고 결론을 내리지만,  
#### 빈도주의자는 동전을 계속해서 뒤집으면 대략 1/2의 동전이 앞면이 될 것이라고 결론을 내림

<br>

### 빈도주의의 결함
#### 여론 조사원이 A 후보가 다가오는 선거에서 승리할 확률이 0.9라고 계산하는 경우
- **선거는 일회성 사건이기 때문에 선거를 계속해서 관찰하는 장기적 상대빈도 개념은 적용되지 않음**. 
    - 매우 엄격한 빈도주의적 해석은, 여론 조사원이 틀렸다고 결론을 내릴 수도 있음
- 후보자가 이기거나 질 것이기 때문에 그들의 승리 확률은 1 또는 0이어야 하기에,

    다소 어색하기는 하지만 덜 극단적인 빈도주의적 해석이 더 합리적
    
    **즉, 선거가 장기적으로 반복됨다면, 후보 A가 대략 90%의 비율로 승리할 것**
    
    <br>

- **하지만 대부분의 관심 있는 이벤트가 반복되지 않음**
    
    **정치인이 선거에서 승리하든지, 내일 비가 올지 말지, 화성에 인간이 살 수 있을지 여부는 모두 일회성 사건**
    
    **확률에 대한 빈도주의적 해석은 이러한 일회성 설정에서 어색함**
    
    <br>
    
- **하지만 베이지안 해석은 이러한 사건의 불확실성을 표현하는 경로를 제공**
    
    **예를 들어, 베이지안은 "승리 확률 0.9"를 선거 모델에 기초하여 상대적인 승리 타당성이 높다는 의미로 해석**
    
    **즉, 후보자가 패배할 확률보다 승리할 확률이 9배 더 높다라는 결론**

<br>

#### 베이지안 분석은 사전 지식을 대변

- 개인의 이전 경험이 통계 분석에서 공식적인 역할을 하도록 허용한다는 아이디어는 다소 주관적

    따라서 사전 데이터 (주관)과 일부 데이터를 결합하고 사후 결과를 도출
    
    실제로 베이지안 철학은 그러한 지식 창출을 위한 공식적인 틀을 제공하며, 이는 사전 정보, 데이터 간의 균형 에 따라 달라짐
    
    <br>
    
- 베이지안 분석은 이전 경험과 새로운 데이터의 균형을 맞추며, 설정에 따라 사전에 데이터보다 더 많은 가중치가 부여되거나(왼쪽), 사전에 데이터에 동일한 가중치가 부여되거나(가운데), 사전에 데이터보다 가중치가 덜 부여(오른쪽).



- 사후 구축에서 사전 정보와 데이터 간의 균형은 각각의 상대적 강도에 따라 결정

- 사후가 사전과 데이터의 균형을 맞추도록 허용하는 것은 베이지안 지식 구축 프로세스에 매우 중요

    - **데이터가 거의 없을 때, 우리는 사전 지식의 힘을 활용할 수 있으며 더 많은 데이터를 수집할수록 이전 항목은 영향력을 상실**
    
    <br>

    - **빈도주의 확률 계산은 p-값 과 유사**
    
        - **일반적으로 p-값은, '실제로 원래 가설이 잘못된 경우, 관찰된 데이터가 우리의 데이터와 같거나 그보다 더 극단적일 가능성을 측정'**
        
        - p-값은 수십 년 동안 빈도주의 관행에서 두드러졌지만, 빈도주의와 베이지안 스펙트럼 전반에 걸쳐 점차 덜 강조되고 있음. 
        
        - 본질적으로 이는 매우 일반적으로 잘못 해석되고 오용되는 경우가 많아, 미국 통계 협회(American Statistical Association)는 그 사용에 관해 공식적인 지침을 발표 ( Wasserstein 2016 )
        
        - **p-값이 흔히 잘못 해석되는 이유: 이미 관찰한 데이터의 불확실성(희귀 질환에 대해 양성 반응을 보인 경우)보다, 아직 입증되지 않은 가설의 불확실성(희귀 질환이 있는지 여부)을 더 중시하기 때문**

## 01.06. 베이지안 역사

- 최근 인기가 높아지고 있음에도 불구하고, 베이지안 통계는 통계학자이자 목사이자 철학자인 Thomas Bayes 목사의 18세기 작업에 뿌리를 둠
* 베이즈는 1740년대에 자신의 철학을 발전시켰지만, 이것이 알려진 건 20세기 후반

    그 사이 200년이 넘는 기간 동안 빈도주의 철학이 통계 연구와 실천을 지배
- 베이지안 철학은 이 기간 동안 대중의 지지를 받지 못함

    > 1990년에 Marilyn vos Savant는 현재 고전적인 Monty Hall 확률 퍼즐 에 대한 베이지안 솔루션을 제시했을 때, Parade 잡지 칼럼의 일부 독자들로부터 조롱을 받음. 분명히 10,000명 이상의 독자가 그녀의 솔루션에 대해 이의를 제기하고 조롱하기 위해 글을 썼습니다. 
    >
    > 7 Vos Savant의 베이지안 해법은 나중에 논란의 여지 없이 정확한 것으로 입증되었습니다. 
    
    - 그럼에도 불구하고, 베이지안 기법은 역사속에서 등장 : 
        - Alan Turing은 제2차 세계 대전에서 독일의 Enigma 코드를 해독했으며, John Tukey는 베이지안 방법을 사용하여 1960년대 선거일 예측을 개선
      
      <br>
      
#### 베이지안 기법의 재 등장
- (1) 컴퓨팅의 발전 : 베이지안 애플리케이션에는 1990년대까지 광범위하게 사용할 수 없었던 정교한 컴퓨팅 리소스가 필요
- (2) 전통에서 벗어남 : 일반적으로 사람들이 배우는 내용이기 때문에 빈도주의적 방법이 실제로 뿌리 깊게 자리잡고 있음

    - 사람들이 실제로 사용하는 경우가 많기 때문에 빈도주의적 방법은 통계 커리큘럼에 뿌리 깊게 자리잡고 있음
    
- (3) '주관성'의 재평가 : 현대 통계 관행은 계몽주의의 산물

    - 계몽주의 이상을 반영한 빈도주의적 방법은 주관적인 베이지안 철학에 대한 우월하고 객관적인 대안 으로 받아들여져왔음
    
        이러한 낙인은 여러 가지 이유로 서서히 사라지고 있음. 
        
        첫째, 빈도주의적 분석이든 베이지안적 분석이든 모든 통계분석 에는 "주관적"이라는 라벨이 붙을 수 있음. 
        
        둘째, 계몽주의 이후 '주관적'이라는 단어는 더 이상 더러운 단어가 아님. 

<br>

- Thomas Bayes가 1740년대에 베이지안 철학을 개발
    
    200년 이상이 지난 1953년이 되어서야 Arianna W. Rosenbluth가 최초의 완전한 마르코프 체인 몬테 카를로 알고리즘을 작성하여 실제로 베이지안 통계를 수행 할 수 있게 되었으며 
    
    David Blackwell이 최초의 베이지안 교과서 중 하나를 소개하여 이러한 방법을 더 많은 청중에게 제공한 것은 훨씬 뒤인 1969년 (Blackwell 1969)
