## 선형 회귀(Linear Regression)

<br>

### 단순 선형 회귀 분석(Simple Linear Regression Analysis)

$$y\;=\;wx+b$$

* 독립 변수 $x$와 곱해지는 값 **$w$는 가중치(weight)**, **별도로 더해지는 값 $b$는 편향(bias)**

<br>

### 다중 선형 회귀 분석(Multiple Linear Regression Analysis)

$$y\;=w_1 x_1 + w_2 x_2 + ... w_n x_n\;+b$$

<br>

### 가설 설정 (Hypothesis, $H(x)$)
$$H(x) = wx + b$$

<br>

### 비용 함수(Cost function) : 평균 제곱 오차(MSE)
- **실제값과 예측값에 대한 오차에 대한 식 : 목적 함수(Objective function) 또는 비용 함수(Cost function) 또는 손실 함수(Loss function)**
  - **목적 함수(Objective function)** : 함수의 값을 최소화하거나, 최대화하거나 하는 목적을 가진 함수 
  - **비용 함수(Cost function)  or 손실 함수(Loss function)** : 함수의 값을 최소화하는 함수
- 엄밀히는 같은 의미는 아니지만, 목적 함수, 비용 함수, 손실 함수란 용어를 혼용해서 사용
- **비용 함수는 단순히 실제값과 예측값에 대한 오차를 표현하면 되는 것이 아니라, 예측값의 오차를 줄이는 일에 최적화 된 식이어야 함** 
  - 회귀 문제의 경우에는 주로 평균 제곱 오차(Mean Squared Error, MSE)가 사용

  $$w와\;b에\;의한\;비용함수\;Cost(w, b)\;=\;{1 \over n}∑^n_{i=1} [y^{(i)}-H(x^{(i)})]^2$$

  - 모든 점들과의 오차가 클 수록 MSE는 커지며, 오차가 작아질 수록 MSE는 감소. 즉, $Cost(w,b)$를 최소가 되게 만드는 $w$와 $b$를 구하면 결과적으로 $y$와 $x$와 의 관계를 가장 잘 나타내는 직선을 그릴 수 있음

  $$w, b\;→ minimize\;Cost(w, b)$$

<br>

<br>

### 옵티마이저(Optimizer) : 경사하강법(Gradient Descent)
* 선형 회귀를 포함한 수많은 머신 러닝, 딥 러닝의 학습은 결국 비용 함수를 최소화하는 매개 변수인 $w$와 $b$를 찾기 위한 작업을 수행
- **이때 사용되는 알고리즘은 옵티마이저(Optimizer) 또는 최적화 알고리즘**
- **그리고 이 옵티마이저를 통해 적절한 $w$와 $b$를 찾아내는 과정을 머신 러닝에서 훈련(training) 또는 학습(learning)**

<br>

#### 경사하강법
- **$Cost$와 기울기 $w$와의 관계 :**
  $w$는 가중치라고 불리지만, 직선의 방정식 관점에서 보면 직선의 기울기를 의미
- 아래의 그래프는 기울기 $w$가 지나치게 높거나, 낮을 때 어떻게 오차가 커지는 모습을 보여줌
  - 주황색선은 기울기 $w$가 20일 때, 초록색선은 기울기 $w$가 1일 때의 직선. (각각 $y=20x$, $y=x$에 해당) 
  - ↕는 각 점에서의 실제값과 두 직선의 예측값과의 오차
  - **기울기가 지나치게 크면 실제값과 예측값의 오차가 커지고, 기울기가 지나치게 작아도 실제값과 예측값의 오차가 증가**
  - **$b$ 또한 지나치게 크거나 작으면 오차가 증가**

* 이에 따라 $w$와 $Cost$의 관계를 그래프로 표현할 시
  * **기울기 $w$가 무한대로 커지면 커질 수록 cost의 값 또한 무한대로 커지고, 반대로 기울기 $w$가 무한대로 작아져도 cost의 값은 무한대로 커짐**
  * 위의 그래프에서 cost가 가장 작을 때는 볼록한 부분의 맨 아래 부분입니다. 기계가 해야할 일은 cost가 가장 최소값을 가지게 하는 $w$를 찾는 일이므로, 볼록한 부분의 맨 아래 부분의 $w$의 값을 찾음

* **기계는 임의의 랜덤값 $w$값을 정한 뒤에, 맨 아래의 볼록한 부분을 향해 점차 $w$의 값을 수정** 

* **이를 가능하게 하는 것이 경사 하강법(Gradient Descent)**
  * 그림에서 초록색 선은 $w$가 임의의 값을 가지게 되는 네 가지의 경우에 대해서, 그래프 상으로 접선의 기울기를 보여줌. 
  * 맨 아래의 볼록한 부분으로 갈수록 접선의 기울기가 점차 감소. **맨 아래의 볼록한 부분에서는 결국 접선의 기울기가 0이 됨.** (그래프 상으로는 초록색 화살표가 수평이 되는 지점)
  * **즉, cost가 최소화가 되는 지점은 접선의 기울기가 0이 되는 지점이며, 또한 미분값이 0이 되는 지점**
  * **경사 하강법의 아이디어는 비용 함수(Cost function)를 미분하여 현재 $w$에서의 접선의 기울기를 구하고, 접선의 기울기가 낮은 방향으로 $w$의 값을 변경하고 다시 미분하고 이 과정을 접선의 기울기가 0인 곳을 향해 $w$의 값을 변경하는 작업을 반복하는 것**

  ![선형회귀3[1].JPG]()


<br>

$$Cost(w, b)\;=\;{1 \over n}∑^n_{i=1} [y^{(i)}-H(x^{(i)})]^2$$

<br>

* **비용(Cost)를 최소화하는 $w$를 구하기 위해 $w$를 업데이트하는 식 (접선의 기울기가 0이 될 때까지 반복)**
  - **현재 $w$에서의 접선의 기울기와 $α$를 곱한 값을, 현재 $w$에서 빼서 새로운 $w$의 값으로 한다는 것을 의미**
  - $\alpha$ : 학습률(learning rate). 
$$w\;:=\;w-\alpha {{δ \over {δw}} Cost(w)}$$

*  **접선의 기울기가 음수일 때의 수식** : $w\;:=\;w-\alpha\;(음수 기울기)$
  - 기울기가 음수면 '음수를 빼는 것'은 곧 '해당 값을 양수로 바꾸고 더하는 것'과 동일
  - (가령, 어떤 수에서 -2를 뺀다는 것은 해당 숫자에 2를 더하는 것과 동일) 
  - **결국 음수 기울기를 빼면 $w$의 값이 증가하게 되는데 이는 결과적으로 접선의 기울기가 0인 방향으로 $w$의 값이 조정됨을 의미**


*  **접선의 기울기가 음수일 때의 수식** : $w\;:=\;w-\alpha\;(양수 기울기)$
  - 기울기가 양수면 $w$의 값이 감소하게 되는데 이는 결과적으로 기울기가 0인 방향으로 $w$의 값이 조정됨을 의미

  ![선형회귀4-2[1].JPG]()

<br>

### 학습률 (Learning Rate, $\alpha$)
- **학습률 $\alpha$는 $w$의 값을 변경할 때, 얼마나 크게 변경할지를 결정하며 0과 1사이의 값을 가지도록 함**
  - 학습률은 $w$를 그래프의 한 점으로보고 **접선의 기울기가 0일 때까지 경사를 따라 내려간다는 관점에서는 얼마나 큰 폭으로 이동할지를 결정 (≈ 보폭)**합니다. 
  - **직관적으로 생각하기에 학습률 $\alpha$의 값을 무작정 크게 하면 접선의 기울기가 최소값이 되는 $w$를 빠르게 찾을 수 있을 것같지만 바람직 하지 않음**
    -  **학습률 $\alpha$가 지나치게 높은 값을 가질 때, 접선의 기울기가 0이 되는 $w$를 찾아가는 것이 아니라, **$w$ 의 값이 발산하는 상황을 초래할 수 있음**. 
    - 반대로 학습률 $\alpha$가 지나치게 낮은 값을 가지면 학습 속도가 느려지므로 적당한 $\alpha$의 값을 찾아내는 것도 중요** 
  - 실제 경사 하강법은 $w$와 $b$에 대해서 동시에 경사 하강법을 수행하면서 최적의 $w$와 $b$의 값을 탐색

## 자동미분

<br>

#### `tensorflow.GradientTape()` : 자동미분 수행 객체

<br>

- $2w^2 + 5$의 식에서, $w$에 대해 미분

```python
import tensorflow as tf
```

```python
w = tf.Variable(2.)

def f(w):
  y = w**2
  z = 2*y + 5
  return z
```

- `Gradienttape`객체의 `gradients`를 통해 $w$에 대해 미분한 값 저장된 것을 확인 가능

```python
with tf.GradientTape() as tape:
  z = f(w)

gradients = tape.gradient(z, [w])
print(gradients)
```

## 자동미분을 이용한 선형 회귀 구현
- 가중치 변수 $w$와 $b$를 선언 (각각 4, 1)

```python
w = tf.Variable(4.0)
b = tf.Variable(1.0)
```

```python
@tf.function
def hypothesis(x):
  return w*x + b
```

```python
x_test = [3.5, 5, 5.5, 6]
print(hypothesis(x_test).numpy())
```

<br>

- 손실함수 정의 (MSE)

<br>

#### `tensorflow.reduce_mean()` : 특정 차원을 제거하고 평균 계산
#### `tensorflow.reduce_sum()` : 특정 차원을 제거하고 합계 계산


```python
@tf.function
def mse_loss(y_pred, y):
  return tf.reduce_mean(tf.square(y_pred - y))
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [11, 22, 33, 44, 55, 66, 77, 87, 95]
```

<br>

- 경사 하강법 옵티마지어 적용 (학습률 : 0.01)

<br>

#### `tensorflow.optimizers.SGD`(학습률)

```python
optimizer = tf.optimizers.SGD(0.01)
```

<br>

- 약 300번에 걸쳐서 경사 하강법 수행
  * $w$와 $b$값이 업데이트 됨에 따라, Cost가 지속적으로 감소하는 것을 확인 가능

```python
for i in range(301):
  with tf.GradientTape() as tape:
    # 현재 파라미터에 기반한 입력 x에 대한 예측값을 y_pred
    y_pred = hypothesis(x)

    # 평균 제곱 오차를 계산
    cost = mse_loss(y_pred, y)

  # 손실 함수에 대한 파라미터의 미분값 계산
  gradients = tape.gradient(cost, [w, b])

  # 파라미터 업데이트
  optimizer.apply_gradients(zip(gradients, [w, b]))

  if i % 10 == 0:
    print("epoch : {:3} | w의 값 : {:5.4f} | b의 값 : {:5.4} | cost : {:5.6f}".format(i, w.numpy(), b.numpy(), cost))
```

<br>

- 학습된 $w$와 $b$값에 대해서 임의의 입력값을 넣었을 경우의 예측값

```python
print(hypothesis(x_test).numpy())
```
