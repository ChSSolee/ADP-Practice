### `Raster` 대수학

```r
r <- rast(ncol = 10, nrow = 10)
values(r) <- 1:ncell(r)
```

- `Raster`객체 값에 대한 연산

```r
s <- r + 10
s <- sqrt(s)
s <- s * r + 5
s
```

```r
values(r) <- runif(ncell(r))
r <- round(r)
r <- r == 1

r
```

<br>

- 값 대체

```r
s[r] <- -0.5
s[!r] <- 5
s[s == 5] <- 15

s
```

<br>

- 여러 `SpatRaster` 객체를 사용하는 경우, **동일한 해상도와 원점을 가져야 함**
- **`Raster` 객체의 원점은 `SpatRaster`의 모서리에서, 입력된 x 및 y 해상도 기준, 해당 지점을 향해 이동하면 얻을 수 있는 (0, 0)에 가장 가까운 지점**

- **동일한 범위를 갖지 않는 경우, 반환 값은 사용된 객체들의 공간 교차점을 포함하게 됨**



```r
r <- rast(ncol=5, nrow=5)
values(r) <- 1
r[[1]]
dim(r)
```

```r
s <- c(r, r+1)
s
dim(s)
```

```r
q <- c(r, r+2, r+4, r+6)
dim(q)
```

```r
x <- r + s + q
dim(x)
```

<br>

- 요약함수 (`min`, `max`, `mean`, `prod`, `sum`, `median`, `cv`, `range`, `any`, `all`)는 항상 `SpatRaster` 객체를 반환

```r
a <- mean(r, s)
dim(a)
a
```

<br>

- **서로 다른 레이어를 가진 다중 레이어 객체들을 사용하면, 가장 차원 값을 가진 객체가 기준이 됨**

```r
st <- c(r, s, a)
sst <- sum(st)
sst; dim(sst)
```

#### 도움 함수
- `Raster`데이터는 행렬로 간주되지만, `SpatRaster` 벡터로 간주
- 셀 번호는 왼쪽 상단 부터 오른쪽 상단까지 매겨지며, 다음 행의 왼쪽에서 동일 방식이 적용

<br>

#### `rowFremCell()` / `colFromCell()` / `cellFromRowCol()` / `xyFromCell()` / `cellFromXY` / `cellFromX` / `cellFromY`: 셀의 열 또는 행 번호를 확인하는 도움 함수

- 5행, 5열의 값

```r
cellFromRowCol(r,5,5)
```

- 100번째 셀에 대한 좌표

```r
xyFromCell(r, 100)
```

- (0, 0) 좌표에 대한 셀 번호

```r
cellFromXY(r, cbind(0,0))
```

```r
colFromX(r, 0); rowFromY(r, 0)
```
