# 05. 잠재변수가 있는 구조방정식 모형

<br>

## 05.01. 구조방정식모형의 구성
- **관찰변수 / 측정변수** : 측정의 정의가 명확하게 내려져 잇어, 그 값들이 직접 관찰된 변수
* **잠재변수 : 직접 측정이나 관찰이 불가능하고 관찰변수들의 저변에 잠재되어 있는 변수**

<br>

### 측정모형과 구조모형
- 잠재변수가 있는 구조방정식 모형의 체제는 관찰변수와 잠재변수들 간의 인과관계를 나타내는 측정모형과 잠재변수들 간의 인과관계를 나타내는 구조모형으로 구분
- **측정모형은 잠재변수들이 관찰변수들을 통해 어떻게 측정되어지는가를 규정하며, 관찰된 변수들이 추상적/가설적 개념을 구축함에 있어 측정변수로서의 특성들이 관심대상에 포함**
    - **측정과정에서 불가피하게 수반되는 측정오차 등이 반영**
    * 구조방정식모형을 설정함에 있어 측정모형의 타당성이 일차적으로 중요
    
<br>

### 잠재변수 규정시 주의사항
- 구조방정식모형에서는 관찰변수들을 통해 잠재변수를 비추어보는 형식의 우회적 접근이 이루어짐
    - 관찰변수를 잠재변수에 대한 지지자 (indicator)
    
1. **잠재변수는 단순히 관찰변수들의 묶음을 의미하거나 관찰변수들에 대한 설명이 아님**
    
2. **잠재변수는 단일한 개념(의미)을 나타내도록 규정되어야 함**
    - 변수들이 각각 서로 다른 개념들을 측정할 수 있기 때문
    - 복합적 의미를 가지는 잘못된 잠재변수 : [행복감, 자신감, 우울, 불안] $\rightarrow$ 심리상태

3. 잠재변수들 간의 인과(연관)관계를 묘사하는 경로계수들은 방향성을 가지고 해석할 수 있어야 하므로, 크기를 말할 수 있어야 함
    - 크기가 정의되지 않는 잘못된 잠재변수 : [성별, 나이, 직업, 학력] $\rightarrow$ 일반적 특성

4. 반영적 지표라는 개념과 관련하여, 하나의 잠재변수를 규정하는 관찰변수들은 동일한 개념을 측정한다고 말할 수 있어야 함
    - [성별, 나이] $\rightarrow$ 일반적으로 서로 관련성이 없금 $\rightarrow$ 동일한 잠재변수를 측정한다고 주장할 수 없음
    
<br>

### 반영적 지표와 조형적 지표
- 잠재변수로부터 관찰변수 방향으로 화살표가 존재 $\rightarrow$ 관찰변수의 의미가 반영적 지표 (reflective indicator)
- 관찰변수로부터 잠재변수 방향으로 화살표가 존재 $\rightarrow$ 잠재변수에 대한 관찰변수의 의미가 조형적 지표 (formative indicator)

<br>

#### 반영적 지표 (reflective indicator)
- **관찰변수의 값이 잠재변수의 영향을 받아 결정되는 구조**
    - **잠재변수가 관찰변수의 원인임을 가정**
- **관찰변수의 값이 잠재변수들의 영향으로 인하여 변화될 수 있기에, 관찰변수들 사이에 상관관계가 높을 수 있음**
    - 관찰변수의 변화가 잠재변수인 개념을 변화시키지 않는다는 특징
- **관찰변수들은 해당 잠재변수의 함수이므로, 동일한 잠재변수의 관찰변수 값들 사이에는 내적 일관성이 기대됨**
    - **측정모형과 관련된 계수들의 통계적 유의성은 구조방정식모형의 주요 관심 대상**
    
- 전통적인 형태의 구조방정식모형에서는 잠재벼순가 모두 반영적 지표임을 가정
    
    (잠재변수들이 반영적 지표만으로 구성된 모형을 '인자-기반 구조방정식모형')
    
    - **이 경우 잠재변수는 요인(factor)의 의미를 가짐**

<br>

**반영적 지표**

$$\begin{cases}
국어점수 = \lambda_1 언어능력+\delta_1\\
영어점수 = \lambda_2 언어능력+\delta_2\\
한문점수 = \lambda_3 언어능력+\delta_3\\
\end{cases}$$

- 각 $\lambda$는 각 관찰변수가 잠재변수와 얼마나 관련되어 있는지를 나타내며, (인자적재값 (factor loading)또는 요인 부하량) 

    각 $\delta$는 관찰변수의 측정값 중 잠재변수와 구별되는 부분 (측정오차 등)을 나타내며, 해당 관찰변수의 특수성 또는 유일성으로 부름

<br>

#### 조형적 지표 (formative indicator)
- **관찰변수가 잠재변수의 값을 결정하는 구조**
    - **관찰변수가 잠재변수를 형성하는 원인임을 가정**
- **관찰변수의 변화에 잠재변수가 변화하지만, 잠재변수의 변화가 관찰변수의 변화를 수반하지 않음** 
- **잠재변수에 의해 공통적으로 영향을 받아 관찰변수의 값이 결정되는 것이 아니기에, 관찰변수들 간의 높은 상관관계와 내적일치도를 반드시 필요로 하지 않음**
- **잠재변수가 조형적 지표로 구성되는 대표적인 모형으로는 PLS (partial least squares)가 있으며, 조형적 지표가 포함된 모형을 '성분-기반 구조방정식모형'**
    - 이 경우 잠재변수는 성분, 지수, 점수의 의미를 가짐
    
<br>

**조형적 지표**

$$SES(사회경제적\;지위) = w_1 소득수준 + w_2 직업수준 + w_3 학력수준$$

- **잠재변수는 관찰변수들의 가중합으로 정의되며, 여기에서 각 $w$는 각 관찰변수가 붙는 가중치를 나타냄**
- 사전에 정의된 가중치를 이용하여 잠재변수의 값을 계산하고, 이후의 분석에서 하나의 점주(성분)로 취급하여 사용



<br>

### 측정모형과 관찰변수의 측정



- (a)에서 키에 대한 한 번의 측정값 또는 행복감의 한 문항에 대한 측정값은 신뢰할 수 없기에 회귀분석, 분산분석, 구조방정식모형과 같은 통계분석의 직접적인 대상이 될 수 없음
- **구조방정식모형은 관찰변수(측정변수)로 설정된 변수들의 분산과 공분산이 직접적인 분석대상이 되므로, (a)와 같이 설정하면 매우가 불안정**

## 05.02. 초기모형

<br>


- **외생 잠재변수 (Exogenous Latent Variable)** : ACHMOT (성취동기), TSSE (직무관련 자존감), VERBINT (대화능력)
- **내생 잠재변수 (Endogenous Latent Variable)** : PERFORM (업무능력), JOBSATIS (직무 만족도)
- **관찰변수** : 
    - Achmot (성취동기) : 성취동기1, 성취동기2, 성취동기3
    - Tsse (직무관련 자존감) : 직무관련 자존감1, 직무관련 자존감2
    - Verbint (언어적 기능)
    - Perform (업무능력)
    - Jobsatis (직무만족도)
- **내생 오차변수** : $\zeta_1, \zeta_2$
- **측정 오차변수** : $\delta_1, \delta_2, \delta_3, \delta_4, \dots, \epsilon_1, \epsilon_2. \epsilon_3$



<br>



**잠재변수들의 인과관계**

$$\begin{cases}
PERFORM = \gamma_{12}TSSE + \gamma_{13}VERBINT+\zeta_{1}\\
JOBSATIS = \gamma_{21}ACHMOT + \gamma_{23}VERBINT+\zeta_{2}\\
\end{cases}$$

```r
library(lavaan)
library(sem)
library(semPlot)
library(semptools)
library(psych)
library(tidyverse)
```

```r
detach("package:sem",unload=TRUE)
```

```r
Jobsatis <- read.csv(paste0(getwd(), "/Data/Jobsatis.csv"), header = TRUE)
Jobsatis %>% head
```

```r
GGally::ggpairs(Jobsatis)
```

```r
Jobsatis.A <- "
    # easurement
    ACHMOT = ~ Achmot1 + Achmot2 + Achmot3
    TSSE = ~ Tsse1 + Tsse2
    VERBINT = ~ Verbint1
    PERFORM = ~ Perform1
    JOBSATIS = ~ Jobsatis1 + Jobsatis2 + Jobsatis3
    
    # regression
    PERFORM ~ TSSE + VERBINT
    JOBSATIS ~ ACHMOT + VERBINT
    
    # error covariance (PERFORM과 JOBSATIS의 편상관이 존재하지 않음을 가정하므로, 편상관이 0이라는 것을 설정)
    PERFORM ~~ 0 * JOBSATIS
"
```

```r
Jobsatis.A.fit <- sem(Jobsatis.A, Jobsatis)
```

* 일부 변수들의 분산이 음의 값으로 추정됨 $\rightarrow$ 추정치가 이론적이느 범위를 넘어서는 상황 'Heywood 상황'
    - 어떤 변수의 분산이 음수가 되는 경우
        - 오차변수의 분산이 음수로 추정되는 경우, 오차변수의 분산을 아주 작은 수치로 고정하여 분석 (0.001)
        - 이 경우, 적합도나 추정치가 좋지 않은 결과를 보일 수 있음
    - 상관계수의 절댓값이 1보다 큰 경우
    - SMR가 1을 넘는 경우 등
    - 원인은 다양하지만, 모형이 데이터를 잘 표현하지 못하는 경우에 많이 나타남
    

<br>

## 05.03. 잠재변수에 대한 측정단위 규정

- 외생 잠재변수에 대한 측정모형

$$\begin{cases}
Achmot1 = \lambda^x_{11} ACHMOT + \delta_1\\
Achmot2 = \lambda^x_{21} ACHMOT + \delta_2\\
Achmot3 = \lambda^x_{31} ACHMOT + \delta_3\\
Tsse1 = \lambda^x_{42} TSSE + \delta_4\\
Tsse2 = \lambda^x_{52} TSSE+ \delta_5\\
Verbint1 = \lambda^x_{63} VERBINT + \delta_6\\
\end{cases}$$

<br>

- 내생 잠재변수에 대한 측정모형

$$\begin{cases}
Perform1 = \lambda^y_{11}PERFORM + \epsilon_{1}\\
Jobsatis1 = \lambda^y_{22}JOBSATIS + \epsilon_{2}\\
Jobsatis2 = \lambda^y_{32}JOBSATIS + \epsilon_{3}\\
Jobsatis3 = \lambda^y_{42}JOBSATIS + \epsilon_{4}\\
\end{cases}$$

<br>

#### $\lambda^x_{ij}$와 $\lambda^y_{ij}$는 인자적재값 (factor loading)
- 관찰변수와 잠재변수가 가지는 연관관계의 정도를 나타냄

<br>

**잠재변수는 관찰되지 않은 것이기 때문에, 정해진 척도(scale)를 가지고 있지 않으며 그 측정단위가 임의적**
- 그러나 관련된 모수를 추정하고 경로계수의 상대적 비교나 모형의 적절한 해석을 위해서는 각 잠재변수를 특정한 단위로 규정하거나, 측정단위를 없앨 필요가 있음

<br>

#### 잠재변수에 대한 측정단위를 규정하기 위해서는 일반적으로 2가지 방법 중 하나 사용

#### 1. 인자적재값을 1로 고정
* 잠재변수의 측정단위를 규정하기 위해서 관찰변수 중의 어느 하나를 그 잠재변수에 대한 참조변수로 두어 해당하는 인자적재값을 0이 아닌 1로 고정하는 것
    - **이는 잠재변수의 측정단위가 참조 관찰변수의 측정단위와 같다고 가정하는 것**
    - 잠재변수의 측정단위가 서로 다른 경우, 잠재변수들 간의 인과관계를 나타내는 회귀게수들의 크기를 상대비교하기가 여려울 수도 있음 $\rightarrow$ 표준화추정치 (standardized estimate)에 의해서 추정치들을 해석
    
- **`sem()`함수에서는 기본적으로 인자적재값을 1로 고정시키는 방법을 적용**

```r
print(summary(Jobsatis.A.fit, fit.measures = TRUE, rsquare = TRUE, standardized = TRUE))
```

<br>

#### 2. 잠재변수들의 표준화
- 잠재변수의 분산이 1이 되도록 표준화되어 있다고 가정하여, 측정단위 자체를 없애버림으로써, 잠재변수가 취하는 값이 단순한 숫자가 되도록 하는 것
<br>

* `sem()`함수에 이를 적용할 때, 문제가 발생하여 해를 구하지 못하는 경우 존재

```r
Jobsatis.A.fit <- sem(Jobsatis.A, data= Jobsatis, std.lv = TRUE)
```

<br>

#### 어떤 방식이든 잠재변수의 측정단위를 규정하지 않으면, 추정치를 구할 수 없게 되는 식별불능 문제에 빠지게 됨

## 05.04. 단일 지시자 문제
- 측정모형이 식별되기 위해서는 한 잠재변수당 2개 이상의 관찰변수가 규정되어야 함
    - 그러나 현실적으로 하나의 잠재변수를 1개의 관찰변수만으로 규정해야 하는 경우가 존재 $\rightarrow$ 단일지시자 (single indicator)문제
    - 이 경우에도 식별불능의 문제 발생 가능
    
- 단일 지시자 문제 발생 경우, 보통 오차변수의 분산을 특정 값으로 고정하여 식별문제를 해결

<br>

### 1. 오차변수의 분산을 0으로 고정
- **`sem()`함수에서는 단일지시자 문제에서 기본적으로 오차변수의 분산을 0으로 고정**
- **잠재변수 VERBINT와 PERFORM에 대한 측정오차가 0으로 가정함 $\rightarrow$ 두 잠재변수에 대해 관찰변수와 잠재변수가 실제 동일한 것으로 간주**

```r
(summary(Jobsatis.A.fit, fit.measures = TRUE, rsquare = TRUE, standardized = TRUE))$pe %>%
    filter(op == "~~") %>%
    filter(lhs %in% c("Verbint1", "Perform1"))
```

### 측정의 신뢰도를 이용
- 관찰변수가 여러 항목들의 합(또는 평균)으로 측정된 경우 신뢰도(reliability) $\alpha$가 주어질 수 있음
    - 신뢰도의 정의에 의하면 $1-\alpha$는 측정값에서 측정오차가 차지하는 비중을 의미하므로, 측정오차의 분산을 다음과 같이 추정
    
$$측정오차의\;분산 = (1-\alpha) \times 관찰변수의\;분산$$

<br>

* 예를들어, 잠재변수 VERBINT를 측정하는 관찰변수 Verbint1의 여러 문항들의 합으로 측정되었고, 이에 대한 측정의 신뢰도가 $\alpha =0.85$라고 가정 시
    - 변수 Verbint1의 분산은 3.393이므로, 대응되는 오차변수 $\delta_6$의 분산을 다음과 같이 추정
    
    $$\hat{Var}(\delta_6)=(1-\alpha)\times\hat{Var}(Verbint1) = (1-0.85)\times3.393=0.509$$

```r
sd(Jobsatis$Verbint1)^2
```

<br>

* 관찰변수 Verbint1에 대응되는 오차변수 $\delta_6$에 대한 분산을 0.509로 고정하고, Perform1에 대응되는 오차변수 (\epsilon_1)에 대한 분산을 0으로 고정하는 예

```r
Jobsatis.A <- "
    # easurement
    ACHMOT = ~ Achmot1 + Achmot2 + Achmot3
    TSSE = ~ Tsse1 + Tsse2
    VERBINT = ~ Verbint1
    PERFORM = ~ Perform1
    JOBSATIS = ~ Jobsatis1 + Jobsatis2 + Jobsatis3
    
    # regression
    PERFORM ~ TSSE + VERBINT
    JOBSATIS ~ ACHMOT + VERBINT
    
    # error covariance
    PERFORM ~~ 0 * JOBSATIS

    # error variance
    Verbint1 ~~ 0.509 * Verbint1
    Perform1 ~~ 0 * Perform1
"
```

```r
Jobsatis.A.fit <- sem(Jobsatis.A, data = Jobsatis)
semPaths(Jobsatis.A.fit, whatLabels = "std", style = 'lisrel', nCharNodes = 0, rotation = 2)
```

## 05.06. 확증적 인자분석을 이용한 측정모형 평가
- 관찰변수와 잠재변수의 수가 많고, 모형이 복잡한 구조를 가지는 경우, 모형의 적합도가 나쁠 때, 모형수정을 위해 어떤 부분을 먼저 검토해야 하는지 판단하기 어려운 경우가 존재
    **$\rightarrow$ 측정모형과 구조모형을 구분하여 차례로 검토**
    
<br>

#### 확증적 인자모형 (confirmatory factor model)
- **잠재변수들을 모두 외생변수로만 고려하는 모형** 
    - **따라서 모든 잠재변수들 간에 연관관계(양방향 화살표)로 설정되는 모형을 의미**
    


- 적합도의 관점에서는 잠재변수들 간에 연관관계 (양방향 화살표)를 설정하든 인과관계 (단방향 화살표)를 설정하든 차이가 없음

- 확증적 인자모형에서는 모든 잠재변수들이 서로 화살표로 연결되어 있으므로, 구조모형에서는 더 이상의 적합도 증가를 기대할 수 없게됨
    - 확증적 인자모형의 적합도가 나쁘게 나타난다면, 측정모형 부분만을 살펴봄으로써 모형수정의 범위를 좁힐 수 있음
    - 그리고 측정모형 부분에서 문제점을 수정한 후, 구조모형을 다시 원래대로 설정하고 수정지수 등을 이용하여 구조모형 부분에서의 문제점들을 검토
    
- **확증적 인자분석 (CFA)를 이용하면, 측정모형을 평가하고 수정하는 작업을 조금더 효율적으로 진행 가능**
    * 간단한 모형에 대해서는 이러한 절차를 수행하지 않아도 모형수정을 쉽게 진행할 수 있음

## 05.10. 정규분포를 따르지 않는 데이터의 경우
- 구조방정식모형에 대한 계수를 추정할 때 기본이 되는 추정법은 최대우도(ML)이며,

    이 방법은 관찰변수들이 모두 정규분포를 따를 때 타당한 방법
    
- 정규분포로부터의 작은 이탈에는 최대우도법이 잘 작동하나, 관찰변수들 중 일부가 정규분포로부터 심하게 이탈되어 있는 경우, 최대우도법은 올바르지 않을 수 있음

- **외생 잠재변수 (Exogenous Latent Variable)** : Expect (기대수준)
- **내생 잠재변수 (Endogenous Latent Variable)** : Quality (인지품질), Value (인지가치), Satisfact (고객만족)
- **관찰변수** : 
    - Expect (기대수준) : f1_1(품질기대), f1_2(요구기대), f1_3(문제기대)
    - Quality (인지품질) : f2_1 (품질평가), f2_2 (요구평가), f2_3 (문제평가)
    - Value (인지가치) : f3_1 (가격가치), f3_2 (품질가치)
    - Satisfact (고객만족) : f4_1 (전반적 만족), f4_2 (기대 대비 만족), f4_3 (이상 대비 만족)
- **내생 오차변수** : $\zeta_1, \zeta_2, \zeta_3$
- **측정 오차변수** : $\delta_1, \delta_2, \delta_3, \epsilon_1, \epsilon_2. \epsilon_3\dots, $

<br>

### WLS (가중최소제곱) 방법
* **관찰변수들에 대하여 특별한 분포를 가정하지 않는 비모수적 방법**

<br>

#### `sem()`의 `estimator`을 `"WLS"`로 입력

```r
Satisfact.A <- "
    # measurement model
    Expect =~ f1_1 + f1_2 + f1_3
    Quality =~ f2_1 + f2_2 + f2_3
    Value =~ f3_1 + f3_2
    Satisfact =~ f4_1 + f4_2 + f4_3
    
    # regression
    Quality ~ Expect
    Value ~ Expect + Quality
    Satisfact ~ Expect + Quality

    # error covariance
    Value ~~ 0 * Satisfact
"
```

### 측정 오차변수들 간의 연관관계
- 수정지수들을 기초로 오차변수들 간의 연관관계 3개가 추가

* **f1_1과 f2_1, f1_2과 f2_2, f1_1과 f2_1의 상관관계가 높음 $\rightarrow$ 관찰변수들의 측정오차 간에 서로 연관관계가 존재할 가능성 존재**

```r
cor(Satisfact %>% select(-Group))
```

<br>

**'$\delta_1 \leftrightarrow \epsilon_1$, $\delta_2 \leftrightarrow \epsilon_2$, $\delta_3 \leftrightarrow \epsilon_3$과 'Satisfact $\leftarrow$ Value'**를 추가하여 수정 모형 구축

```r
Satisfact.B <- "
    # measurement model
    Expect =~ f1_1 + f1_2 + f1_3
    Quality =~ f2_1 + f2_2 + f2_3
    Value =~ f3_1 + f3_2
    Satisfact =~ f4_1 + f4_2 + f4_3
    
    # regression
    Quality ~ Expect
    Value ~ Expect + Quality
    Satisfact ~ Expect + Quality + Value

    # error covariance
    f1_1 ~~ f2_1
    f1_2 ~~ f2_2
    f1_3 ~~ f2_3
"
```

```r
Satisfact.B.fit <- sem(Satisfact.B, Satisfact, estimator = "WLS")
semPaths(Satisfact.B.fit, whatLabels = "std", style = "lisrel", nCharNodes = 0, rotation = 2)
```

* 적합도 지수들이 향상

```r
tidy(fitMeasures(Satisfact.B.fit)) %>%
    filter(names %in% c("chisq", "df", "gfi", "agfi", "cfi", "tli", "rmsea", "srmr"))
```

- 인자적재값 : 모든 인자적재값의 표준화추정치는 0.5보다 크며, 통계적으로 유의

    (인자적재값이 유의하지 않으면 측정의 신뢰도가 의심)

```r
summary(Satisfact.B.fit, fit.measures = TRUE, rsquare = TRUE, standardized = TRUE)$pe %>%
    filter(op == '=~')
```

* 관찰변수에 대한 모든 SMC가 0.3보다 높음

```r
summary(Satisfact.B.fit, fit.measures = TRUE, rsquare = TRUE, standardized = TRUE)$pe %>%
    filter(op == 'r2') %>% filter(str_detect(lhs, "_"))
```

* 오차변수의 연관관계 : 3개의 경로에 대해서 공분산 추정치가 통계적으로 유의하며, 상관계수가 0.2이상의 값을 가짐

```r
summary(Satisfact.B.fit, fit.measures = TRUE, rsquare = TRUE, standardized = TRUE)$pe %>%
    filter(op == '~~') %>% head(3)
```

- 구조모형의 회귀계수 추정치 : 통계적으로 유의하지 않은 계수 존재 $\rightarrow$ 모형 수정 고려 가능

```r
summary(Satisfact.B.fit, fit.measures = TRUE, rsquare = TRUE, standardized = TRUE)$pe %>%
    filter(op == '~')
```

### 내생 오차변수들 간의 연관관계
- **내생 오차변수들 간의 연관관계 : 오차들 간의 연관성이라는 단순한 의미가 아닌, 관련된 설명변수들이 통제된 상황에서, 내생변수들 간의 편상관으로 해석되어야 함**



* **위 모형은 'Satisfact $\leftarrow$ Value' 대신, '$\zeta_2 \leftarrow \zeta_3$'가 설정되어 있음**
    
    **'Satisfact $\leftarrow$ Value'가 설정된 모형과 위 모형은 동치관계 $\rightarrow$ 동일한 적합도 지수**

```r
Satisfact.C <- "
    # measurement model
    Expect =~ f1_1 + f1_2 + f1_3
    Quality =~ f2_1 + f2_2 + f2_3
    Value =~ f3_1 + f3_2
    Satisfact =~ f4_1 + f4_2 + f4_3
    
    # regression
    Quality ~ Expect
    Value ~ Expect + Quality
    Satisfact ~ Expect + Quality

    # error covariance
    Value ~~ Satisfact
    f1_1 ~~ f2_1
    f1_2 ~~ f2_2
    f1_3 ~~ f2_3
"
```

```r
Satisfact.C.fit <- sem(Satisfact.C, Satisfact, estimator = "WLS")
semPaths(Satisfact.C.fit, whatLabels = "std", style = "lisrel", nCharNodes = 0, rotation = 2)
```

```r
tidy(fitMeasures(Satisfact.C.fit)) %>%
    filter(names %in% c("chisq", "df", "gfi", "agfi", "cfi", "tli", "rmsea", "srmr"))
```

* Value $\sim\sim$ Satisfact의 공분산과 편상관 계수 : 두 오차변수 $\zeta_2, \zeta_3$간의 상관계수 0.487은 대응되는 내생변수 간의 편상관(연관관계)로 해석

```r
summary(Satisfact.C.fit, fit.measures = TRUE, rsquare = TRUE, standardized = TRUE)$pe %>%
    filter(op == '~~') %>% head(1) %>% select(lhs, op, rhs, est, std.lv)
```

* 회귀계수 추정치와 SMC는 조금씩 달라짐
