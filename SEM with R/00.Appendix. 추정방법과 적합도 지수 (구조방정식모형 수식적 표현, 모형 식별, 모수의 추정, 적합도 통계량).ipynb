## A.01. 구조방정식모형의 수식적 표현

<br>

### 잠재변수가 없는 모형

<br>

#### 방정식 체계의 작성
- 외생변수와 내생변수들 간의 인과관계를 기호를 사용하여 여러 개의 선형방정식들로 표현
- **선형방정식의 개수는 일반적으로 내생변수들의 개수와 돋일**

<br>

**방정식체계를 만드는 단계 및 규칙**

1. 한 개의 내생변수를 방정식의 등호 왼쪽에 위치
2. 등호 횐쪽의 내생변수에 영향을 주는 외생변수 및 내생변수의 각 변수에 대한 계수를 곱한 항들을 더하여 방정식의 오른쪽에 위치
3. 방정식 왼쪽의 내생변수에 대응되는 오차변수를 (2)에 더함
4. 위 과정을 모형에 포함되어 있는 모든 내생변수에 대해서 수행

<br>

#### 경로분석모형의 모수행렬

| 행렬구분 | 행렬 | 이름 | 내용 |
| - | - | - | - |
| 계수행렬  | $\Gamma$ | gamma | 외생관찰변수 $x$와 내생관찰변수 $y$간의 인과관계에 대응되는 경로계수 $\gamma$들로 이루어진 행렬 |
|  | $B$ | beta | 내생관찰변수 $y$들 간의 인과관계에 대응되는 경로계수 $\beta$들로 이루어진 행렬 |
| |
| 공분산 행렬 | $\Phi$ | phi | 외생관찰변수 $x$들의 분산 및 공분산으로 이루어진 행렬 |
|  | $\Psi$ | psi | 오차변수 $\zeta$들의 분산 및 공분산으로 이루어진 행렬 |

<br>

$$\begin{cases}
y_1\;=\;\gamma_{12}x_2 + \;\zeta_1\\
y_2\;=\;\gamma_{21}x_2 + \beta_{21}y_1 + \;\zeta_2\\
y_3\;=\;\gamma_{31}x_1 + \beta_{31}y_1 + \beta_{32}y_2 + \zeta_3\\
\end{cases}$$

<br>




<br>

#### 계수행렬의 표현
- **방정식체계에서 경로계수들만을 따로 분리하여 행렬 형태로 표현한 것을 계수행렬 (coefficient matrix)라고 하며, 경로분석모형에서는 일반적으로 두 개의 계수행렬 $\Gamma$와 $B$가 존재**
- 위 방정식에서 내생변수 $y_1, y_2, y_3$에 대한 외생변수 $x_1, x_2$의 인과관계를 나타내는 $3\times2$의 $\Gamma$-행렬

$$\Gamma = \begin{pmatrix}0&\gamma_{12}\\0&\gamma_{22}\\\gamma_{31}&0 \end{pmatrix}$$

- 내생변수 $y_1, y_2, y_3$들 간의 인과관계는 크기 $3\times3$ $B$-행렬로 표현

$$B = \begin{pmatrix}0&0&0\\\beta_{21}&0&0\\\beta_{31}&\beta_{32}&0 \end{pmatrix}$$

<br>

- **$\gamma_{12}$또는 $\beta_{21}$ 등은 연구자의 모형에 포함되어 있어서, 데이터에 기초한 통계적 추론의 대상이 되는 모수들로서 자유모수 (free parameter)**
    - 이런 모수들은 그 값 또는 범위가 고정되거나 제한되어 있지 않음
- **$\gamma_{11}$ 또는 $\beta_{12}$등은 연구자의 가설에 의해 0으로 고정되어 있어서, 데이터를 통한 추정에서 제외되는 모수들로 고정모수 (fixed parameter)**

<br>

#### 공분산행렬의 표현
- 구조방정식모형에는 외생변수 또는 오차변수들 간의 연관관계가 포함되며, 경로분석모형에서 해당 변수들의 변이정도를 나타내는 분산 (variance) 과 연관관계를 나타내는 공분산 (covariance)는 두 개의 공분산행렬 $\Phi$와 $\Psi$로 표현

$$\Phi = \begin{pmatrix}\phi_{11}&\\\phi_{21}&\phi_{22} \end{pmatrix}$$

$$\Psi = \begin{pmatrix}\psi_{11}&&\\0&\psi_{22}&\\0&0&\psi_{33} \end{pmatrix}$$

<br>

- $\phi_{ii}=Var(x_i)$, $i=1, 2$, $\phi_{21}=Cov(x_2, x_1)$, $\psi_{ii}=Var(\zeta_i)$, $i=1, 2, 3$
    - **구조방정식모형에서는 일반적으로 모든 외생변수들은 서로 연관되어 있는 것으로 가정 $\rightarrow$ 모든 $\Phi$-행렬의 모든 원소들이 자유모수**
    - 위 예시에서 오차변수들은 서로 연관되어 있지 않은 것으로 가정 $\rightarrow$ $\Psi$-행렬의 모든 비대각원소들은 0으로 고정
    
        (연구자의 가설에 따라서 오차변수들 간의 연관성으로 고려할 수 있음)

<br>

### 잠재변수가 있는 모형

<br>

#### 방정식 체계의 작성

<br>

**구조방정식모형의 모수행렬**

| - | - | - | - |

|  | $\Gamma$ | gamma | 외생잠재변수 $\xi$와 내생잠재변수 $\eta$간의 인과관계에 대응되는 경로계수 $\gamma$들로 이루어진 행렬 |
|  | $B$ | beta | 내생잠재변수 $\eta$들 간의 인과관계에 대응되는 경로계수 \$beta$들로 이루어진 행렬 |
| |

|   |   |   |   |
| - | - | - | - |
| 공분산 행렬 | $\Phi$ | phi | 외생잠재변수 $\xi$들의 분산 및 공분산으로 이루어진 행렬 |
|  | $\Psi$ | psi | 오차변수 $\zeta$들의 분산 및 공분산으로 이루어진 행렬 |
|  | $\Theta_{\delta}$ | theta delta | 오차변수 $\delta$들의 분산 및 공분산으로 이루어진 행렬 |
|  | $\Theta_{\epsilon}$ | theta epsilon | 오차변수 $\epsilon$들의 분산 및 공분산으로 이루어진 행렬 |

<br>


**$x$-변수에 대한 측정 모형**

$$\begin{cases}
x_1\;=\;\lambda^x_{11}\xi_1 + \delta_1\\
x_2\;=\;\lambda^x_{21}\xi_1 + \delta_2\\
x_3\;=\;\lambda^x_{32}\xi_2 + \delta_3\\
x_4\;=\;\lambda^x_{42}\xi_2 + \delta_4\\
x_5\;=\;\lambda^x_{53}\xi_3 + \delta_5\\
\end{cases}$$

<br>

**$y$-변수에 대한 측정 모형**

$$\begin{cases}
y_1\;=\;\lambda^y_{11}\eta_1 + \epsilon_1\\
y_2\;=\;\lambda^y_{22}\eta_1 + \epsilon_2\\
y_3\;=\;\lambda^y_{32}\eta_2 + \epsilon_3\\
\end{cases}$$

<br>

**잠재변수에 대한 측정 모형**

$$\begin{cases}
\eta_1\;=\;\gamma_{12}\xi_2 + \zeta_1\\
\eta_2\;=\;\gamma_{21}\xi_1 + \gamma_{23}\xi_3 + \beta_{21}\eta_1 + \zeta_2\\
\end{cases}$$

<br>

<br>

#### 계수행렬의 표현

<br>

$$\Lambda_x = \begin{pmatrix}\lambda^x_{11}&0&0\\\lambda^x_{21}&0&0\\0&\lambda^x_{32}&0\\0&\lambda^x_{42}&0\\0&0&\lambda^x_{53} \end{pmatrix}$$

$$\Lambda_y = \begin{pmatrix}\lambda^y_{11}&0\\0&\lambda^y_{22}\\0&\lambda^y_{32} \end{pmatrix}$$

$$\Gamma = \begin{pmatrix} 0&\gamma_{12}&0\\\gamma_{21}&0&\gamma_{23} \end{pmatrix}$$

$$B = \begin{pmatrix} 0&0\\\beta_{21}&0 \end{pmatrix}$$

<br>

#### 공분산행렬의 표현
- 일반적으로 구조방정식모형에는 외생잠재변수들 간의 연관관계 및 오차변수들 간의 연관관계가 포함
- 각 변수들의 변이를 나타내는 분산(variance)과 변수들 간의 연관관계를 나타내는 공분산

$$\Phi = \begin{pmatrix} \phi_{11}&&\\\phi_{21}&\phi_{22}&\\\phi_{31}&\phi_{32}&\phi_{33} \end{pmatrix}$$

$$\Psi = \begin{pmatrix} \psi_{11}&\\0&\psi_{22} \end{pmatrix}$$

$$\Theta_\delta = \begin{pmatrix} \theta^\delta_{11}&&&&\\0&\theta^\delta_{22}&&&\\0&0&\theta^\delta_{33}\\0&0&0&\theta^\delta_{44}\\0&0&0&0&\theta^\delta_{55} \end{pmatrix}$$

$$\Theta_\epsilon = \begin{pmatrix} \theta^\epsilon_{11}&&\\0&\theta^\epsilon_{22}&\\0&0&\epsilon^\epsilon_{33} \end{pmatrix}$$

<br>

- 위 예에서는 모든 외생잠재변수 $\xi$들이 서로 연관되어 있는 것으로 가정 $\rightarrow$ $\Phi$-행렬의 원소가 추정대상
- 위 예에서, 모든 오차변수들은 서로 연관되어 있지 않은 것으로 가정 $\rightarrow$ 오차변수에 대한 세 개의 공분산행렬은 분산을 나타내는 대각원소만 추정대상, 공분산인 비대각원소는 추정대상이 아니므로 0으로 고정

## A.02. 모형에 대한 식별



- 연구자에 의해서 가정된 모형에 대해 언제나 모든 해를 구할 수 있는 것은 아님
- **모형에 포함된 모수들에 대한 정확한 유일해의 존재여부를 다루는 모형식별 (model identification)의 문제는 규정된 모형의 적정성과 연계**
    - 가급적 모형구축 직후 자료적합 이전 단계에서 검토
    
<br>

#### 단순회귀모형(simple regression model)
$$y=\gamma x + \zeta$$

- **관찰변수들 간의 구조적 관계는 분산 $Var(y), Var(x)$와 공분산 $Cov(y, x)$로 이루어진 모공분산행렬 (population covariance matrix) $\Sigma$를 통해 묘사**

$$\Sigma = \begin{pmatrix} Var(y)&\\Cov(y, x)&Var(x) \end{pmatrix}$$

- 연구자에 의해 규정된 구조방정식 (이 경우 단순회귀모형)모형에 대해서는 $x$와 $\zeta$가 독립이라는 조건하에서 다음 관계가 성림

$$\begin{cases}
Var(y)=\;\gamma^2 Var(x)+Var(\zeta)=\gamma^2 \phi + \psi\\
Cov(y, x)=Cov(\gamma x + \zeta, x)=\gamma Var(x) = \gamma \psi\\
\end{cases}$$

- 여기에서 $\theta = \{\gamma, \phi, \psi\}$와 같이 $\theta$를 모형모수들의 집합이라고 할 때, 위 식의 결과를 이용하여 단순회귀모형하에서, $x$, $y$의 공분산행렬을 $\theta$의 함수로 표현

$$\Sigma(\theta) = \begin{pmatrix} \gamma^2 \phi+\psi&\\\gamma \phi&\phi \end{pmatrix}$$

- 만약 연구자의 모형이 적절하고 $\Sigma = \Sigma(\theta)$가 성립한다면, 아래의 연립방정식에 의하여 모수 $\gamma, \phi, \psi$에 대한 해를 구할 수 있음

$$\begin{cases}
Var(y) = \gamma^2 \phi + \psi\\
Cov(y, x) = \gamma \phi\\
Var(x) = \phi\\
\end{cases}$$

- 이 경우 모형모수들은 아래와 같이 $\Sigma$의 원소들로부터 해를 구할 수 있음

$$\begin{cases}
\phi = Var(x)\\
\gamma = {{Cov(y, x)}\;/\;{Var(x)}}\\
\psi = Var(y) Var(x)\;/\;[Cov(y, x)]^2\\
\end{cases}$$


<br>

### 식별가능

- **관찰변수의 개수를 $k = p + q$(여기서 $p$는 $x$-변수의 개수이고, $q$는 $y$-변수의 개수)라고 하고,**
    
    **모공분산행렬 $\Sigma$와 연구자의 모형에 의해 모수들의 함수로 재표현된 공분산행렬 $\Sigma(\theta)$의 ($i,j$)원소를 각가 $\sigma_{ij}$와 $\sigma(\theta)_{ij}$이고,**

    **$\Sigma = \Sigma(\theta)$에 있는 $k(k+1)/2$개의 연립방정식 $\sigma_{ij}=\sigma(\theta)_{ij}$를 풀었을 때 ($i \geq j$),**
    
    **모든 모수들이 공분산행렬 $\Sigma$의 원소들로만 표현되면 모형이 식별가능하다고 판단**
    
<br>

### 식별의 필요조건과 포화모형
- 모형의 식별을 위한 필요조건은 추정해야 할 독립적인 미지수의 개수 $t$가 연립방정식 $\sigma_{ij} =\sigma(\theta)_{ij}$의 개수 $k(k+1)/2$보다 작거나 같아야 함 ($i\geq j$)

    **즉, 자유도 (degree of freedom) $d$를**
    
    $$d = {{k(k+1)}\over{2}}-t$$
    
    **와 같이 정의할 때 $d$가 0보다 크거나 같아야 함**
    
<br>

<hr> 



<hr>

**모형에서 추정해야 할 모수** 

| 행렬 | (a) | (b) | (c) | (d) | 
| ---- | --- | --- | --- | --- |
| $\;\Gamma\;$ | $\gamma_{11},\gamma_{12},\gamma_{21},\gamma_{22}$ | $\gamma_{11},\gamma_{12},\gamma_{21},\gamma_{22}$ | $\gamma_{11},\gamma_{12},\gamma_{21},\gamma_{22}$ | $\gamma_{11},\gamma_{12}$ |
| $\;B\;$ |  | $\beta_{21}$ | $\beta_{12}, \beta_{21}$ | $\beta_{12}, \beta_{21}$ |
| $\;\Phi\;$ | $\phi_{11},\phi_{12},\phi_{22}$ | $\phi_{11},\phi_{12},\phi_{21}$ | $\phi_{11},\phi_{12},\phi_{21}$ | $\phi_{11},\phi_{12},\phi_{22}$ |
| $\;\Psi\;$ | $\psi_{11}, \psi_{22}$ | $\psi_{11}, \psi_{22}$ | $\psi_{11}, \psi_{22}$ | $\psi_{11}, \psi_{21}, \psi_{22}$ |
| 자유도 | 1 | 0 | -1 | 0 |

<br>

- 위 모형들은 각각 2개씩의 외생변수/내생변수를 가지고 있으므로 $k(k+1)/2 = 4(5)/2 = 10$이고,

    (a), (b), (c)에 대해서는 자유도가 0보다 크거나 같음
    
    **따라서 이 세 모형은 식별의 필요조건을 만족하며, 특히 모형 (b), (d)와 같이 자유도가 0인 모형을 포화모형 (saturated model)**
    
    - 포화모형은 식별의 필요조건을 만족하기는 하지만, 식별가능한 가장 복잡한 모형 중의 하나
    - 통계적 모형의 모형구축이 자료에 내재 되어있는 복잡한 구조를 단순화하여 설명하려는 것에 있다고 할 때, 포화모형은 모수의 절약이 시도하지 않으므로 간명성의 측면에서 모형으로서 가치가 떨어짐
    

<br>

### 과다모수와 식별불능
- **자유도 $d$가 0보다 작으면, 방정식의 개수보다 미지수의 개수가 많다는 것을 의미하고, 해가 무핸히 존재하여 유일한 해를 구할 수 없게됨**
- **일반적으로 식별불가능 (under-identified)하다는 것은 불충분한 정보를 의미하며, 곧 모수가 유일하게 정의될 수 있도록 모형을 더욱 구체화하여 제약하거나 더 많은 변수들이 모형에 도입되어야 함을 나타냄**

- 모형식별을 위한 충분조건이나 필요충분조건이 존재하지만, 식별가능성은 모형모수의 유형과 구성에 의존하며 표본크기등과는 관련이 없다는 점에서 모형구축에서의 간명성과 신축성이란 양 측면과 밀접히 연계

<br>

### 일부 모수의 식별불능
- 자유도가 0보다 크거나 같아서 모형이 식별의 필요조건을 만족한다고 해도, 모든 모수에 대해 식별이 가능한 것은 아님
- 즉, 연립방정식 내에서 부분적인 식별불가능 문제가 발생하여 일부 모수를 추정할 수 없는 경우가 발생

<br>

## A.03. 모수의 추정
- 연구자가 모공분산행렬 $\Sigma$를 알고 있다면 식별가능한 모형에 대해 $\Sigma = \Sigma(\theta)$를 풀어봄으로써 해를 구할 수 있음
    - 하지만 실제로는 $\Sigma$를 알 수 없기 때문에 N개의 표본으로부터 얻은 표본공분산행렬(sample covariance matrix) $S$를 이용하여 $\Sigma(\theta)$의 추정치 $\hat{\Sigma(\theta})$이 가급적 $S$와 가깝도록 해야 하는데, 
    
    $\hat{\Sigma(\theta})$와 $S$가 가까운 정도를 나타내는 함수 $F[S, \Sigma(\theta)]$를 적합함수 (fit function)이라고 하며 이 적합함수를 어떻게 정의하느냐에 따라 추정방법이 달라짐
    
<br>

### 적합함수
- 구조방정식모형에 대한 분석에서 고려되는 적합함수들은 일반적 적합함수 (general fit function)의 특별한 형태

$$F[S, \Sigma(\theta)] = (s - \sigma)' W^{-1} (s-\sigma) = \sum^k_{g=1} \sum^g_{h=1} \sum^k_{i=1} \sum^i_{j=1} w^{g,h,ij} (s_{gh}-\sigma_{gh}) (s_{ij}-\sigma_{ij})$$

- 여기에서 $s_{ij}$는 전체 관찰변수에 대한 표본공분산행렬
$$s_{ij} = \begin{pmatrix} S_{yy}&S_{yx}\\S_{xy}&S_{xx}\end{pmatrix}$$의 $(i, j)$번째 원소

- $\sigma_{ij}$는 모수들에 의해서 재표현된 공분산행렬의 $(i, j)$번째 원소


$$\Sigma(\theta) = \begin{pmatrix} {\Lambda_y (I-B)^{-1} (\Gamma\Phi\Gamma' + \Psi)(I - B')^{-1} \Lambda'_y+\Theta_\epsilon }&{\Lambda_y (I-B)^{-1} \Gamma \Phi \Lambda' _x}\\{\Lambda_x \Phi \Gamma ' (I-B')^{-1} \Lambda'_y}&\Lambda_x\Phi\Lambda'_x + \Theta_\delta \end{pmatrix}$$

- $s'$와 $\sigma '$는

$$\begin{cases}
s' = (s_{11}, s_{21}, \dots, s_{kk})\\
\sigma' = (\sigma_{11}, \sigma_{21}, \dots, \sigma_{kk})\\
\end{cases}$$

와 같이 $s$와 $\Sigma(\theta)$의 대각원소를 포함한 하삼각행렬의 원소들을 행으로 늘어놓은 행벡터를 표시한 것
    
$w^{gh,ij}$는 크기 $u \times u$ (단, $u = k(k+1)/2$)인 양정치행렬 $W^{-1}$의 원소를 나타냄

- 만약 모형이 모집단에 대해서 적절하고 표본분산 및 공분산들이 표본크기가 증가함에 따라 모공분산행렬 $\Sigma$에 대응되는 원소들에 확률적으로 수렴한다면, 임의의 양정치행렬 $W$를 사용한 어떠한 적합함수도 모수에 대해 일치추정량을 제공한다
- 그러나, 만약 근사적으로 정확한 추정치의 표준오차나 적합도 검정을 위한 카이제곱 값을 필요로 한다면, 추가적인 가정들이 필요

<br>

- 구조방정식모형에 대한 '고전적 이론'은 $S$의 원소들의 점근적 (asymtotic)분산과 공분산이 다음과 같은 형태임을 가정

$$ACov(s_{gh}, s_{ij}) = {{1} \over {N}} (\sigma_{gi} \sigma_{hj} + \sigma_{gh} \sigma_{hi})$$

- 특히 관찰변수들이 다변량 정규분포에 따르거나 $S$가 Wishart-분포를 따르면 위와 같은 가정이 만족되며 비정규분포의 경우라도 첨도가 지나치게 크지 않는 한 이 가정에는 특별히 문제가 되지 않음이 알려짐
- GLS와 ML, 카이제곱값과 표준오차는 이러한 가정에 기초

<br>

### ML (Maximum Likelihood, 최대우도법)
- 최대우도법에 대한 적합함수는 관찰변수들이 다변량 정규분포에 따른다는 가정에 기초를 두고 있으며, 적합함수는

$$F = log|\Sigma| + tr(S\Sigma^{-1})-log|S| - (p+q)$$

- 최대우도법은 정규성의 가정하에서는 불편성, 일치성, 최소분산성, 정규분포성 등을 가지는 가장 효율적인 추정량을 제공

<br>

### ULS (Unweighted Least Square, 비가중 최소제곱법)
- 적합함수 : 

$$F = {1 \over 2} tr[(S-\Sigma)^2]$$

- 행렬 $S-\Sigma$의 모든 원소들의 제곱합을 2로 나눈 것으로, 모든 잔차에 동일한 가중치를 부여한 경우
- ULS 방법은 확률분포에 관한 특정 가정이 요구되지는 않으나, 모든 관찰변수들이 동일한 단위로 측정되었을 때만 정당화

<br>

### GLS (Generalized Least Square, 일반화 최소제곱법)
- 적합함수 :

$$F = {1 \over 2} tr[(I-S^{-1}\Sigma^2)]$$

- 식 $F[S, \Sigma(\theta)] = (s - \sigma)' W^{-1} (s-\sigma) = \sum^k_{g=1} \sum^g_{h=1} \sum^k_{i=1} \sum^i_{j=1} w^{g,h,ij} (s_{gh}-\sigma_{gh}) (s_{ij}-\sigma_{ij})$에서 $W=S$로 취한 것과 같음

- GLS는 적합함수에서 ULS를 단순히 일반화시킨 형태이나, 몇 가지 조건하에서 ML 방법의 결과와 같은 수준의 최적성을 가짐
- GLS의 추정량의 정규분포성과 효율성은 ML방법의 경우보다 덜 제한적인 경우에서 성립한다는 등의 로버스트 (robust) 성질을 가짐
- 정규성하에서는 ML과 동등한 추정량을 제공하며, 추정량의 표준오차나 카이제곱 값에 대한 추론결과는 정규분포 가정하에서 유도된 것



<br>

### ADF (Asymtotically Distribution-free, WLS, 일반화 가중최소제곱법)
- $W$행렬의 아래와 같이 정하며, 이는 $ACov(s_{gh}, s_{ij})$의 일치 추정량

$$w_{gh, ij} =m_{ghij} -s_{gh}s_{ij}$$

여기에서 $m_{ghij}$는 4차 중심적률
    
$$m_{ghij} = {1\over N} \sum^N_{a=1}(z_{ag}-\bar{z_g})(z_{ah}-\bar{z_h})(z_{ai}-\bar{z_i})(z_{aj}-\bar{z}_j)$$

<br>

- ADF 방법은 $W$를 식 $F[S, \Sigma(\theta)] = (s - \sigma)' W^{-1} (s-\sigma) = \sum^k_{g=1} \sum^g_{h=1} \sum^k_{i=1} \sum^i_{j=1} w^{g,h,ij} (s_{gh}-\sigma_{gh}) (s_{ij}-\sigma_{ij})$의 적합함수에서 가중행렬로 사용한 것
    - 점근적으로 분포무관한(distribution free) 추정방법이며 정규분포성으로부터 큰 이탈을 보이는 자료에 적용할 수 있는 최적의 방법
    
<br>

**실제 응용에 있어서의 한계점**
1. 많은 양의 계산시간
2. 정확한 4차 적률을 계산하기 위해서는 많은 표본 필요
3. 결측값이 있을 때는 4차 적률들이 서로 다른 개수의 표본에 기초

<br>

### DWLS(Diagonally Weighted Least Square)
- ADF에 대해 $W$의 모든 원소 대신 $W^{-1}$의 대각원소만을 가중치로 이용

- 적합함수 : 

$$F = \sum^k_{g=1}\sum^k_{h=1} {1\over {w_{gh}}} (s_{gh} - \sigma_{gh})^2$$

- 모수에 대하여 효율적인 추정치를 제공하지는 않지만, ULS와 WLS의 절충안

<br>

## A.04. 적합도 통계량
- 구조방정식모형의 일차적인 목적은 연구자가 가정한 모형이 자료에 의해서 얼마나 지지되는지를 평가
- 연구자의 모형에 의해서 추정된 공분산행렬 $\hat{\Sigma(\theta)}$가 $S$와 얼마나 부합하는지를 평가하는 것이라고 할 수 있으며, 만약 부합정도 (적합도)가 좋지 않다면 모수에 대한 추가적인 해석은 의미가 없음

- 표본공분산행려 $S$의 $(i, j)$번째 원소를 $s_{ij}$라고 하고 $\hat{\Sigma}(\theta)$의 $(i, j)$번째 원소를 $\hat{\sigma_{ij}}$라고 할 때 $(i, j=1, \dots, k)$, 다음과 같이 정의도는 적합값 (fit criterion)은 적합함수 $F[S, \Sigma(\theta)]$를 최소화한 값

$$F[S, \hat{\Sigma}(\theta)] = \sum^k_{g=1} \sum^g_{h=1} \sum^k_{i=1} \sum^i_{j=1} w^{g,h,ij} (s_{gh}-\hat{\sigma}_{gh}) (s_{ij}-\hat{\sigma}_{ij})$$

- $s_{ij}-\hat{\sigma_{ij}}$는 적합잔차 (fitted residual) 이며, $F[S, \hat{\Sigma}(\theta)]$는 모형이 실제 데이터와 불일치하는 정도를 나타냄
    - 가중치 $w^{gh,ij}$는 추정방법에 따라서 다름

### 카이제곱 통계량

$$\chi^2 = nF[S, \hat{\Sigma}(\theta)]$$

$$자유도\;d = k(k+1)/2 -t\;,\; n=N-1(N은\;표본크기)$$

$$k=\;관찰변수\;개수,\;t=서로\;독립적인\;자유모수\;개수$$

<br>

- **모형이 적절하고 표본의 개수가 충분히 크다면, 카이제곱통계량은 아래의 가설을 검정** 

$$H_0:\Sigma=\Sigma(\theta)\;vs\;H_1:\Sigma>0$$
    
**귀무가설 = 모공분산행렬이 모형에 의해서 정의되는 공분산행렬과 같다**

**대립가설 : 모공분산행렬이 어떠한 제약도 가지고 있지 않음**

<br>

- **표본크기가 매우크면 모공분산행렬에 잘 근사함에도 불구하고, 사소한 차의를 유의적으로 판단함으로서 귀무가설 채택이 어려워짐**

- **모수를 추가하여 복잡한 모형을 만들수록 카이제곱 통계량의 값은 작아져, 적합도는 증가하지만 지나치게 복잡하다는 문제 발생**

- 일반적으로 카이제곱 통계량에 의한 검정은 관찰변수들이 다변량정규분포에 따르고, 표본공분산행렬에 근거하여 분석이 수행되어야 하며, 표본크기가 적절히 커야한다 라는 가정이 만족될 경우에만 유효
    - 실제 분석에서 이러한 가정이 모두 만족되기는 쉽지 않음
- 카이제곱 통계량은 검정통계량보다는 모형이 데이터에 의해서 지지되는 정도를 나타내는 하나의 측도로써 유용하게 사용

<br>

#### **표준카이제곱 : 카이제곱 값을 자유도로 나눈 값**
- **얼마나 작아야 모형을 수용할 수 있는지에 대한 명확한 지침은 없음 (2 혹은 3보다 작으면 좋다는 기준 통용)**

<br>

### RMSEA (근사치 오차평균 제곱근, Root Mean Square Error of Approximation)

$$\epsilon_{\alpha} = \sqrt{max({1\over d} F[S, {\Sigma}(\hat{\theta})]- {1\over n}, 0)}$$

- **모형이 연구 모집단을 얼마나 잘 근사하느냐의 정도를 나타냄 (= 모형의 불일치도)**
- **모형의 간명성조정지수**

<br>

- **RMSEA가 0일 때 완벽한 적합도**
- **RMSEA가 큰 값을 가질수록 좋지 않음**
    - **좋음 : < 0.05**
    - **괜찮음 : 0.05 ~ 0.08**
    - **보통 : 0.08 ~ 0.10**
    - **나쁨 : > 0.1**

<br>

* **신뢰구간 산출 가능 $\rightarrow$ 좋은 적합도에 대한 검정 수행 가능**
    * **신뢰하한이 좋은 기준값 0.05보다 작을 시 $\rightarrow$ '좋은 적합도를 가진다는 가설 ($H_0:\epsilon \leq 0.05$)'이 기각되지 않음**
    * **신뢰상한이 나쁜 기준값 0.1보다 작을 시 $\rightarrow$ '적합도가 좋지 않다는 가설 기각 ($H_1:\epsilon \geq 0.1$)'**
    
- **표본크기가 작을 수록 넓은 신뢰구간이 일어나기 쉬움**

<br>

### GFI, AGFI, PGFI

<br>

#### GFI (절대 적합도 지수)

- **카이제곱 통계량이 표본크기에 의존하지만, GFI, AGFI, PGFI는 표본크기에 무관**
- $F[S, \hat{\Sigma}(\theta)]$를 모형이 적합되기 이전의 적합값 (즉, 모든 모수가 0일 때의 적합값일 때, GFI(Goodness of Fit Index)

$$GFI = 1 - {{F[S, {\Sigma}(\hat{\theta})]}\over{F[S, \hat{\Sigma}(0)]}}$$

- **GFI = 1.0 : 완벽한 모형 적합도**
    - **0에 가까우면 나쁜 적합도**
    - **0.9이상이면 적합도가 좋은 것으로 간주**

- **GFI는 모형의 간명성을 고려하지 않음 $\rightarrow$ 모형을 복잡하게 만들수록 GFI는 증가**

<br>

#### AGFI (수정 적합도 지수)

$$AGFI = 1 - {{k(k+1)}\over{2d}}(1-GFI)$$

- GFI는 모수를 추가하면 계속 증가하지만, AGFI는 증가하는 GFI가 줄어드는 자유도에 비해서 상대적으로 큰 경우에는 증가하고, 그 반대에는 감소

    $\rightarrow$ 모수를 추가하거나 제거하여 모형을 수정하는 경우 여러 수정보형들을 서로 비교하고자 할 떄 유용하게 사용
    
- **AGFI값이 0.85이상이면 좋은 것으로 간주**

(AGFI는 성능상의 문제로, 최근에는 사용 X)

<br>

#### PGFI (간명성 적합도 지수)
$$PGFI = {{2d}\over{k(k+1)}}GFI$$

- 0.5정도 이면 좋은 것으로 간주
- 표본크기에 민감



<br>

### CFI, NNFI, NFI, PNFI

- **독립모형에서의 적합값을 $F_1$, 추정하는 모형에서의 적합값이 $F[S, {\Sigma}(\hat{\theta})]$일 떄,**

   **독립모형을 기준으로 추정모형의 적합도를 상대비교하는 다양한 지수**
   
- 관찰변수들의 공분산이 모두 0 (관찰변수들이 서로 아무런 연관관계가 없음)이라는 가정은 지나치게 비현실적이기에,

    증분 적합도지수가 이론모형을 독립모형과 비교한다는 점이 문제로 지적
    
    (독립모형은 너무나 나쁜 성능의 모형이기에, 이론모형이 독립모형보다 꽤 나은 적합도를 가진다고 해도 이것을 우수하다고 판정하기 어려울 수 있기 때문)

<br>
<hr>

- $d_i$ = 독립모형의 자유도
- $f=nF/d$
- $f_1 = nF_I/d_I$
- $\tau=max(nF-d, 0)$
- $\tau=max(nF_I-d_I, 0)$

일 때, **증분 적합도지수 (incremental fit index, 혹은 상대적합도지수)는**

$NFI = 1 - F/F_I$

$TLI = (f_1-f)/(f_I-1)$

$PNFI = (d/d_I)(1-F/F_I)$

$CFI = (1-\tau/\tau_I)$

$IFI = (nF_I-nF)/(nF_I-d)$

$RFI = (f_I-f)/f_I$

<hr>

<br>

#### CFI (comparative fit index)
- 표본크기에 영향을 받지 않지만, 간명성을 고려하지 않음
- 보통 0과 1사이에 있음 (범위를 벗어나면 0또는 1로 간주)
- 0.9이상이면 좋은 것으로 간주
- CFI는 관찰변수들의 상관계수의 평균 크기에 의존 $\rightarrow$ 관찰변수들의 평균 상관계수가 크지 않다면 CFI도 크지 않음

<br>

#### TLI (Tucker-Lewis index, NNFI)
- 표본크기에 영향을 받지는 않지만, 간명성을 고려하여 적합도를 평가
- 보통 0과 1사이에 있음 (범위를 벗어나면 0또는 1로 간주)
- 0.9이상이면 좋은 것으로 간주
- CFI는 관찰변수들의 상관계수의 평균 크기에 의존 $\rightarrow$ 관찰변수들의 평균 상관계수가 크지 않다면 CFI도 크지 않음

<br>

#### NFI (normed fit index)
- TLI가 0과 1사이를 벗어날 수 있따는 점을 수정
- 0.9이상이면 좋은 것으로 간주
- 표본크기가 충분히 크지 않을 경우, 편의오차 (biased error)를 가질 수 있음

(가급적으로 사용X)

<br>

#### PNFI (parsimony-adjusted NFI)
- 모형의 간명성을 고려하여 NFI 값을 조정
- 모형의 크기에 상당히 민감 $\rightarrow$ 관찰변수의 개수가 10개 이하인 경우처럼 모형의 크기가 상대적으로 작을수록 모형의 복잡성에 대해 가해지는 조정이 더 커지는 경향

<br>



### AIC, ECVI, CAIC, BIC
- **정보기준 통계량들은 모수의 추정과 모형선택을 하나의 개념적 틀 아래 결합시킨 것**
    - 대부분 간명성을 고려하여, 단순한 모형에 유리하도록 고안
    
- 이 적합도 지수들은 여러 개의 경쟁 모형들을 비교하여 평가할 때 유용
- 위계관계(nested, 포함관계)에 있는 모형들 이외의 다른 모형도 비교할 수 있다는 장점

<br>

#### AIC (Akaike information criterion)
- 작은 모형일 수록 좋은 모형 (= 상대적으로 적합도가 더 우수하면서 간명한 모형)

<br>

#### ECVI (expected cross-validation index)
- AIC와 유사하게 경쟁모형들의 우수성을 비교

<br>

#### CAIC (consistent Akaike information criterion)
- AIC와는 달리 표본의 크기를 직접 고려

<br>

#### BIC (Bayes information criterion)
- 표본크기가 클 경우 AIC나 CAIC보다 모형의 간명성을 더 크게 고려

<br>

### SRMR

<br>

#### RMR (root mean square residual)
- 적합잔차의 제곱(= ($s_{ij}-\hat{\sigma_{ij}})^2$)들의 평균에 제곱근을 취한 것

$$RMR=\sqrt{{2\over{k(k+1)}} \sum^k_{i=1} \sum^i_{j=1} (s_{ij} - \hat{\sigma_{ij}})^2}$$

- RMR이 작을수록 적합정도가 좋음을 의미
- RMR은 관찰변수들의 측정단위에 의존 $\rightarrow$ 해석하거나, RMR에 대해 좋은 기준을 제시하기가 어려움

<br>

#### SRMR (standardized RMR)
- 표준화 잔차에 기초한 RMR의 표준화 형태 $\rightarrow$ 관찰변수의 측정단위에 의존하지 않음
    - 해석적 측면에서 RMR보다는 SRMR이 선호
- SRMR이 0.08보다 작을 경우 좋은 적합도를 가진 것으로 간주
- 0.1보다 작을 경우 양호

<br>

### CN (Critical N)

$$CN = {{\chi^2_{0.05}(d)}\over{F}} +1$$

- 5% 유의수준하에 $S$와 $\Sigma(\theta)$의 차이를 유의적으로 하는 표본크기를 의미
- 200이상이면 적합정도가 좋은 것으로 간주
