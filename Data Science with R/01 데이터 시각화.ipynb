## 1.5 facet 면분할(데이터 각 서브셋을 표시하는 하위 플롯)

### - class별 displ과 hwy의 산점도 (3행으로)

````r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy)) + facet_wrap(~class, nrow = 3)
````

### - drv과 cyl별, displ과 hwy의 산점도

````r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy)) + facet_grid(drv ~ cyl)
````

````r
# 1.5.1
ggplot(mpg) + geom_point(aes(x = displ, y = hwy)) + facet_grid(drv ~ .)
````

````r
ggplot(mpg) + geom_point(aes(x = displ, y = hwy)) + facet_grid(. ~ drv)
````

````r
ggplot(data = diamonds) + stat_count(aes(x = cut))
# ggplot(diamonds) + geom_bar(aes(x = cut, y = ..count.., group = 1))
````

````r
demo <- tribble(~cut, ~freq,
               "Fair", 1610,
               "Good", 4906,
               "Very Good", 12082,
               "Premium", 13791,
               "Ideal", 21551)
````

### 요약통계량 시각화

````r
# cut의 범주별 요약통계량
ggplot(data = diamonds) + stat_summary(aes(x = cut, y = depth),
                                      fun.min = min,
                                      fun.max = max,
                                      fun = median) # mean
````

````r
# 1.7.1
ggplot(data = diamonds) +
  geom_pointrange(
    mapping = aes(x = cut, y = depth),
    stat = "summary",
    fun.min = min,
    fun.max = max,
    fun = median)
````
