### 데이터 생성 프로세스
- ㄱ와 ㄴ은 각각 담당하는 웹페이지에 관해 2가지 디자인안을 가지고 있으며, 디자인안에 따라 클릭률이 달라진다는 주장을 하고 있음
  
  $\rightarrow$ **디자인안에 따라 클릭률이 달라진다는 것은 각 디자인안이 고유한 클릭률을 가지고 있는 것으로 생각할 수 있음**

  $\rightarrow$ **데이터로부터 그 진위를 추론**

<br>

- **어떤 디자인안이 가지고 있는 고유의 클릭률 : $\theta$**
  - $\theta$ : 클릭의 비율 (확률). 0에서 1 사이의 값 (0 $\leq$ $\theta$ \leq 1)

<br>

- 이 디자인안이 사용자에게 노출되면 사용자는 반응하며, 의도한 버튼을 클릭 혹은 클릭하지 않을 수 있음
  - **클릭 유무 $r$ : 클릭했을 때 1, 클릭하지 않았을 때 0으로 표시**

  - $\theta$가 커지면 높은 확률로 $r=1$이 되고, 작아지면 높은 확률로 $r=0$이 됨
  
  - 극단적으로 $\theta=1$이면 아무리 많은 사용자에게 노출되어도 클릭이 되는 $r=1$이 되며,

    $\theta=0$이면 아무리 많은 사용자에게 노출되어도 클릭되지 않음 $r=0$

- **$r$과 같이 확률적으로 다양한 값을 가지는 변수를 확률변수 $random variable$**

<br>

- 사용자에게 어떤 디자인안을 노출시키는 것, 어떤 실험을 함으로써 확률변수의 값을 측정하는 것을 **시행 (trial)**
  - 모든 시행에는 반드시 결과가 따름 (클릭이 발생/미발생)
  
    $\rightarrow$ 값이 0이든 1이든 $r$의 구체적인 값이 관측




  
<br>

#### 베르누이 시행


<br>그림1.jpg



- 각 디자인안은 고유의 클릭률 $\theta$를 가지고 있으며, 사용자에게 표시하는 시행마다 클릭 여부 $r$을 생성

$$\theta \sim P(r|\theta)$$
$$r=0, 1$$
  
- **결과가 0 또는 1 두 가지로 표시되고, 성공 확률 $\theta$가 몇 번을 반복해도 변하지 않는 시행은 베르누이 시행**
- 분석자 : 이러한 시행을 관찰
  - 디자인안 내부에 고유의 클릭률 $\theta$가 감춰져 있기 때문에 이를 직접 알지는 못하지만,

    디자인안을 시행한 횟수와 그 결과로 얻은 클릭 횟수만 알고 있으며, 이 데이터로부터 $\theta$를 추정

  - 분석자가 감춰져 있는 클릭률 $\theta$에 관해 가지고 있는 생각
    
    예) '$\theta=0$명 정도일 확률은 0.1, $\theta=0.5$정도일 확률은 0.3

    **등의 생각을 정량적으로 표시한 것이 신념 (Belief)이며, 이는 확률 분포 (Probabiltiy Distribution)의 형태로 표현**

- **통계적 관점 : 실제로 클릭하는 주체는 사용자이지만, 디자인안이 클릭을 만들어내는 것 처럼 관측**

  - **사용자마다 행동 경향이 다르고, 시간대나 장소 등 다양한 요소가 복잡하게 얽혀 있겠지만,**
  
    **각 디자인안을 무작위로 사용자에게 표시함으로써 그와 같은 요소들을 원래 알고자 하던 효과와는 관계없는 불규칙한 요소로 처리**

## 01.03. 확률 분호
- 확률 변수가 특정한 값을 가질 확률을 정해놓은 것

<br>

### 이산값의 확률분포
- 주사위를 던졌을 때 나오는 면 $X$를 확률 변수라고 하면, 확률 변수 $X$의 확률 분포는 동일하게 1/6
  
  **$\rightarrow$ 모든 값이 같은 확률을 가지는 분포는 균일 분포 (Uniform Distribution)**


- **주사위나 제비뽑기처럼 여러 선택지에 대해 정의된 확률 분포를 카테고리컬 분포 (Categorical Distribution, 범주형 분포)**

<br>

- 확률분포가 가지는 형태는 결정되어 있지 않지만 만족해야 할 조건이 존재
1. 확률분포가 1,2, 3...과 같이 유한한 개수의 비연속적인 값 (이산값)으로 표시되는 경우,

  이산값에 관한 확률분포는 **이산확률분포 (Discrete Prob Dist)**

  - **이산확률분포가 만족해야 하는 조건 중 하나는 확률 변수가 여러 가지 값을 가질 확률의 총합이 1이 되어야 하는 것**

$$p(X = X_1) + \dots + p(X= X_n) = \sum^n_{k=1} p(X= X_k)=1$$
$$p(X=X_i)\text{는 확률변수 X가 특정값 }X_i\text{를 가지는 확률}$$

2. 어떤 현상이 발생할 확률은 음수가 되지 않음

$$p(X=X_i) \geq 0 \;\;(i=1, \dots, n)$$

<br>

#### 파라미터
- 어떤 확률분포를 대표하는 변수

  - 예) 각 상품이 나오는 범주형 분포의 파라미터 $P_a = 0.6, p_2 = 0.3, p_3=0.09,p_4=0.01$

### 연속값의 확률분포
- 확률변수가 연속값일 때의 확률 분포 (Continuous Prob Dist)
1. 확률변수가 가지는 모든 값의 확률을 합하면 1

$$\int^{+\infty}_{-\infty} p(x) dx = 1$$

2. 어떤 현상이 발생할 확률은 음수가 되지 않음
- 연속확률분포의 경우, 확률분포로부터 확률변수가 특정값을 가질 확률을 직접 읽을 수 없음
  - 이산확률분포에서는 확률분포가 각 현상이 발생하는 확률에 대응하지만, 연속확률분포에는 그렇지 않음

$$p(x) \geq 0$$

3. **확률분포의 세로축에 해당하는 값은 확률의 밀도**

- **확률분포함수의 값이 1을 넘는 것도, 그 값이 확률을 의미하는 것이 아니라 밀도를 나타냄**

  - 연속확률분포를 나타내는 함수는 확률밀도함수 (Probability Density Function, PDF)
  - 이산확률분포를 나타내는 함수는 확률질량함수 (Probability Mass Function, PMF)

<br>

### 덧셈정리 & 곱셈정리
- 이산값의 확률 변수 $X$와 $Y$가 각각 {$X_1, \dots, X_n$ }, {$Y_1, \dots Y_m$}의 값을 가질 때,

  두 확률변수가 동시에 어떤 특정한 값 $X=X_i, Y=Y_j$를 가질 확률은 $p(X=X_i, Y = Y_j)$

  $→$ **$p(X=X_i, Y = Y_j)$은 결합확률 (Joint Prob), 그 분포 $p(X, Y)$는 결합분포 (Joint Dist)**


<br>

##### 확률의 덧셈정리 (Addition Theorem)
$$P(Y=Y_j) = p(X=X_1, Y=Y_j) + \dots + p(X=X_n, Y=Y_j) = \sum^n_{i=1}p(X=X_i, Y=Y_j) = \sum_{X}p(X,Y)$$

- 결합분포 $p(X, Y)$를 하나의 확률변수 $Y$의 확률분포 $p(Y)$로 나누어 결과를 얻을 수 있는 분포 $p(X|Y)$

$$p(X|Y)=\frac{p(X,Y)}{p(Y)}$$

- $p(X|Y)$는 조건부 분포 (Conditional Dist)이며, 확률변수$Y$의 값이 주어진 상태에서의 확률변수 $X$의 확률분포를 의미

<br>



#### 주변분포 (Margin Dist)
- **결합분포에서 다른 하나의 확률변수를 제거 $\rightarrow$ 주변화 (Marginization)**

<br>

### 베이즈 정리
- 덧셈정리는 확률변수를 교환해도 성립하므로

$$p(x, y) = p(x|y)p(y) = p(y|x)p(x)$$

- 이 식 전체를 $p(y)$로 나누면

$$p(x|y) = \frac{p(y|x)p(x)}{p(y)}$$$\rightarrow$ 베이즈 정리 (Bayes Theorem)

<br>

- 베이즈 정리를 이용하여 어떤 확률분포의 알려지지 않은 파라미터를 관측한 데이터로부터 추론 $\rightarrow$ 베이츠 추론 (Bayesian Inference)

<br>

#### 베이즈 추론
- 관측된 데이터를 의미하는 확류변수 $D$와 알려지지 앟은 파라미터를 의미하는 확률변수 $\theta$

$$p(\theta|D) = \frac{p(D|\theta)p(\theta)}{p(D)}$$

- 베이즈 추론을 이용하여 구하려는 것은 $p(\theta|D)$인 조건부 분포
  - 이는 데이터 $D$가 주어진 상태에서 파라미터 $\theta$의 확률분포 $\rightarrow$ 사후분포 (Posterior Dist)

<br>

- 베이즈 정리에 따르면, 사후분포는 $p(D|\theta)$인 조건부 확률과 $p(\theta)$인 확률분포를 곱한 뒤, 이를 $p(D)$인 확률분포로 나누어 구함

  $\rightarrow$ **$p(\theta)$는 사전분포 (Prior Dist) 이며, 데이터를 관측하기 이전의 파라미터 $\theta$의 확률분포를 의미**

<br>

- **$p(D|\theta)$는 가능도 함수 (likelihood)이며, 파라미터 $\theta$에 대해 데이터 $D$가 관측될 당연함의 정도**

- **$p(D$)는 증거 혹은 정규화 상수 (Normalizing Constant)이며, 사후분포가 확률분포로서의 조건을 만족하도록 정규화하는 역할**

  - **이 값은 $p(D|\theta)p(\theta)$를 $\theta$에 관해 주변화하여 얻을 수 있음 (주변 가능도, Marginal Likelihood)**

  - 파라미터 $\theta$를 추론하기 위해 베이즈 정리를 이용할 때는 데이터 $D$가 주어진 것으로 취급

    $\rightarrow$ 확률변수 $\theta$의 확률분포에만 관심이 있을 때는 베이즈 정리를 구성하는 요소 중에서 유일하게 $\theta$를 포함하지 않는 $p(D)$는 상수로 간주하여

$$p(\theta|D) ∝ p(D|\theta)p(\theta)$$


<br>

- **베이즈 정리와 관측된 데이터 $D$를 이요하여 관심 파라미터 $\theta$에 관해 사후분포를 얻는 프레임이 베이즈 추론**

- 여러 데이터가 순차적으로 주어질 때는 이전 데이터에 따라 얻을 수 있는 사후분포를 사전분포로하여 베이즈 정리를 다시 적용함으로써,

  여러 데이터를 고려한 $\theta$의 사후분포를 얻을 수 있음

  $→$ **베이즈 업데이트 : 반복하여 베이즈 정리를 적용하는 프레임**

## 01.04. 베이즈 정리를 이용한 클릭률 추론
- 디자인안별 고유의 클릭률  $\theta$는 분석자에게는 알려지지 않았지만, 그에 관한 신념인 확률분포 $p(\theta)$는 존재
  - 분석가가 생각하는 파라미터 $\theta$의 확률분포이며, 사후분포에 해당
  - 분석자가 가진 신념을 관측된 데이터와 베이즈 정리를 이용해 업데이트





<br>그림1.jpg



<br>

1. 최초에는 분석자가 임의로 $\theta$에 대한 사전분포를 결정
- 사전에 단서가 될 만한 정보를 가지고 있다면, 이를 신념에 반영

<br>

2. $\theta$에 따른 가능도 함수 정의

$$p(r=1 |\theta) = \theta$$

$$p(r=0| \theta) = 1-\theta$$

$$\rightarrow p(r|\theta) = \theta^r (1-\theta)^{1-r} = Bernoulli(\theta)$$

- 일반적으로 가능도함수는 확률분포로서의 조건을 만족하지 않음

<br>

3. 베이즈 정리를 통한 사후분포 계산

$$p(\theta | r) = \frac{p(r|\theta) p(\theta)}{p(r)}$$

<br>

4. 관측값의 크기만큼의 베이즈 업데이트
- **여러번의 베이즈 업데이트를 통해 얻는 사후분포가 $\theta$에 대한 신념**

<br>

### Python 구현


```python
import numpy as np
from matplotlib import pyplot as plt
```

- $\theta$에 대한 사전분포의 범위 : (0 ~ 1)

```python
thetas = np.linspace(0, 1, 1001)
print(thetas)
```

- 베르누이 가능도 $p(r|\theta)$

```python
likelihood = lambda r: thetas if r else (1 - thetas)
```

- 베이즈 갱신 1회 시행

```python
def posterior(r, prior):
  lp = likelihood(r) * prior
  return lp / lp.sum() # 정규화
```

```python
p = np.array([1 / len(thetas) for _ in thetas])
p = posterior(1, p)  # 클릭
print(p)
```

```python
plt.plot(thetas, p)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.show()
```

### 데이터 프로세스 재구축


그림3.jpg

- 어떤 디자인안을 $N$번 표시했을 때 얻어지는 클릭 횟수 합계 $\alpha$를 고려
  - 단, 여전히 분석자는 디자인안 고유의 클릭률 $\alpha$를 알지 못함
- $N$번 모두 동일한 디자인안을 표시하고, 각 시행은 독립시행

  $→$ 독립 동일 분포 (Independent and Identically Distributed, iid)

<br>

#### 이항분포
- $\theta$의 확률인 시행이 발생하는 베르누이 시행을 $N$번 반복했을 때, 그 시행이 발생한 합계 횟수 $\alpha$는 이항분포를 따름

$$p(\alpha|\theta, N) = \binom{N}{a} \theta^\alpha (1-\theta)^{N-\alpha} $$

- 이항분포는 베르누이 시행의 성공확률 $\theta$, 시행횟수 $N>0$을 파라미터로 가짐
- $N=1$일 떄의 이항분포는 베르누이 분포와 동일

<br>


- $\theta$ 사전분포가 $\theta \sim Uniform(0, 1)$,

  $\alpha$의 가능도가 $Binomial(\theta, N)$일 때,

  $\theta$의 사후분포

```python
likelihood = lambda a, N: thetas ** a * (1 - thetas) ** (N - a)
```

```python
def posterior(a, N, prior):
  lp = likelihood(a, N) * prior
  return lp / lp.sum()
```

```python
prior = 1 / len(thetas)
plt.subplot(2, 1, 1)
plt.plot(thetas, posterior(2, 40, prior), label='A - A')
plt.plot(thetas, posterior(4, 50, prior), label='A - B')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.xlim(0, 0.2)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(thetas, posterior(64, 1280, prior), label='B - A')
plt.plot(thetas, posterior(128, 1600, prior), label='B - B')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.xlim(0, 0.2)
plt.legend()
plt.tight_layout()

plt.show()
```

## 01.06. 베타분포

- $\theta$의 사전분포가 $Beta(\alpha, \beta)$

$$p(\theta | \alpha, \beta) = \beta = \frac{\theta^{\alpha-1}(1-\theta)^{\beta-1}}{\int^{1}_{0} \theta^{\alpha-1} {(1-\theta)}^{\beta-1}} d\theta$$

- $alpha=1, \beta=1$일 때 Uniform Dist

<br>

- $\theta$ 사전분포가 $\theta \sim Beta(1, 1)$,

  $\alpha$의 가능도가 $Binomial(\theta, N)$일 때,

  $\theta$의 사후분포

- 사전분포와 사후분포는 켤레관계에 따라 모두 베타분포의 형태를 가짐

  - 사전베타분포의 파라미터가 각각 $\alpha. \beta$일 때,

    사후베타분포의 파라미터는 $\alpha^{'}=a+1, \beta^{'}=N-a+1$

```python
def betaf(alpha, beta):
  numerator = thetas ** (alpha - 1) * (1 - thetas) ** (beta - 1)
  return numerator / numerator.sum()
```

```python
def posterior(a, N):
  return betaf(a + 1, N - a + 1)
```

```python
plt.subplot(2, 1, 1)
plt.plot(thetas, posterior(2, 40), label='Alice - A')
plt.plot(thetas, posterior(4, 50), label='Alice - B')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.xlim(0, 0.2)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(thetas, posterior(64, 1280), label='Bob - A')
plt.plot(thetas, posterior(128, 1600), label='Bob - B')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.xlim(0, 0.2)
plt.legend()
plt.tight_layout()

plt.show()
```

## 01.07 사후분포를 이용한 의사결정
- **베이즈 추론으로 얻은 사후분포의 특징을 요약 통계량으로 수치화하거나, 샘플을 활용하여 정량적인 평가**

<br>

### 요약 통계량

<br>

#### 기댓값 (Expected Value, Mean)
- 확률변수에 확률의 가중치 (Weight)를 곱하여 평균을 낸 값

$$E[x] = \int^{\infty}_{-\infty} xp(x) dx\;\;\;(연속확률분포)$$

$$E[x] = \sum_{x}xp(x)\;\;\; (이산확률분포)$$


- 확률변수에 어떤 함수 $f(x)$를 시행해 변환한 것에 대한 기댓값

$$E[f(x)] = \int^{\infty}_{-\infty} f(x) p(x) dx \;\;\; (연속확률분포)$$

$$E[f(x)] = \sum_x f(x) p(x)\;\;\; (이산확률분포)$$

<br>

### 분산 (Variance)
- 확률변수와 그 기댓값 사이의 거리에 관한 기댓값

$$V[x] = E(x- E[x])^2$$

- **어떤 확률분포에 따르는 확률변수 $x$의 대부분이 기댓값 $E[x]$주변에 모여있으면, 그 사이의 거리 $(x-E[x])^2$는 작아지며,**

  **기댓값에서 떨어져 있을수록 그 거리는 커짐**

- **분산은 그 거리의 기댓값으로 대부분의 경우 '확률변수가 얼마나 퍼져 있는지' 알기 위한 지표로 사용**
- **분산의 양의 제곱근을 표준편차 (Standard Deviation)**

<br>

### 표본평균, 표본분산

<br>

#### 표본 통계량
- 확률분포로부터 얻은 샘플에 대한 통계량

<br>

#### 표본평균
- $n$개의 표본에 대하여, 표본평균은

$$\bar{x} = \frac{x_1 + \dots + x_n}{n} = \frac{\sum^n_{i=1} x_i}{n}$$

- 샘플이 독립동일분포고 $n$이 매우 크면, 샘플을 생성한 확률분포 $p(x)$의 기댓값 $E[x]$에 수렴

  $\rightarrow$ 대수의 법칙

$$n → \infty$$
$$\bar{x} \longrightarrow E[x]$$

<br>

#### 표본분산
- 샘플과 그 표본평균 사이의 거리의 평균

$$s^2 = \frac{\sum^n_{i=1} (x_i- \bar{x})}{n}$$

<br>

### HDI (Highest Density Interval)
- 확률변수의 값이 높은 확률로 나타나는 구간
- HDI는 연속확률분포에 대해 정의되는 값

- 일반적으로 확률분포 $p(x)$의 $\alpha \%$는 어떤 한계값 $t$보다 확률밀도가 높은 구간 $p(x) > t$로 표현

$$\int_{p(x) > t} p(x) dx = \alpha / 100$$




<br>

### HDI를 통한 가설검정

<br>

#### 각 디자인안의 클릭률의 사후분포 HDI

- 확률질량이 큰 순서대로 상윗값을 반환하는 HMV

```python
def hmv(xs, ps, alpha=0.95):
  xps = sorted(zip(xs, ps), key=lambda xp: xp[1], reverse=True)
  xps = np.array(xps)
  xs = xps[:, 0]
  ps = xps[:, 1]

  return np.sort(xs[np.cumsum(ps) <= alpha])
```

- ㄱ의 A안에 클릭률에 대한 베타사후분포 계산

```python
thetas = np.linspace(0, 1, 1001)

def posterio(a, N):
  alpha = a + 1
  beta = N - a + 1
  numerator = thetas ** (alpha - 1) * (1 - thetas) ** (beta - 1)

  return numerator / numerator.sum()
```

```python
ps = posterior(2, 40)
```

- 베타사후분포의 95% HDI
  - ㄱ의 A안의 클릭률 95% HDI는 대략 0.009에서 0.148의 구간 $→$ 클릭률 0.9% ~ 14.8% 사이에서 95%의 확률을 차지

```python
hm_thetas = hmv(thetas, ps, alpha=0.95)
```

```python
plt.plot(thetas, ps)
plt.annotate('', xy=(hm_thetas.min(), 0), xytext=(hm_thetas.max(), 0),
             arrowprops=dict(color='black', shrinkA=0, shrinkB=0, arrowstyle='<->', linewidth=2))
plt.annotate('%.3f' % hm_thetas.min(), xy=(hm_thetas.min(), 0), ha='right', va='bottom')
plt.annotate('%.3f' % hm_thetas.max(), xy=(hm_thetas.max(), 0), ha='left', va='bottom')
plt.annotate('95% HDI', xy=(hm_thetas.mean(), 0), ha='center', va='bottom')
hm_region = (hm_thetas.min() < thetas) & (thetas < hm_thetas.max())
plt.fill_between(thetas[hm_region], ps[hm_region], 0, alpha=0.3)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$p(\theta)$')
plt.xlim(0, 0.3)
plt.tight_layout()
plt.show()
```

<br>

#### ㄱ과 ㄴ의 각 디자인안 클릭률에 대한 95% HDI


```python
def plot_hdi(ps, label):
  hm_thetas = hmv(thetas, ps, 0.95)
  plt.plot(thetas, ps)
  plt.annotate('', xy=(hm_thetas.min(), 0),
                  xytext=(hm_thetas.max(), 0),
                  arrowprops=dict(color='black', shrinkA=0, shrinkB=0,
                                     arrowstyle='<->', linewidth=2))
  plt.annotate('%.3f' % hm_thetas.min(), xy=(hm_thetas.min(), 0),
                  ha='right', va='bottom')
  plt.annotate('%.3f' % hm_thetas.max(), xy=(hm_thetas.max(), 0),
                  ha='left', va='bottom')
  hm_region = (hm_thetas.min() < thetas) & (thetas < hm_thetas.max())
  plt.fill_between(thetas[hm_region], ps[hm_region], 0, alpha=0.3)
  plt.xlim(0, 0.3)
  plt.ylabel(label)
  plt.yticks([])
```

- **전체적인 경향을 볼 때, ㄴ의 디자인안은 ㄱ의 디자인안에 비해 클릭률의 HDI가 더 좁음**

- ㄱ의 A안과 B의 HDI는 겹치는  영역이 넓지만, ㄴ의 A안과 B안에서는 2개의 HDI가 겹치지 않음

### ROPE (Region of Practical Equivalence, 실질적 동등 구간)
- **웹사이트나 애플리케이션에 따라서 작은 오차는 오차범위라고 판단 가능**

  (예: [4.5, 5.5]는 5%와 같다고 판단하고, 이 구간과 HDI를 비교)

  $→$ **특정 수치 주변에 정의한 '실질적으로 같은'구간을 ROPE**


<br>

#### ROPE와 HDI에 관한 의사결정규칙 적용
- **대상값의 ROPE 전체가 HDI밖에 있는 경우에는 그 대상값을 기각(배제). 즉 다르다는 결론**
- **대상값의 ROPE 안에 전체가 포함되는 경우에는 그 값을 채택(선택). 즉, 같다고 결론**

    - **특히 95% HDI의 최댓값이 4.5보다 작다면, 디자인 B안의 클릭률은 5%보다 낮다**
  - **ROPE 안에 디자인 B안의 클릭률 95% HDI가 포함된다면, 디자인 B안의 클릭률은 5%**

### 신규확률변수
- 가설검정을 위해 새로운 확률변수 도입하여, 그 통계량을 판단에 활용

> **가설 : B안의 클릭률은 A안의 클릭률보다 높다**

- **A안과 B안의 차이를 나타내는 새로운 확률변수 정의**
  
  $→$ $\delta$의 확률분포 $p(\delta)$를 정의한 후, 95% HDI를 계산하여 클릭률의 차이를 얻는 범위를 고려하거나, 0과의 차이를 검토

$$\delta = \theta_B - \theta_A$$

#### 샘플 & 시뮬레이션

| 보고서 | 디자인 | $\alpha =\alpha+1$ | $\beta = N-\alpha+1$|
| - | - | - | - |

| ㄴ | A | 65 | 1217 |

  $→$ **샘플로부터 샘플을 생성한 원래 데이터의 확률분포 추론**
  
  (ㄱ의 A안의 클릭률 사후분포의 샘플에 대한 히스토그램)

#### 클릭률의 차이 $\delta$의 확률분포 추론
- ㄱ의 A안 베타사후분포 $Beta(3, 39)$, ㄱ의 B안 베타사후분포 $Beta(5, 47)$에서 각각 샘플을 생성

```python
theta_a = np.random.beta(3, 39, size=100000)
theta_b = np.random.beta(5, 47, size=100000)
```

- 두 샘플의 차이 $\rightarrow$ $\delta$의 샘플

```python
delta = theta_b - theta_a
```

- $\delta$의 분포는 -0.2 ~ 0.2 사이에 넓게 분포

```python
plt.hist(delta, range=(-0.3, 0.3), bins=60)
plt.xlabel(r'$\delta$')
plt.ylabel('Frequency')
plt.show()
```

- $\delta$의 분포에서 0보다 큰 영역의 넓이는 분포의 68.11%

  $→$ 95%보다 작음

  $→$ **B안의 클릭률은 A안의 클릭률보다 크지않다고 결론**

```python
print((delta > 0).mean())
```

### 지시함수 (Indicator Function)
- 지시함수 : 주어진 조건 $x$를 만족하면 1, 그렇지 않으면 0을 반환

$$I(x) = \begin{cases} 1\;\; \text{조건 만족} \\ 0 \;\;\text{조건 불만족}\end{cases}$$

<br>

- **0보다 큰 $\delta$의 샘플 비율을 계산하는 것은, $\delta$의 분포를 통하여 $I(\delta > 0)$의 표본평균을 계산하는 것과 동일**
- 또한, 대수의 법칙에 따라 기댓값 $E(I(δ > 0))$에 수렴

$$\frac{I(\delta_1 > 0) + \dots + I(\delta_n > 0)}{n} ⟶^{n → \infty} E(I(\delta > 0))$$

$$= \int^{\infty}_{-\infty} p(\delta) I(\delta > 0) dδ = \int^{\infty}_{-\infty} p(\delta) dδ$$

- **클릭률의 차이 $\delta$가 양수가 되는 확률을 구하는 것은,**

  **\$delta$의 분포에서 $\delta >0$의 영역의 넓이를 구하는 것에 해당**

<br>

### 확률변수의 변형

## 실험 설계의 기본 원칙
- 대조실험 (control Experiment)는 어떤 조작에 의해 주어지는 결과들을 비교해서 검증하는 것
- 일반적으로는 어떤 조건을 가한, 실험그룹 (Treatment)와 조건을 가하지 않은 대조 그룹 (Control)의 두 그룹을 준비하고, 두 그룹에 나타난 결과를 비교함으로써 조작의 효과를 검증
  - 웹사이트를 대상으로 실시하는 A/B테스트도 어떤 디자인이나 기능의 변경이 주는 효과를 비교해 검증한다는 점에서, 사용자를 샘플로 하여 대조 실험을 하는 것과 동일

- 이러한 비교 대조 실험을 효율적으로 수행하기 위한 기술을 실험 설계 (Desing Of Experiment, DoE)
  - 실험 설계는 원래 농업에서 적절하게 비료의 효과를 평가하기 위해 Fisher가 고안


<br>

### 실험설계의 3가지 원칙

1. **국소 관리의 원칙**
- 대상의 조작 원인만 평가하기 위해 실험 환경을 작게 나누는 것을 의미
- **즉, 실험환경을 작게 나누어, 평가하고 싶은 효과 이외에는 동일한 조건으로 억제**
  - 웹 최적화의 관점에서, 사용자를 작은 그룹으로 나누어, 각각 다른 안을 보여주는 것에 해당

<br>

2. **반복의 원칙**
- 같은 조건 하에 여러 샘플을 배치하여, 그룹 안의 샘플 사이의 변동 (오차 변동, 그룹내 변동)과 조작에 의해 발생하는 사이의 변동 (요인 변동, 그룹 간 변동)을 나누어 평가

<br>

3. **무작위 원칙**
- 국소 관리를 통해 제어할 수 없는 원인의 경우에는 무작위화
- 국소 관리의 원칙에 따라 사용자를 작은 그룹으로 나누고, 각각 다른 디자인안을 보여주었다 하더라도, 그룹을 나눈 방법에 어떤 규칙성이 있으면 결과를 올바르게 평가할 수 없음
