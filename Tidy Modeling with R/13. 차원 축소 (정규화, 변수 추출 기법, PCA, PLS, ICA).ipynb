# 13. 차원 축소
- 차원 축소는 데이터 세트를 고차원 공간에서 저차원 공간으로 변환하며 설명변수가 "너무 많다"고 의심되는 경우 좋은 선택
- 일반적으로 설명변수가 너무 많으면 더 높은 차원의 데이터를 이해하거나 시각화하기 어렵기 때문에 문제가 될 수 있음

<br>

## 13.01. 차원 축소
- 차원 축소는 변수 엔지니어링이나 탐색적 데이터 분석에 사용할 수 있음
    - 예) 고차원 생물학 실험에서 모델링하기 전에 첫 번째 작업 중 하나는 데이터에 원치 않는 추세(예: 연구실 간 효과와 같이 관심 문제와 관련되지 않은 효과)가 있는지 확인하는 것
        
        수십만 개의 차원이 있으면 데이터 디버깅이 어려우며, 차원 축소는 탐색적 데이터 분석에 도움이 될 수 있음
- **설명 변수가 많을 때 발생할 수 있는 또 다른 잠재적 결과는 모델에 해를 끼칠 수 있다는 것**
    - 예) 설명 변수의 수가 모델을 맞추는 데 사용되는 데이터 포인트의 수보다 작아야 하는 일반적인 선형 회귀와 같은 경우
    - 예) 설명 변수 간 상관 관계가 존재하는 데이터의 경우 모델에 부정적인 영향 (다중 공선성)
 
<br>

- **주성분 분석(PCA)은 선형 방법에 의존하고 지도되지 않기 때문에(즉, 결과 데이터를 고려하지 않음)** 
    
    데이터 세트의 열 수를 줄이는 가장 간단한 방법 중 하나
- 고차원 분류 문제의 경우 주요 PCA 구성 요소의 초기 플롯에서 클래스 간의 명확한 구분을 보여줄 수 있음
    
- PCA와 같은 방법은 새로운 변수의 더 작은 하위 집합을 사용하여 원래 설명 변수를 표현 
    
    $\rightarrow$ **새로운 변수를 생성하려면, 주성분을 구성하는 원래의 설명 변수가 모두 필요**
    
    $\rightarrow$ **이에 대한 예외는 새 변수를 생성할 때 설명 변수의 영향을 완전히 제거할 수 있는 능력이 있는 희소 방법**

## 13.02. 변수간 상관관계
- Koklu 및 Ozkan( 2020 )은 말린 콩의 시각적 특성에 대한 데이터 세트를 발표하고 이미지에서 말린 콩의 품종을 결정하는 방법을 설명
    
    (이러한 데이터의 차원은 많은 실제 모델링 문제에 비해 그리 크지 않지만 변수의 수를 줄이는 방법을 보여주는 예제를 제공)
    
> 본 연구의 1차 목적은 개체군 형태의 작물 생산으로부터 균일한 종자 품종을 얻을 수 있는 방법을 제공하는 것으로, 종자가 단독 품종으로 인증되지는 않는다. 따라서, 균일한 종자 분류를 얻기 위해 유사한 특징을 가진 7개의 서로 다른 등록 품종을 구별하는 컴퓨터 비전 시스템이 개발. 분류 모델을 위해 등록된 7종의 마른 콩 13,611개에 대한 이미지를 고해상도 카메라로 촬영
> 

<br>

> 각 이미지에는 여러 개의 콩이 포함
>
> 특정 콩에 해당하는 픽셀을 결정하는 프로세스를 '이미지 분할' 이라고 하며, 이러한 픽셀을 분석하여 색상 및 형태(예: 모양)와 같은 각 콩의 변수를 생성
>
> 다양한 콩 품종이 다르게 보이기 때문에 이러한 변수는 결과(콩 품종)를 모델링하는 데 사용
>
> 훈련 데이터는 수동으로 레이블이 지정된 이미지 세트에서 나오며, 이 데이터 세트는 Cali, Horoz, Dermason, Seker, Bombay, Barbunya 및 Sira 등 7가지 콩 품종을 구별할 수 있는 예측 모델을 만드는 데 사용
>
> 됩니다. 효과적인 모델을 생성하면 제조업체가 원두 배치의 균질성을 정량화하는 데 도움이 될 수 있음

<br>

- 물체의 모양을 정량화하는 방법에는 여러 가지가 있으며, ( Mingqiang, Kidiyo, Joseph 2008 ) . 대다수가 관심 대상의 경계나 영역과 관련되어 있음

    - 면적 (area, 또는 크기) 은 물체의 픽셀 수나 물체 주위의 볼록 껍질의 크기를 사용하여 추정할 수 있음

    - 경계의 픽셀 수와 경계 상자(객체를 둘러싸는 가장 작은 직사각형)의 면적을 사용하여 둘레(perimeter)를 측정할 수 있음

    - 주축 (major axis) 은 물체의 가장 극단적인 부분을 연결하는 가장 긴 선을 수량화하며, 단축(minor axis)은 주축에 수직인 선

    - 물체의 면적과 둘레가 같은 원의 면적의 비율을 사용하여 물체의 컴팩트함 (compactness) 을 측정할 수 있음
        - 예) “•”와 “×” 기호는 간결성이 매우 다릅니다.

    - 물체가 얼마나 길거나 직사각형인지에 대한 척도 (elongated)
        - 예) 이심률 (eccentricity) : 물체가 얼마나 원에서 벗어나는 정도 
        - 평면도(roundness) 와 볼록도 (convexity)에 대한 관련 추정치가 존재
        

- 각 이미지와 이심률 (eccentricity) 통계
    - 원이나 사각형과 같은 모양은 이심률이 낮은 반면 직사각형 모양은 높은 값을 가짐
    - 측정항목은 개체의 회전에 영향을 받지 않음
    - 이미지 특징 중 상당수는 높은 상관관계를 가지고 있음. 
        - 넓은 면적을 가진 물체는 둘레가 더 클 가능성이 높음

#### 정규화
- **대부분의 설명변수들은 비율변수이므로, 데이터의 분포가 치우쳐 있을 가능성이 존재**

    $\rightarrow$ **이러한 분포의 치우침은 분산의 계산 (예: PCA에서 사용)에 큰 영향을 줄 수 있음**
    
    $\rightarrow$ **설명변수 분포에 대해 정규화 적용**
    
<br>

1. 모든 수치형 설명변수에 대해 분산이 0인 경우 해당 변수를 제거
2. 모든 수치형 설명변수에 대해 순서 정규화를 수행. 변수의 값들을 순서대로 정규화하여 이상치의 영향을 줄임
3. 모든 수치형 설명변수에 대해 정규화를 수행하여, 각 변수의 평균을 0으로, 표준편차를 1로 만듬

```r
library(bestNormalize)
```

```r
bean_rec <- recipe(class ~ ., data = bean_train) %>%
    step_zv(all_numeric_predictors()) %>%
    step_orderNorm(all_numeric_predictors()) %>% 
    step_normalize(all_numeric_predictors())
```

## 13.05. 변수 추출 기법

- `recipe`를 검증용 데이터에 적용하여 산점도로 표시하는 함수 생성

```r
library(ggforce)
```

```r
plot_validation_results <- function(recipe, dat = bean_validation) {
    recipe %>%
        prep() %>%
        bake(new_data = dat) %>%
        ggplot(aes(x = .panel_x, y = .panel_y, color = class, fill = class)) +
        geom_point(alpha = 0.4, size = 0.5) +
        geom_autodensity(alpha = .3) +
        facet_matrix(vars(-class), layer.diag = 2) + 
        scale_color_brewer(palette = "Dark2") + 
        scale_fill_brewer(palette = "Dark2")
}
```

### PLS (부분최소제곱)
- PCA의 지도학습 버전 

<br>

#### 수학적 원리
- $X$를 설명변수, $Y$를 반응변수, $t$를 $X$의 선형조합, $w$가 선형조합의 가중치 일 때,

$$ t= Xw$$

$$cov(t, Y) = \frac{cov(t, Y)}{\sqrt{(var(t))}\sqrt{(var(Y))}} \sqrt{var(t)}\sqrt{var(Y)}$$

$$= corr(t, Y) \sqrt{var(t)}\sqrt{var(Y)}$$

$$\rightarrow \text{t와 Y의 공분산을 최대화하려면, t와 Y의 상관계수와 t의 분산을 최대화 해야 함}$$

$$max(cov(t, Y)) \propto max(corr(t, Y) \sqrt{var(t)}\sqrt{var(Y)})$$


- **$t$와 $Y$의 공분산을 최대하는 가중치 $w$**

$$cov(t, Y) = cov(Xw, y) = E[(Xw - E[Xw])(Y-E[Y])]$$

$$\text{PLS에 사용되는 데이터는 모두 정규화되어있기에,}\;\;E[Xw] = 0, E[Y] = 0$$

$$\rightarrow\; E[(Xw - E[Xw])(Y-E[Y])] = E[(Xw)(Y)] = \frac{1}{n}\sum^n_{i=1} (Xw)_i Y_i$$

$$= \frac{1}{n}(Xw)^T Y = \frac{1}{n}w^T(X^T Y)$$

$$∴ cov(t, Y) = \frac{1}{n}W^T(X^T Y)$$

<br>

$$\text{여기서}\; w^T(X^T Y)는\;w와\;X^TY의\;\text{내적으로 볼 수 있기에}$$

$$\rightarrow\; w^T(X^TY) = |w||X^T Y| cos\theta\text{이며}$$

$$\theta=0\text{일 때}\;cov(t, Y) \text{는 최대가 되어}$$

$$\rightarrow w=X^TY$$

$$\rightarrow \;w를\;t = Xw\;에 대입$$

<br>

#### PLS 성분 계산
1. 설명변수와 반응변수를 모두 정규화
2. 초기값인 $X_1$, $Y_1$를 설정하고, 위의 절차를 통하여 첫 번째 $t_1$값을 구함
    
    - 공분산이 최대가 되도록 하는 선형조합 가중치 $w_1$ 계산 ($w_1 = X_1^T Y$)
    
    $$w_1 = \frac{X_1^T Y}{X_1^T Y} \rightarrow |w_1| =1$$
    
    <br>
    
    - 가중치 $w_1$를 통하여 첫 번쨰 PLS 변수 $t_1$ 산출 $\rightarrow$ $t_1 = X_1 w_1$
        
        $$Y_1 = \beta_1 t_1 + F_1$$
    
    <br>
    
    - $\beta_1$은 회귀계수, $F_1$은 절편 $\rightarrow$ 최소제곱법을 적용하여 $\beta_1$를 추정
    
    $$\hat{\beta_1} = (t_1^T t_1)^{-1}t_1 Y_1$$

<br>

3. $t_1$값과 $Y_1$, $X_1$을 이용해 두 번째 $X_2$, $Y_2$를 설정
             
    - $t_1$으로 설명하지 못한 부분 (절편) $F_1$을 $Y_2$로 설정 (이전 단계에서 설명하지 못한 부분을 회귀식으로 다시 추정)
    
    $$Y_2 = F_1 = Y_1 - \beta_1 t_1 $$
    
    <br>
    
    - $t_1$이 기존 $X_1$에 대해서 설명하는 부분을 제외
    
    $$X_1 = t_1 P_1^T + E_1,\;\;\; P_1^T = (t_1^T t_1)^{-1}t_1 X_1$$
    
    $$X_2 = E_1 = X_1 - t_1 P_1^T$$
    
    - 공분산이 최대가 되도록 하는 선형조합 가중치 $w_2$ 계산
    
    $$w_2 = \frac{X_2^T Y_2}{|X_2^T Y_2|} \rightarrow |w_2|=1$$
    
    - $w_2$를 통하여 두 번쨰 PLS 변수 $t_2$ 산출 $\rightarrow$ $t_2 = X_2 w_2$
    
    - $t_2$의 회귀계수 $\beta_2$를 계산
    
    $$Y_2 = t_2 \beta_2 + F_2,\;\;\; \beta_2 = (t_2^T t_2)^{-1}t_2^T Y_2$$

<br>

4. 같은 방식으로 $k$번째 PLS 변수까지 추출
- 결국 계속해서 전체 데이터의 설명하지 못한 부분들 (partial)에 대해서 최소제곱법을 수행

$$\hat{Y} = \sum^k_{i=1} t_i \beta_i = t_1 \beta_1 + t_2 \beta_2 + \dots + t_k \beta_k$$

5. 가장 높은 성능을 보이는 $k$를 최적 PLS 성분 개수로 설정

<br>

### ICA (독립성분 분석)


- 아래의 식 $x = As$에서

    $$x_1(t) = a_{11}s_1(t)+a_{12}s_2(t)$$

    $$x_2(t) = a_{21}s_1(t)+a_{22}s_2(t)$$
    
    <br>
    
    **$s_1 (t)$와 $s_2 (t)$를 복원 해야 할 때, $A$가 mixing matrix라고 한다면,**
    
    $$s = A^{-1}x = Wx$$
    
    **unmixing matrix $W=A^{-1}$를 탐색**
    
<br>

#### 중심극한정리(CLT)와 ICA의 관계
- **CLT : 서로 독립적인 변수들의 선형조합으로 구성된 새로운 변수의 분포는 정규분포를 따름**
    
    (선형조합에 들어가는 변수들이 많을수록 더 정규분포에 가까워짐)

<br>

- **즉, ICA를 중심극한정리의 반대 과정이라고 할 수 있다면, ICA는 설명변수들의 조합으로 얻어진 $x$에 적절한 행렬 $W=A^{-1}$를 곱하여**
    
    **원래의 변수들인 $s$를 탐색하는 것**
    
    $\rightarrow$ **$s$들이 서로 독립적이라는 가정을 최대한 만족할 수 있도록 하는 $W=A^{-1}$를 탐색**

<br>

#### 선형변환에 따른 분포변화
- **$s$의 확률밀도함수가 $p_s(s)$일 때, $x = As$이고, $s = A^{-1}x = Wx$라면**
    
    **$x$의 확률밀도함수 $p_x(x)$는**
    
    $$p_x(x) = p_s(A^{-1}x) \cdot |A^{-1}|$$
    
    $$= p_s(s) \cdot |A^{-1}| = p_s(Wx) \cdot |W| $$
  

<br>

#### Bell-Sejnowski 알고리즘
- $s$의 원소 $s_i$들이 확률밀도함수 $p_s$를 따른다면, 모든 $s_i$로 구성된 결합분포는

    $$p(s) = \prod_{i=1}^{n}p_s(s_i)$$
    
    이 때, $x=As$의 확률밀도함수 $p_x$는
    
    $$p(x) = \prod_{i=1}^{n}p_s(w_i^Tx)\cdot|W|$$
   

- $m$개의 데이터 $\lbrace x^{(i)}; i = 1, 2, \cdots, m \rbrace$ 와 $n$개의 $s$에 대해 로그 가능도를 계산

    $$l(W) = \sum_{i=1}^{m}\left(\sum_{j=1}^{n}\log p_s(w_j^Tx^{(i)})+log|W|\right)$$
    
    $w_j^T$는 $W$행렬의 $j$번째 열
    
    $$W = \begin{bmatrix}
        { }- { } w_1^T { }- { } \\
        { }- { } w_2^T { }- { } \\
        \vdots \\
        { }- { } w_n^T { }- { } \\
        \end{bmatrix}$$

    <br>
    
- $s$의 분포에 대한 사전지식이 없다면 $\rightarrow$ $p_s$의 CDF를 $g(s) = 1 / (1+e^{-s})$로 가정 
    
    $p_s$의 확률밀도함수 $g^{'}(s)$이므로, $l(W)$는
    
    $$l(W) = \sum_{i=m}^{m}\left(\sum_{j=1}^{n}\log g'(w_j^Tx^{(i)})+log{|}W{|}\right)$$
    
<br>

-    행렬 $W$가 정방행렬(square matrix)이고 역행렬(inverse) $W^{-1}$이 존재하기에, 행렬식의 미분 규칙에 따라 $\nabla_W|W|=|W| (W^{-1})^T$
    
    $\rightarrow$ $l(W)$를 $W$에 대해 편미분
    
    $$\frac{\partial l}{\partial W} = \sum_{i=1}^{m}\left(\sum_{j=1}^{n}\frac{1}{g'(w_j^Tx^{(i)})}\cdot g''(w_j^Tx^{(i)})\cdot x^{(i)^T} + \frac{1}{|W|}|W|(W^{-1})^T\right)$$
    
    $$= \sum_{i=1}^{m}\left(\sum_{j=1}^{n}\frac{1}{g(w_j^Tx^{(i)})(1-g(w_j^Tx^{(i)}))}\cdot g(w_j^Tx^{(i)})(1-g(w_j^Tx^{(i)}))(1-2g(w_j^Tx^{(i)}))\cdot x^{(i)^T} + (W^{-1})^T\right)$$
    
    $$= \sum_{i=1}^{m}\left(\sum_{j=1}^{n}(1-2g(w_j^Tx^{(i)}))\cdot x^{(i)^T} + (W^{-1})^T\right)$$

- 경사하강법을 이용하여 $W$를 update ($\alpha$ = 학습률)
    
    $$W := W + \alpha \frac{\partial l}{\partial W}$$
    
    따라서, 각 $j=1, 2, \dots, n$에 대해
    
    $$W := W + \alpha\left(\begin{bmatrix}
        1 - 2g(w_1^Tx^{(i)}) \\
        1 - 2g(w_2^Tx^{(i)}) \\
        \vdots \\
        1 - 2g(w_n^Tx^{(i)})
        \end{bmatrix} x^{(i)^T} + (W^T)^{-1}\right)$$이며, 알고리즘이 수렴한 뒤에 얻은 $W$를 이용하여 $s$를 얻을 수 있음

<br>

#### Natural Gradient 알고리즘
- **Bell-Sejnowski 알고리즘은 $W$를 업데이트 할 때마다 계산해야하는 $(W^T)^{-1}$이 있어 계산속도가 느리다는 단점이 존재**

    $\rightarrow$ **Natural Gradient 알고리즘은 이를 보완**

<br>

- Bell Sejnowski 알고리즘의 $W$ 갱신에 대하여, 학습률 $\alpha$ 오른쪽 괄호항의 우측에 $W^T W$를 곱하면

    $$\left(\begin{bmatrix}
        1 - 2g(w_1^Tx^{(i)}) \\
        1 - 2g(w_2^Tx^{(i)}) \\
        \vdots \\
        1 - 2g(w_n^Tx^{(i)})
        \end{bmatrix} x^{(i)^T} + (W^T)^{-1}\right)W^TW$$
    
    $$=\left(\begin{bmatrix}
        1 - 2g(w_1^Tx^{(i)}) \\
        1 - 2g(w_2^Tx^{(i)}) \\
        \vdots \\
        1 - 2g(w_n^Tx^{(i)})
        \end{bmatrix} x^{(i)^T}W^TW + (W^T)^{-1}W^TW\right)$$
        
    $$=\left(\begin{bmatrix}
        1 - 2g(w_1^Tx^{(i)}) \\
        1 - 2g(w_2^Tx^{(i)}) \\
        \vdots \\
        1 - 2g(w_n^Tx^{(i)})
        \end{bmatrix} x^{(i)^T}W^T + I\right)W$$이며, $W$에 대한 update는 
    
    $$W := W + \alpha \frac{\partial l}{\partial W}W^TW$$,
    
    $$W:= W + \alpha \left(\begin{bmatrix}
        1 - 2g(w_1^Tx^{(i)}) \\
        1 - 2g(w_2^Tx^{(i)}) \\
        \vdots \\
        1 - 2g(w_n^Tx^{(i)})
        \end{bmatrix} x^{{(i)^T}}W^T + I\right)W $$


<br>

#### PCA와의 비교
- PCA와 ICA모두 공통적으로 데이터를 대표하는 기저벡터를 탐색한다는 점에서는 동일

<br>

1. **목적**
- PCA: 데이터의 공분산 구조를 기반으로 주성분(Principal Components)을 찾아 차원 축소 및 데이터를 변환하는 것이 목적. PCA는 데이터의 분산을 최대화하는 새로운 좌표계를 탐색
- ICA: 데이터를 잠재적인 독립적인 성분으로 분해하는 것이 목적. ICA는 데이터의 확률 분포에서 독립성을 추출하려고 시도

<br>

2. **성분의 성격**
- PCA: **PCA는 주어진 데이터 세트의 주요 변동성이 가장 큰 방향(주성분)을 탐색. 주성분은 데이터의 통계적 특징을 대표하는 축이며, 주성분 사이에는 상관 관계가 있을 수 있음.
- ICA: ICA는 데이터의 독립 성분을 찾으며, 이 **독립 성분은 서로 상관되지 않고 독립적인 신호 또는 요인**

<br>

3. **가정**
- **PCA: PCA는 데이터가 정규 분포를 따른다고 가정. 또한, PCA는 주성분 간에는 서로 직교한다는 가정**
- **ICA: ICA는 데이터가 정규 분포가 아니라고 가정. 또한, ICA는 성분 간에는 서로 독립이라는 가정**

## 13.06 모델링
- PLS와 UMAP 활용 모델링

<br>

#### 워크플로 생성 & 하이퍼파라미터 정의

```r
library(baguette)
library(discrim)
```

- 신경망

```r
mlp_spec <- mlp(hidden_units = tune(), penalty = tune(), epochs = tune()) %>%
    set_engine('nnet') %>%
    set_mode('classification')
```

- 배깅

```r
bagging_spec <- bag_tree() %>%
    set_engine('rpart') %>%
    set_mode('classification')
```

- 유연 판별분석 (FDA)

```r
fda_spec <- discrim_flexible(prod_degree = tune()) %>%
    set_engine('earth')
```

- 정규화 판별분석 (RDA)

```r
rda_spec <- discrim_regularized(frac_common_cov = tune(), frac_identity = tune()) %>%
    set_engine('klaR')
```

- 나이브 베이즈

```r
bayes_spec <- naive_Bayes() %>%
    set_engine('klaR')
```

<br>

#### 전처리 레시피 생성
- 분산이 0인 변수 제거
- 모든 수치형 변수에 대한 순서정규화
- 모든 수치형 변수에 대한 정규화

```r
bean_rec <- recipe(class ~ ., data = bean_train) %>%
    step_zv(all_numeric_predictors()) %>%
    step_orderNorm(all_numeric_predictors()) %>%
    step_normalize(all_numeric_predictors())
```

- PLS

```r
pls_rec <- bean_rec %>% 
    step_pls(all_numeric_predictors(), outcome = "class", num_comp = tune())
```

- UMAP

```r
umap_rec <- bean_rec %>%
    step_umap(
        all_numeric_predictors(),
        outcome = "class",
        num_comp = tune(),
        neighbors = tune(),
        min_dist = tune())
```

<br>

#### 모델 적합
- 하이퍼파라미터 튜닝 연산에 대한 병렬처리 설정

```r
ctrl <- control_grid(parallel_over = "everything")
```

- 평가지표를 AUC값으로 설정하여,
    
    각 차원 축소 방법 별 각 모델에 대한 하이퍼파라미터 튜닝

```r
bean_res <-  workflow_set(
    preproc = list(basic = class ~., pls = pls_rec, umap = umap_rec), 
    models = list(bayes = bayes_spec, fda = fda_spec,
                  rda = rda_spec, bag = bagging_spec,
                  mlp = mlp_spec)
    ) %>% 
    workflow_map(
        verbose = F,
        seed = 1603,
        resamples = bean_val,
        grid = 10,
        metrics = metric_set(roc_auc),
        control = ctrl)
```

<br>

#### 모델 평가


```r
rankings <- rank_results(bean_res, select_best = TRUE) %>% 
    mutate(method = map_chr(wflow_id, ~ str_split(.x, "_", simplify = TRUE)[1])) 
```

- 신경망 모델과, PLS-RDA모델이 가장 높은 성능을 보이는 것을 확인

```r
tidymodels_prefer()
filter(rankings, rank <= 5) %>% dplyr::select(rank, mean, model, method)
```

<br>
