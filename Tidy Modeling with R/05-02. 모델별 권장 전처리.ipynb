# 05-02. 모델별 권장 전처리

- 필요한 전처리 유형은 모델의 유형에 따라 상이
- `tidymodels`에서 적합 가능한 모델 목록
    
    https://www.tidymodels.org/find/parsnip/

<br>

### 모델별 필요한 전처리 기본수준에 대한 권장사항
- **`dummy`** : 범주형 변수에 대한 더미인코딩
- **`zv`** : 분산이 0인 수치형 변수 / 단일 수준 범주형변수 제거
- **`impute`** : 결측값 대체
- **`decorelate`** : 다중공선성 / 상관관계 완화 (차원축소, 변수제거, 정규화 회귀 등)
- **`normalize`** : 표준화 (Standard Scaling)
- **`transform`** : 변수의 분포 비대칭에 따른 변수변환 (박스-칵스, 로그 등)

<br>

**✔ : 설명변수에 해당 전처리가 필요**

**× : 설명변수에 해당 전처리가 불필요**

**◌ : 해당 전처리가 모델에 도움이 될 수 있지만, 필수는 아님**
