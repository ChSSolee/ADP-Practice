# 11. 경쟁위험 생존분석
#### 경쟁위험 : 여러 종류의 사건 중 하나만 경험할 수 있는 경우, 이들 사건의 발생확률
- 경쟁위험 생존데이터는 콕스모형에 적합시킬 수 있고, 모수적 생존모형에 적합 시킬 수 있으며, 생존이 아닌 누적발생률을 사용하는 모형에 적합시킬 수 있음

# 
#### 원인별 접근법 (cause-specific)
- **각 이벤트 종류별로 생존분석을 하되, 다른 종류의 이벤트는 중도절단된 것으로 처리하는 방법으로, 크게 2가지 문제가 존재**
1. **경쟁위험들이 서로 독립적이라는 가정이 필요**
    - 각 경쟁위험이 서로 독립적이라는 가정이 맞지 않는 경우, 경쟁위험을 동시에 분석할 수 있는 직접적인 방법은 없음
    - **민감도분석 : 경쟁위험 사건들을 모두 이벤트가 없는 것으로 처리하거나, 전부 우리가 관심 있는 이벤트를 경험한 것으로 처리**
2. **각 이벤트 종류별로 적합된 콕스모형의 Kaplan-Meier 생존곡선은 경쟁위험이 있을 경우 해석하는데 문제가 존재**
    - 누적위험모형에서 KM생존곡선의 일차적인 대안은 누적발생률함수 (cumulative incidence function, CIF)로 이벤트의 주변확률 (marginal probability)를 추정하는 곡선
    - CIF는 KM생존율과는 달리, 승법극한(product-limit)추정량이 아님
- 경쟁위험의 독립성 가정이 필요하기는 하지만, 경쟁위험의 독립성 여부와는 관계없이 생존곡선에 대한 의미 있는 해석을 가능하게 함
- CIF의 변형으로, 조건부확률함수 (conditional probability function, CPF)는 시간 t에서 다른 경쟁위험을 경험하지 않은 대상에서 위험의 조건부확률을 제공

# 
# 
### 11-1 경쟁위험의 예
#### 경쟁위험은 한 대상환자에서 한 가지 이상의 이벤트(실패)가 발생할 수 있지만, 그중 하나의 이벤트만 일어날 때 발생
- 어떤 환자는 폐암으로 사망하거나 심혈관질환으로 사망할 수 있으나, 두 가지 사망이 동시에 일어날 수 없다
- 암환자가 수술 후 병원감염으로 사망하거나, 병원감염이 일어나기 전 수술로 인해 사망할 수 있다
- 참전한 군인은 전투로 사망할 수도 있으나, 사고로 사망할 수 있다

#### 그러나, 사망사건이 아닌 경우에는, 추적기간 중 이벤트가 여러번 발생할 수도 있다 (경쟁위험의 재발사건)

# 
#### 경쟁위험 모형분석의 목적 : 관련있는 예측변수들과 가능한 몇 가지 경쟁위험의 발생률 또는 생존율의 관계를 평가. 또한, 관련 있는 예측변수들의 영향을 통제하고 경쟁위험들의 위험비를 비교
- 폐암과 심혈관질환의 경쟁위험 연구 
    - 심혈관질환으로 인한 사망가능성을 열어놓은 채, 어떤 물질에 노출된 환자의 폐암으로 인한 사망률과 노출되지 않은 환자의 폐암으로 인한 사망률의 차이가 있는지 연구
    - 다른 예측변수들의 영향을 통제한 상태에서 폐암으로 인한 사망률과 심혈관질환으로 사망률을 비교
- 암환자의 수술 후 사망과 병원내감염에 대한 경쟁위험
    - 감연관리에 관심이 있는 연구자에게는 원내감염이 관심사
    - **수술후 사망은 원내감염으로 인한 사망의 경쟁위험**
- 참전군인의 사망
    - 전투로 인한 사망과 사고로 인한 사망은 경쟁위험
    - 연구의 주된 관심사는 노출된 군과 노출되지 않은 군의 전투로 인한 사망의 위험비
    - 하지만, 부대 전체가 시간 t에 헬리콥터를 타고 전투지역으로 이동 중 사고로 모두 사망했을경우, 시간 t까지 전투로 인한 사망은 없었으므로, 전투에 의한 생존가능성은 1
    - 하지만 연구자가 관심있는 결과를 **전투 혹은 사고로 인한 사망**으로 할 경우, 시간 t이후 이벤트 없는 생존확률 (event-free survival rate)은 0

### 11-3 각각의 경쟁위험을 분리된 모형에 적합 (가장 전형적인 방법)
- 다른 경쟁위험들은 중도절단된 것으로 간주
    - 한 가지 실패 유형만이 일차적 관심사라면 다른 경쟁위험들은 중도절단된 것으로 처리, 실패 유형에 대한 위험비를 추정
#### 원인별 위험함수
### $h_c (t) = \lim_{\Delta t -> 0 } {{P(t \leq T_c < t + \Delta t | T_c \geq t)}\over{\Delta t}}$
- $T_c$ : 이벤트 c의 time-to-failure, c는 경쟁위험의 개수만큼 1, 2, ..., C까지 존재
- $h_c(t)$ : 시간 t까지 이벤트 c가 일어나지 않았다는 조건하에 시간 t에서의 순간적인 실패확률

# 
#### 콕스모형에 적합시키기 위해 p개의 예측변수 $X = X_1, X_2, ..., X_p$를 고려한 원인별 위험함수
### $h_c(t, X) = h_{0c}(t) \times exp(\sum^p_{i=1} \beta_{ic} X_i),\;\;\;\; c= 1, ..., C$

# 
#### 경쟁위험 중 status = 1에 대한 원인별 위험함수로 만든 콕스모형

### 11-4 독립성 가정
#### 중도절단에서 중요한 가정 : '중도절단은 독립적'
### $h(t | G, Ce) = h(t|G,NCe)$

($G$ = 시간 t에서 임의의 서브그룹에 속해 있는 대상 / $h(t|G, Ce)$ = 서브그룹 G의 중도절단된 대상들의 위험)
#### **시간 t에서 임의의 서브그룹 G에 속한 중도절단 대상의 위험은, 중도절단되지 않고 남아 있는 대상들의 위험과 같다는 가정**
- 독립적이지 않은 중도절단은 생존분석에서 치우친 결과(biased result)를 생성
- **중도절단된 사람들이 남아 있는 사람들에 비해 실패할 가능성이 더 많다면, 임의의 시간 t에서 생존확률의 참값보다 추정된 생존확률이 과대추정되어 편향된 결과를 만들 수 있음**

# 
#### 그렇다면 중도절단의 독립성 가정이 필요하다면, 독립성 가정이 만족되는지 어떻게 결정할 것인가
- 명시적으로 증명할 수 없음
# 

#### 독립성 가정이 만족되지 않을 경우 어떻게 분석을 진행할 것인가?
- 첫 번째 대안 : 데이터 분석에 의존하지 않고 다른 근거에 의해 독립성 가족을 만족한다고 판단 (임상적, 생물학적 등)
- 두 번째 대안 : 경쟁위험들에 공통되는 위험요인들을 고려하여, 생존모형에 포함
- **세 번째 대안 : 민감도분석**
    - 민감도분석은 독립성 가정을 만족하는지 명시적으로 증명할 수는 없지만, **독립성 가정에 위배되는 '최악의 사례'들로부터 넓은 추정구간을 결정하고, 그 결과가 독립성 가정하에 분석한 결과와 의미있게 다르지 않다면, 독립성 가정하에 얻은 결과에 약간의 편향만 존재한다고 결론 가능**

# 
#### 민감도분석
#### 독립성 가정하에 구한 모형과, 최악의 사례하의 모형간의 위험비를 비교 

# 

- 암으로 인한 사망을 주 관심사건으로 원인별 위험모형을 생성
    - 2가지의 '최악의 사례' 정의
    1. 기타원인에 의한 사망을 암으로 인한 사망으로 처리
    2. 기타원인에 의한 사망을 중도절단으로 처리하되, 그 데이터에서 가장 오래 기록된 생존시간까지 생존한 것으로 처리

# 
1. 기타원인에 의한 사망을 암으로 인한 사망으로 처리한 경우

````r
Byar2 = Byar %>% mutate(status2 = ifelse(status2 > 1, 1, status2)) # 기타원인으로 인한 사망을 암으로 인한 사망으로 처리
fit4 = coxph(Surv(dtime, status2 == 1) ~ Rx + Age + Wt + PF + Hx + Hb + SZ + SG, Byar2)
gaze(fit4) 
````

# 
2. 기타원인에 의한 사망을 중도절단으로 처리하고, 가장 오래 기록된 생존시간까지 생존한 것으로 처리한 경우

````r
Byar3 = Byar %>% mutate(dtime = ifelse(status2 > 1, max(dtime), dtime),
                        status2 = ifelse(status2 > 1, 0, status2))
fit5 = coxph(Surv(dtime, status2 == 1) ~ Rx + Age + Wt + PF + Hx + Hb + SZ + SG, Byar3)
gaze(fit5)
````

# 
- **fit1 Independent : 독립성 가정하에 구한 HR은 2가지 최악의 사례(fit4, fit5)에서 구한 HR의 구간 밖에 위치 (0.827 ~ 0.877)**
- **경쟁위험이 독립적이지 않다면, 치료효과에 대한 결론이 달라질 수 있음**
- **최악의 사례 1(fit4)은 최악의 사례 2(fit5)에 비해 독립성 가정으로부터 멀리 떨어져 있으나, 최악의 사례 2가 더 비현실적이므로 더 강조될만한 가치 존재**

````r
bind_rows(gaze(fit1)[1, ], gaze(fit4)[1, ], gaze(fit5)[1, ]) -> bind
rownames(bind) <- c("fit1 Independent", "fit4 WC1", "fit5 WC2")
bind
````

#### 결론적으로, 민감도분석은 독립성 가정을 평가하는 몇 가지 대안 중 하나이며, 독립성 가정이 위배되었을 때 분석결과가 얼마나 편향될 수 있는지 평가할 수 있게 해줌
#### 전형적인 생존분석 방법에서는 독립성 가정을 만족한다고 가정하고 분석을 진행

### 11-5 누적발생률함수
- 이전 장의 KM생존곡선은 한 종류의 실패(이벤트)만 있는 상황을 가정
- **경쟁위험이 있을 때 KM생존곡선은 위험이 하나일 때 처럼 유의한 결과를 제공하지 못함**

# 

#### 예1 : 암환자 100명을 5개월간 추적하고, 99명의 환자가 3개월째에 심혈관질환으로 사망하였고, 마은 1명의 환자는 암으로 사망(심혈관질환은 중도절단으로 간주)
    
| Months | Died | Cause | 
| - | - | - |
| 3 | 99 | 심혈관질환 |
| 5 | 1 | 암 |

# 
- $S(t_f)$ : 각 실패시간의 원인별 KM생존율
    - 5개월째 위험대상 수는 1명, 이 환자가 5개월째에 이벤트가 발생했으므로, 5개월째의 KM 생존율
    - **$KM_{Ca}:S_{Ca}(t = 5) = S(t = 4) \times Pr(T > 5 | T\geq 5) = 1 \times 0 = 0$**
    - 5개월째에 암으로 인한 원인별 사망률 : $1 - S_{Ca} = 1$, 99명의 대상환자가 암이 아닌 심혈관질환으로 사망하였으므로, 처음 100명의 대상환자 중 암으로 사망한 비율은 0.01
# 

| $t_f$ | $n_f$ | $m_f$ | $q_f$ | $S(t_f)$ |
| - | - | - | - | - |
| $0$ | $100$ | $0$ | $0$ | $1$ | 
| $3$ | $100$ | $0$ | $99$ | $1$ |
| $5$ | $1$ | $1$ |  |  $0$ |

# 
#### 예2 : 3월째에 사망한 99명의 대상환자들이 3개월째 심혈관질환으로 사망하지 않았다면, 모두 5개월째에 암으로 사망하였을 것이라고 가정
- 5개월째의 KM생존율은 원래 데이터와 마찬가지로 0
    - 5개월 생존율이 0으로 같은 이유는 KM방법은 중도절단이 독립적이라고 가정(independent censoring)
    

| $t_f$ | $n_f$ | $m_f$ | $q_f$ | $S(t_f)$ |
| - | - | - | - | - |
| $0$ | $100$ | $0$ | $0$ | $1$ | 
| $3$ | $100$ | $0$ | $0$ | $1$ |
| $5$ | $100$ | $100$ | $0$ |  $0$ |

# 
#### 예3 : 3개월째에 사망한 99명의 환자가 사망하지 않고 5개월까지 생존해 있다가 연구가 종료
- 1 - 0.99 = 0.01은 전체 대상환자 중 심혈관질환으로 사망 여부를 고려하지 않을 경우 암으로 인하여 사망하는 환자들의 비율로 주변확률 (marginal probability) 

| $t_f$ | $n_f$ | $m_f$ | $q_f$ | $S(t_f)$ |
| - | - | - | - | - |
| $0$ | $100$ | $0$ | $0$ | $1$ | 
| $3$ | $100$ | $0$ | $0$ | $1$ |
| $5$ | $100$ | $1$ | $99$ |  $0.99$ |

#### 누적발생률함수 (cumulative incidence function) 
- 경쟁위험 데이터에서 KM생존곡선의 대안 중 하나로 앞 절에서 소개한 주변확률을 사용
    - 경쟁위험이 하나인 경우 $CIF = 1 - KM$
    - 경쟁위험이 2개 이상인 경우 CIF는 원인별 위험함수에서 추정
- 경쟁위험이 있는 경우 위험의 주변확률로 경쟁위험이 독립적이라는 가정이 필요없음

# 
**누적발생률곡선 계산**
1. 관심 있는 이벤트를 c라고 할 때, 정렬된 실패시간 $t_f$에서의 위험
- $\hat{h_c}(t_f)  = m_{cf} / n_f$

($m_f$는 시간 $t_f$에서 발생한 이벤트 c의 개수이며, $n_f$는 $t_f$에서의 위험대상 수)

2. 실패시간 $t_f$에서 실패가 발생하려면 대상환자가 실패가 일어나기 전 시간까지 살아 있어야 하므로 $S(t_{f-1})$을 추정. $S(t_{f-1})$은 바로 전의 실패시간 $t_{f-1}$에서 경쟁위험이 모두 없는 총 생존확률

3. 시간 $t_f$에서의 이벤트 c의 실패발생률 $(\hat{I_c}(t_f))$을 추정

- $\hat{I_c}(t_f) = \hat{S}(t_{f-1}) \times \hat{h_c}(t_f)$

4. **시간 $t_f$에서의 이벤트 c의 누적발생률($CIF_c$)은 실패시간 $t_1$부터 $t_f$까지의 실패발생률의 누적 합계**

- $CIF_c(t_f) = \sum^f_{f' = 1}\hat{I_c}(t_{f'}) = \sum^f_{f' = 1} \hat{S}(t_{f' - 1})\times \hat{h_c}(t_{f'})$

# 
#### 경쟁위험이 없는 경우 $CIF = 1 - KM$이지만, 경쟁위험이 있는 경우에는 $CIF_c(t_f) \neq 1 - KM_c$
- **$1-KM_c$는 원인별 생존함수 $S_c(t)$를 계산하는데 쓰임** (경쟁위험으로 인한 실패를 중도절단으로 처리, **중도절단법(censored method)**)

#### $1 - KM_c = \sum^f_{f' = 1}\hat{S_c}(t_{f' - 1} \times \hat{h_c}(t_{f'}))$

# 
#### 누적발생률함수 계산
- ```status``` : 암에 의해 사망 = 1 / 다른원인에 의해 사망(경쟁위험) = 2 / 생존 = 0

# 
#### Fine & Gray CIF모형
- **콕스비례위험모형과 유사하지만, CIF에서 파생된 sub-distribution 위험함수를 사용**
- **표준적인 콕스비례위험모형과 같이 해석**

# 

#### 경쟁위험 c에 대한 sub-distribution 위험함수 ($h_{c, CIF}(t)$)는 다음과 같이 정의
- $h_{c, CIF}(t) = lim_{\Delta t = 0} {{pr(t < T_c < t + \Delta t | B)}\over{\Delta t}}$
    
    ($Where\;B\;=\;T_c > t\;or\;T_{c'},\;c' \neq c$)
# 
- 즉, B는 시간 t 이후에 경쟁위험 c가 발생하거나 ($T_c > t$)시간 t이전에 c가 아닌 다른 경쟁위험으로 인해 실패한 경우 $(T_{c'}\leq t, c' \neq c)$
#### $h_{c, CIF}(t) = {{\delta \;CIF_c(t)\;/ \;\delta t}\over{1 - CIF_c (t)}}$

# 
#### 경쟁위험이 없는 경우와 경쟁위험이 있는 경우의 유사성
- $CIF_c(t)$는 다른 경쟁위험으로 인한 실패를 중도절단이 아니라, 실패로 간주하여 총 생존 $S(t)$를 사용

| $Competing\;Risks$ | $No\;Competing\;Risks$ |
| - | - |
| $h_{c, CIF}(t)$ | $h(t)$ |
| $\delta CIF_c(t)\;/\;\delta t$ | $f(t) = \delta F(t)\;/\;\delta t\;\;\;where\;\;\;F(t)=Pr(T \leq t)$ |
| $1-CIF_c(t)$ | $S(t) = 1- F(t)$ |

# 
#### Fine & Gray CIF모형에서 시간 t에서의 경쟁위험 c의 위험함수 $h_{c, CIF}(t)$
#### $h_{c, CIF}(t) = h_{0c, CIF}\times exp(\sum^p_{i=1} \gamma_i X_i)$
# 
#### 콕스비례위험모형의 위험함수 $h(t)$ 대신 subdistribution 위험함수 $h_{c, CIF}(t)$를 사용한 점을 제외하고는 콕스모형과 유사
#### $HR_{c, CIF}(X^*, X) = exp[\;\sum^p_{i=1}\gamma_i (X^*_i - X_i)\;]$
- 콕스비례위험모형의 비례위험 **$HR_{CoxPH}(X^*, X) = exp[\;\sum^p_{i=1} \beta_i (X^*_i - X_i)]$**의 $\beta$가 $\gamma$로 바뀐 것을 제외하면 동일
- FIne & Gray CIF모형에서도 시간의존공변량 등 비례위험 가정을 만족하지 않는 변수에 대해 확장하는 것이 가능

### 11-6 누적조건부확률함수 (cumulative conditional probability function, CPF)
- $CPF_c$ : 시간 t까지 다른 경쟁위험들을 경험하지 못한 대상에서 이벤트 c를 경험할 확률
    - $T_C$ : 이벤트 c가 일어날 때까지의 시간
    - $T$ : 다른 경쟁위험 이벤트가 일어날 때까지의 시간
    - $CPF_c$ : $CIF_c$를 $1-CIF_c$로 나눈 값 ($CIF_{c'}$ = 경쟁위험 c이외의 다른 위험들의 누적발생률)
        - **$CPF_c = CIF_c / (1 - CIF_{c'})$**
        
#### $CPF_c = Pr(T_c \leq t | T \geq t)$
