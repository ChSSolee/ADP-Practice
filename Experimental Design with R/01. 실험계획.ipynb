# 01. 실험계획

<br>

## 01.01. 실험계획
**실험** : 실험대상에 어떤 처리를 하여 나온 실험 자료를 분석하여 새로운 결론을 얻는 과정
- 실험의 목적을 이루기 위해서는 면밀한 계획과 실행, 자료에 대한 정확한 분석이 필수

<br>

**실험설계** : 시스템의 성격을 알기 위하여 실험을 수행할 때, 과연 어떤 식으로 실험을 수행해야 효율적인 정보획득이 가능한지, 어떻게 처리해야 얻은 정보에 대한 분석 및 해석이 적법한지 다루는 통계 방법

<br>

**실험의 예시**
> 화학 플랜트에서 정화시간에 유의한 영향을 미치는 중요 요인이 무엇인지를 선별

<br>

> 염색 온도에 민감한 염색 색상의 명암관의 관계 규명

<br>

> 3종류의 진통제 에 대한, 진통제 간 유의한 차이 비교 실험

<br>

#### 실험의 요인
> 입력 $\rightarrow$ 제어 가능한 요인 [$X_1, \dots X_p$] + 잡음 [$Z_1, \dots, Z_p$] $\rightarrow$ 출력 ($y$)

<br>

- **제어 가능한 요인 (factor)** : 제어가 용이하고, 재현성이 있으며, 기술적 의미를 갖는 요인
- **잡음 (noise)** : 출력에 영향을 미치지만, 요인의 상태를 파악할 수 없는 것. 측정오차와 같이 이유를 파악하기 어려우나 출력의 산포에 영향을 미치는 것
- 계획된 실험이란 제어 가능한 요인을 연구자가 조절하였을 때, 출력이 어떻게 변화하는지를 관찰하고 조사하여 프로세스를 보다 잘 이해하려는 시도
- 통계자료는 조사(survey)나 실험자료(experimental)로 분류
    - 조사자료는 어떤 현상이나 시스템에 대한 정보가 관찰이나 면접을 통하여 수집. **어떤 현상을 있는 그대로 파악하는 것이 목적이고, 환경조건을 관리하는 일이 비교적 드물다고 할 수 있음**
    - 실험자료는 통계적으로 계획되고, 관리된 실험을 통하여 얻을 수 있는 자료
    
<br>

#### 어떻게 실험을 수행해야 최소의 노력 또는 비용으로 최대의 정보를 얻는 가에 관심
- 실험계획은 실험에 의한 정보의 획득 및 관리 행위이며, 정보의 효율적 획득에 초점

<br>

## 01.02. 실험계획의 기본원리
- 실험계획 방법에 따라서 얻어진 자료에 대해 통계적 분석방법이 달라지므로, 실험 계획과 통계적 분석은 서로 밀접한 관계
    - 랜덤화, 블록화, 반복, 교락, 직교화 등

<br>

### 랜덤화 (randomization)
- 실험단위의 배정 또는 실험순서를 임의로 또는 랜덤으로 결정
- **선택된 요인 외에 기타 원인들의 영향이 실험결과에 편향되는 것을 막기 위한 방안**
    - 만약 실험을 랜덤하게 하지 않고, 규칮걱으로 차례대로 행한다면, 실험에서 고려하지 않은 원인이 실험결과에 영향을 끼칠 수 있으므로, 어느 것이 참다운 원인인지 모르게 됨 (랜덤화는 선택된 요인 외에 기타 원인들의 영향을 어느 정도 상쇄시킬 수 있으므로 객관성 유지 가능)
- 실험자료의 통계적 분석방법은 실험결과나 오차가 서로 독립인 확률변수라는 가정 $\rightarrow$ 랜덤화는 이러한 가정을 충족

<br>

### 블록화 (blocking)
- 실험 전체가 이질적인 경우, 여러 블록으로 나누어 **(블록 내에서는 동질적으로 / 블록간에는 이질적으로)**, 각 블록 내에서 요인효과를 조사하는 것을 블록화
    - 조사연구에서의 층화(stratification), 품질관리에서의 층별과 같은 개념
    
* 블록화를 하면 블록 내에서는 동질성이 유지되며, 블록 간에는 이질적이 됨
* **블록화는 관심있는 요인의 효과차이를 더 뚜렷하게 밝혀내기 위한 것으로, 블록간 차이가 크면 클수록 오차분산을 줄일 수 있어서 효율성이 증대**

<br>

### 반복 (replication)
- 동일 조건하에서 실험을 두 번이상 시행
- 실험오차를 추정 가능 : 관측된 요인효과의 차이가 통계적으로 유의한 차이인지 판단하는 근거
- 반복으로 통해 요인효과를 추정하는데 평균을 사용할 수 있으므로, 더 정밀한 효과 추정량을 얻을 수 있음
    - 단, 반복실험은 실험의 횟수와 비용이 증가
- 실험이 관리상태하에서 동일한 조건에서 비슷한 실험결과가 나온다면, 비교적 작은 반복수도 무난

<br>

### 교락 (confounding)
- 구할 필요가 없는 2인자 교호작용이나 고차의 교호작용을 블록과 교락화 시키는 방법
- 통상적으로 높은 차수의 교호작용은 크게 의미가 없으며, 높은 차수의 교호작용과 블록을 교락시킴으로써 실험의 효율을 높임

<br>

### 직교화 (orthogonization)
- **직교성** : 상호 독립성. 한 요인의 효과와 다른 요인의 효과를 독립적으로 추정할 수 있음
- **직교화란 실험 계획 시 요인을 배치할 때 요인 간 직교성을 갖도록 실험을 배치하는 것**
- 같은 실험횟수라도 검출력이 높은 실험을 계획할 수 있으며, 정도(precision)가 높은 추정을 할 수 있음

(직교배열표는 요인 간 직교성이 성립하도록 만든 표로 일부실시법에서 유용)

<br>

## 01.03. 실험계획의 순서
<br>

### 문제에 대한 이해와 목적의 설정
- 연구대상과 무엇이 문제인지를 명확히 하여, 실험의 목적을 구체적으로 설정

<br>

### 반응변수의 선택
- **특성값 (characteristiv value)/반응변수 (response variable)** :실험을 실시한 후에 자료의 형태로 얻는 값

<br>

#### 반응변수 선택의 고려사항
1. 반응변수값을 재는 측도 선택
2. 반응변수에 대한 측정시스템의 타당성
3. 반응변수 개수 선택

<br>

### 요인과 수준의 선택
- **요인 (factor)** : 반응변수값에 영향을 미치는 원인 중, 실험에 취급된 원인
- **요인 수준 (level)** : 실험을 하기 위하여 설정한 요인의 실험 조건

<br>

1. 초기단계의 실험에서는 특성요인도 (cause and effect diagram) 등을 작성하여 반응변수에 영향을 미칠 것으로 생각되는 모든 요인 중에서 적절한 요인을 선택
- **이 때 요인은 될 수 있는 대로 서로 독립적이라고 생각되는 것을 택하는 것이 좋음**
2. **선택된 요인의 수가 너무 많은 경우에는 실험의 크기가 커져 비용이 많이 들고, 시간이 오래 걸리며 실험을 관랗기 어려워지는 등 많은 문제가 발생**
- **요인을 너무 축소하여, 중요한 요인이 빠지면 실험 목적 달성에 어려움**
3. **수준은 연구자의 관심영역 (region of interst) 내에서 선택하게 되는데 수준간격을 너무 넓게 잡으면, 요인수준의 조합에서 생기는 효과인 상호작용 또는 교호작용 (interaction)이 지나치게 너무 많이 일어나 실험 자체가 의미가 없어지는 경우가 존재**
- **반면에 너무 좁게 잡으면 수준 간의 차이가 없게 되어 실험의 효율이 떨어짐**
4. 실제로 사용하기 어려운 조건은 실험의 수준에서 제외시키고, 최적이라고 예상되는 조건은 실험의 수준에 포함

<br>

### 실험계획의 선정
- 구체적인 실험계획 수립
1. 요인과 수준이 선택되면 실험오차, 비용, 요구되는 통계적 정밀도 등을 고려하여 반복수를 결정
- 요인의 수준조합과 반복수가 결정되면 총실험횟수를 결정

2. 어떻게 랜덤화할 것인지 결정하여 실험순서를 결정
- **실험 전체를 랜덤한 순서로 행하는 것이 원칙이지만, 어떤 요인이 수준변경이 곤란하다면, 분할구계획을 사용**
3. 블록화를 고려해야 하는지를 결정, 어떤 것을 블록으로 취합할 것인지 결정

<br>

### 실험의 수행

- 실험 수행 시, 계획한 대로 단게별로 제대로 진행되는지 잘 관리

<br>

### 자료분석과 결과 검토
1. 실험결과로 얻은 자료는 실시된 실험계획에 대응되는 모형에 따라 적법하게 분석
2. 그래프 분석을 통하여 결과에 대한 대략적인 결론을 도출하고, 실험계획에서는 주로 분산분석을 통하여 실험자료를 분석

<br>

### 결론과 조치
1. 어떤 시스템의 최적조건을 구하는 것이 목적일 시 : 최적조건에 대하여 기술적, 경제적으로 문제가 없는지 검토하고, 재현성을 갖는지 실험한 후 이 조건이 유지될 수 있도록 사후관리
2. 결론과 실패한 것 까지도 다음 실험에 반영

<br>

<hr> 

### 가상 시나리오
> **1. 문제의식**
> 전자부품을 만드는 회사에서, 로트당 불량률 조사 결과 1%로 확인
> 원가절감을 위해서 로트당 불량률을 줄이는 것이 시급한 문제임을 인식하고, 최적공정조건을 도출하기로 결정

<br>

> **2. 특성값과 요인의 선택**
> 엔지니어와 공정관리자들은 사실파악
> - 불량원인에 대한 파레토 (Pareto Diagram)를 그려본 결과, 불량원인의 80% 정도가 전체 공정 중 마지막 납땜과정에서 일어나는 납땜불량으로 파악 $\rightarrow$ **납땜불량을 개선하기로 프로젝트의 범위를 축소하고, 특성값을 납땜불량률로 결정**
> 납땜불량에 대한 특성요인도를 그려, 납땜온도와 첨가제 함량, 라인속도가 잠재적인 요인으로 예상
> 실험을 통해 납땜불량에 큰 영향을 미치는 핵심요인을 선별하고 최적조건을 구하기로 결정

<br>

> **3. 실험의 계획**
> 요인에 대한 실험조건인 수준이 납땜온도와 첨가제 함량인 경우, 현재 작업조건을 기준으로 하여 3수준으로 하고, 라인속도는 현재 조건과 약간 늦춘 것을 비교하기 위하여 2수준으로 결정
> 공정이 비교적 안정되어 있으므로, 18 ($3\times3\times2$)개의 처리조합에서 반복없이 실험전체를 랜덤한 순서로 실험

<br>

> **4. 실험실시**
> 작업이 없는 일요일에 진행하고, 실험조건만 변화하여 나머지 환경은 가능한 한 균일 조건을 이룸

<br>

> **5. 얻어진 자료의 분석결과**
> 납땜온도와 첨가제의 함량은 불량률에 유의한 영향을 미침
> 라인속도를 늦출 경우 불량률이 줄기는 하지만 예상보다 크게 줄지는 않음
> 얻어진 자료를 바탕으로 반응표면분석을 이용하여, 등고선 그림을 작성한 결과 납땜온도와 첨가제의 함량에 대한 최적조건을 구함

<br>

> **6. 결론 및 조치**
> 라인속도는 생산성에 영향을 미치므로 현행대로 유지
> 납댐온도와 첨가제의 함량에 대한 최적조건에서 재현성 실험을 한 결과, 재현성이 확인
> 구해진 최적조건을 바탕으로 새로운 공정조건과 허용한계를 설정하고, 문서화
> 실험을 통하여 로트당 불량률이 개선됨을 확인

<br>

> **7. 새로운 시작**
> 새로 개선된 공정조건하에서 자재 원산지 간의 비교실험을 실시하여 분산분석한 결과 유의한 차이를 발견하지 못함

<br>

## 01.04. 실험계획 시 유의사항
1. 주어진 문제에 대한 기술적인 지식을 최대한 활용
2. 계획과 분석은 가능한 한 간단한 것을 사용
3. 실제적 차이와 통계적 차이를 구분
4. 통계적 분석 결과가 기술적 지식, 상식 등과 상반되게 나올 경우
- 예) '오차가 서로 독립인 정규분포를 따른다' 등의 통계적 가정에 대한 검토 필요
- 이상치나 영향을 크게 미치는 측정값을 식별함으로 써 분석결과를 음미할 수 있는 자료에 대한 검토가 필요
