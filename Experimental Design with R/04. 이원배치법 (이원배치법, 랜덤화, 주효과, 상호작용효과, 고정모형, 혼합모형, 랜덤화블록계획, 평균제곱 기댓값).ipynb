# 04. 이원배치법
- 반응치에 영향을 미칠 것으로 기대되는 요인의수가 2개인 경우
- 요인 A와 B의간의 상호작용효과, 주효과와 상호작용효과의 유의성을 검토

<br>

## 04.01. 이원배치법 개요
- **반응치에 영향을 미칠 것으로 기대되는 요인의수가 2개인 경우**
    - 요인 $A$와 $B$의 수준수가 $a$, $b$인 경우 수준조합의 수, 총 실험조건의 수는 $a\times b$
    - 각 조합마다 반복수가 $r$인 경우, 전체 실험의 횟수는 $a\times b \times r$
- **두 요인이 고정요인인 경우에는 $a \times b \times r$회의 전체 실험을 랜덤하게 선택**
    - **이는 요인 $A$의 한 수준을 먼저 랜덤하게 선택하고, 요인 $A$의 선택된 수준에서 다음으로 다른 요인 $B$의 수준을 랜덤하게 선택하여 $a\times b$회의 실험을 $r$회 반복 실험하는 분할구계획과 다름**

<br>

**이원배치법의 가장 커다란 관심사의 하나는 요인 $A$, $B$의 주효과 이외 두 요인 간의 상호작용효과가 존재하는가의 여부**
- **요인의 수준 간 차이가 반응변수 값에 영향을 미친다면 $\rightarrow$ 해당 요인은 효과가 존재 (주효과가 존재)**
- **온도 $A$와 습도 $B$의 변화가 반응치(강도)에 미치는 영향을 보고자 할 때,* *

    **두 요인 간 상호작용효과는 한 요인의 낮은 수준 ($A_1$)에서 다른 요인의 수준 간 차이 ($B_1 \rightarrow B_2)$로 인한 반응치의 강도가**

    **그 요인의 높은 수준 ($A_2$에서 다른요인의 수준 간 차이 ($B_1 \rightarrow B_2)$로 인한 반응치의 차이와 서로 얼마나 다른지에 따라 결정**

<br>



1. $A_1$에서 $B_1 \rightarrow B_2$로 변할 때 강도가 줄어드는 정도가

    $A_2$에서 $B_1 \rightarrow B_2$로 변할 때 강도가 줄어드는 정도와 크게 다르지 않음
    
    $\rightarrow$ **두 요인간 상호작용효과가 없음**
    
<br>

2. $A_1$에서 $B_1 \rightarrow B_2$로 변할 때에는 강도가 더욱더 많이 줄어듬

    **$\rightarrow$ 두 요인 간 상호작용은 어느정도 존재**
    
3. $A_1$에서 $B_1 \rightarrow B_2$로 변할 때 강도가 증가하지만

    $A_2$에서 $B_1 \rightarrow B_2$로 변할 때에는 오히려 감소
    
    **$\rightarrow$ 두 요인간 상호작용 효과는 매우 큼**
    

#### 반복측정이 있는 경우, 실험 전체의 최적조건을 구할 때에는 두 요인간 상호작용효과가 있는지 확인
- **두 요인 간 상호작용효과가 거의 없는 경우, 각 요인의 최적조건이 곧 실험 전체의 최적조건**
- **두 요인 간 상호작용효과가 큰 경우에느, 한 요인의 특정 수준에서 다른 요인의 최적수준을 정해야 함**

<br>

## 04.02. 실험의 랜덤화
- 실험계획의 중요한 목적 중 하나는 분산분석의 결과가, **반응치의 모평균 간에 차이가 난 원인이 실험계획에서 고려한 요인수준의 차이에만 기인한다는 결론을 내릴 수 있도록 함**
* **처리효과의 차이를 제외하고, 나머지 실험조건은 동질적이 되도록 노력** 

    **즉, 동질적이 되지 못하게 하는 제 3의 중요한 요인이 존재하지 않도록 해야함**
    
    - **이 효과가 각각의 처리에 골고루 적용해서, 서로 상쇄되도록 도와주는 방법이 실험의 랜덤화**
    
<br>

#### 이원배치법의 실험의 랜덤화는 각 처리가 요인 $A$, $B$의 수준조합인 $A_iB_j$에 의해서 결정된다는 점을 제외하고 일원배치법과 동일
- 일원배치법에서의 실험의 랜덤화와 마찬가지로, 처리를 실험단위에 배치하는 순서 또는 실험실시 순서는 **실험자의 의지에 관계없이 랜덤한 메커니즘에 의해 결정되는 완전확률화계획법**
- 요인 $A$,$B$의 수준수가 2, 3이고 각각의 처리에서 반복수가 2회인 총 실험횟수 $2 \times 3 \times 2 = 12$인 이원배치법 실험의 경우

| 요인 | $B_1$ | $B_2$ | $B_3$ |
| - | - | - | - |
|$A_1$ | 1, 2 | 3, 4 | 5, 6|
|$A_2$ | 7, 8 | 9, 10 | 11, 12 |

- 1단계 : 각각의 실험조건 (요인의 수준조합)에 1에서 12 사이의 일렬번호를 할당
- 2단계 : 1에서 12까지 12개 숫자의 랜덤배열 계산
- 3단계 : 2단계에서 구한 랜덤배열 실험조건의 순서대로 실시

<br>

## 04.03. 고정모형 (A, B : 고정요인)

<br>

### 구조모형 (반복 있는 경우) : 2개의 요인이 모두 고정요인인 경우
#### 이원배치법의 자료구조

| $요인$ | $B_{1}$ | $B_{2}$ | $\dots$ |$B_{b}$ |
| - | - | - | - | - |
| $A_1$ | $x_{111} \dots x_{11r}$ | $x_{121} \dots x_{12r}$ |  $\dots$ | $x_{1b1} \dots x_{1br}$ |
| $\vdots$ | $\vdots$ | $\vdots$ | $\vdots$ | $\vdots$ |
| $A_a$ | $x_{a11} \dots x_{a1r}$ | $x_{a21} \dots x_{a2r}$ | $\dots$ | $x_{ab1} \dots x_{abr}$ |

#### 일원배치법의 자료구조로 표현

| 실험반복 | $A_1 B_1$ | $\dots$ | $A_a B_b$ |
| - | - | - | - |
| 1 | $x_{111}$ | $\dots$ | $x_{ab1}$ |
| $\vdots$ | $\vdots$ | $\vdots$ | $\vdots$ |
| r | $x_{11r}$ | $\dots$ | $x_{abr}$ |

#### 이원배치법 자료에 대한 통계모형
- **$x_{ijk}$는 요인 $A$의 $i$번째 수준과 요인 $B$의 $j$번째 수준에서 $k$번째 관측된 반응값**
- **$r_{ij}$는 요인 $A$의 $i$번째 수준과 요인 $B$의 $j$번째 수준에서의 처리효과**
- **오차는 서로 독립인 $N(0, \epsilon^2_E)$**

$$x_{ijk} = \mu + r_{ij} + \epsilon_{ijk}$$

- 하지만 $r_{ij}$는 $A_i B_j$에서의 처리효과이기 때문에, 요인 $A$의 $i$번째 수준효과 $\alpha_i$와 요인 $B$의 $j$번째 수준효과 $\beta_j$로 설명할 수 있음
- 그럼에도 설명되지 않는 나머지 부분이 $A, B$간의 상호작용효과 $(\alpha \beta)_{ij}$로


$$(\alpha \beta)_{ij} = r_{ij}- \alpha_i - \beta_j $$

- 따라서 이원배치법 자료에 대한 통계모형은

$$x_{ijk} = \mu + \alpha_i + \beta_j + (\alpha \beta)_{ij} + \epsilon_{ijk}$$

<br>

#### 상호작용 : 요인들의 특정한 수준조합에서 일어나는 효과
- 요인 $A$의 효과가 요인 $B$의 수준에 따라 변화하는 경우에 요인 $A,B$ 간에는 상호작용효과가 존재


<br>

### 분산분석 (반복 있는 경우)

#### 귀무가설 

1. **$H_0 : \alpha_1 = \dots = \alpha_a = 0$ vs $H_1 :\;not\;H_0$**

<br>

2. **$H_0 : \beta_1 = \dots = \beta_a = 0$ vs $H_1 :\;not\;H_0$**

<br>

3. **$H_0 : \;모든\;(\alpha \beta)_{ij} = 0$ vs $H_1 :\;not\;H_0$**

<br>

$$T_{i.} = \sum_j \sum_k x_{ijk}\;\; \bar{x_{i..}}=T_{i..}/br\;\;i=1, 2, \dots, a$$

$$T_{.j} = \sum_i \sum_k x_{ijk}\;\; \bar{x_{.j.}}=T_{..j}/ar\;\;i=1, 2, \dots, b$$

$$T_{ij} = \sum_k x_{ijk}\;\; \bar{x_{ij.}}=T_{ij.}/r$$

$$T = \sum_j \sum_k x_{ijk}\;\; \bar{\bar{x}} = T / abr$$



<br>

- **분산분석의 요체는 총변동을 주효과들과 상호작용효과에 기인하는 병동으로 분해하는 것**
- 이원배치법은 요인 $A$, $B$의 수준수가 각각 $a, b$인 반복이 있는 자료로 재배치하면, 처리의 수가 $a\times b$개인 일원배치법으로 볼 수 있음
    - 따라서, 관측치 $x_{ijk}$의 총평균 $\bar{\bar{x}}$로부터의 거리인 총편차는 다음과 같이 분해
    
    $$(x_{ijk} - \bar{\bar{x}})= (\bar{x_{ij.}}-\bar{\bar{x}}) + (x_{ijk}-\bar{x_{ij.}})$$
    
    앞의 식을 제곱하여 모든 $i, j, k$에 대하여 합하면
    
    <br>
    
    $$\sum_i \sum_j \sum_k (x_{ijk} - \bar{\bar{x}})^2 = \sum_i \sum_j \sum_k (\bar{x_{ij.} - \bar{\bar{x}}})^2 + \sum_i \sum_j \sum_k (x_{ijk} - \bar{x_{ij.}})^2$$
    
    $$총변동\;=\;AB간의\;변동\;+\;잔차변동$$
    
    $$SS_t = SS_{AB} + SS_E$$
    
    - **총변동 $SS_T$는 $AB$간 변동 $SS_{AB}$와 잔차변동 $SS_E$로 분해**
    
    <br>
    
    - $CT = T^2/abr$일 때
    
    $$SS_T = \sum_i \sum_j \sum_k (x_{ijk} - \bar{\bar{x}})^2 = \sum_i \sum_j \sum_k x^2_{ijk}-CT$$
    
    $$SS_{AB} = \sum_i \sum_j \sum_k (x_{ij.} - \bar{\bar{x}})^2 = \sum_i {{T^2_{ij.}}\over{r}}-CT$$
    
    $$SS_E = \sum_i \sum_j \sum_k (x_{ijk}-\bar{x_{ij}})^2 = SS_T - SS_{AB}$$
    
    <br>
    
    - **처리가 요인 $A$와 요인 $B$의 수준의 조합임을 주지할 때 처리 간 변동 $SS_{AB}$의 일부는 요인 $SS_A$와 $SS_B$로 설명할 수 있으며, 상호작용효과가 처리효과에서 요인 $A$, $B$의 주효과를 뺀 나머지로 정의된 바와 같이**
    
    <br>
    
        **$SS_{A\times B}$중에서 $SS_A$와 $SS_B$를 제외한 남는 부분, 요인 $A$, $B$의 주효과만으로 설명될 수 없는 부분이 상호작용에 의한 변동 $SS_{A\times B}$**
    
    $$(\bar{x_{ij.}}-\bar{\bar{x}}) = (\bar{x_{i..}}-\bar{\bar{x}}) + (\bar{x}_{.j.} - \bar{\bar{x}}) + (\bar{x_{ij.}}-\bar{x_{i..}}-\bar{x_{.j.}}+\bar{\bar{x}})$$
    
    <br>
    
    위의 식을 제곱하여 $i,j,k$에 대하여 합하면
    
    <br>
    
    $$\sum_i \sum_j \sum_k (\bar{x_{ij.}}-\bar{\bar{x}})^2 = \sum_i \sum_j \sum_k (\bar{x_{i..}}-\bar{\bar{x}})^2 + \sum_i \sum_j \sum_k (\bar{x}_{.j.} - \bar{\bar{x}})^2 + \sum_i \sum_j \sum_k (\bar{x_{ij.}}-\bar{x_{i..}}-\bar{x_{.j.}}+\bar{\bar{x}})^2$$
    
    <br>
    
    - **$AB$간의 변동은 요인 $A$의 변동과, 요인 $B$의 변동, 상호작용 $A\times B$의 변동으로 분할**
    
    <br>
    
    $$SS_{A\times B} = SS_{A} + SS_B + SS_{AB}$$
    
    $$SS_A = \sum_i \sum_j \sum_k (\bar{x_{i..}}-\bar{\bar{x}})^2 = \sum_i {T^2_{i..}\over {br}}-CT$$
    
    $$SS_{B} = \sum_i \sum_j \sum_k (\bar{x}_{.j.} - \bar{\bar{x}})^2 = \sum_i {{T^2_{.j.}}\over{ar}}-CT$$
    
    $$SS_{A \times B} = \sum_i \sum_j \sum_k (\bar{x_{ij.}}-\bar{x_{i..}}-\bar{x_{.j.}}+\bar{\bar{x}})^2 = SS_{AB} - SS_A - SS_B$$
    
    <br>
    
    - **따라서 총변동 $SS_T$는**
    
    $$SS_T = SS_{AB} + SS_E = SS_A + SS_B + SS_{A\times B} + SS_E$$

- 각 변동의 자유도는 제곱을 한 서로 다른 편차의 개수에서 선형제약조건의 수를 뺀 것

$$\phi_T = abr -1$$

$$\phi_{AB} = ab -1$$

$$\phi_E = \phi_T - \phi_{AB} = ab(r -1)$$

$$\phi_A = a -1$$

$$\phi_B = b-1$$

$$\phi_{A\times B} = \phi_{AB}-\phi_A-\phi_B = \phi_A \times \phi_B = (a-1)(b-1)$$

<br>

- 평균제곱 $MS_A$, $MS_B$, $MS_{A\times B}$, $MS_E$는 각 변동을 각각의 자유도로 나눈 값

<br>

#### 검정통계량

1. **$F_0 = MS_A / MS_E > F(\phi_A, \phi_E;\;\;\alpha)$이면,요인 $A$가 반응치에 유의한 영향을 줌**

2. **$F_0 = MS_B / MS_E > F(\phi_B, \phi_E;\;\;\alpha)$이면,요인 $B$가 반응치에 유의한 영향을 줌**

3. **$F_0 = MS_{A\times B} / MS_E > F(\phi_{A \times B}, \phi_E;\;\;\alpha)$이면, A와 B사이 상호작용이 존재**

<br>

#### 분산분석표

| 요인 | 제곱합 | 자유도 | 평균제곱 | $F_0$ |
| - | - | - | - | - |
| $A$ | $SS_A$ | $a$-$1$ | $MS_A$ | $MS_A$/$MS_E$ |
| $B$ | $SS_B$ | $b$-$1$ | $MS_B$ | $MS_B$/$MS_E$ |
| $A\times$$B$ | $SS$$_{AXB}$ | $(a$-$1)(b$-$1)$ | $MS$$_{AXB}$ | $MS_{A\times\;B}/MS_E$ |
| $E$ | $SS_E$ | $ab(r$-$1)$ | $MS_E$ | |
| |
| $T$ | $SS_T$ | $abr$-$1$ |   |  |

#### 이원배치법에서의 가설검정에서 첫 번째 순서는 상호작용효과의 유의성을 검토하는 것
- **유의확률이 대략 0.25이상일 경우 상호작용효과가 존재하지 않아 요인 $A$의 주효과의 크기가 요인 $B$가 어떠한 수준값을 취하든지 간에 관계없이 똑같다고 판단하고 오차항에 풀링**

    **$\rightarrow$$A$, $B$의 주효과에 대한 검토**
- **상호작용효과가 유의하면, 주효과 자체의 유의성 검정은 $A$, $B$의 주효과의 크기 비교에 따른 상대적인 중요성을 가늠한다는 것을 제외하고는 커다란 의미가 없음**

    $A$의 주효과의 크기가 요인 $B$의 수준에 따라서 달라진다는 사실을 알았기 때문

### 분산분석 (반복 없는 경우)
- 반복수 $r = 1$이므로, 잔차변동의 자유도 $\phi_E = 0$
- **구조모형에서 $k$가 항상 1이므로 상호작용과 $(\alpha \beta)_{ij}$와 오차 $\epsilon_{ij1}$간 구분이 불가능하여, 주효과에 대한 검정을 하기 위해서는 상호작용효과와 교락된 오차항을 이용**

$$x_{ij} = \mu + \alpha_i + \beta_j + \epsilon_{ij}$$

- 총변동의 분해는

$$SS_T = SS_A + SS_B + SS_E$$

- $CT = T^2/ab$일 때

$$SS_T = \sum_i \sum_j (x_{ij} - \bar{\bar{x}})^2 = \sum_i \sum_j x^2_{ij}-CT$$

$$SS_{A} = \sum_i \sum_j (\bar{x_{i.}} - \bar{\bar{x}})^2 = \sum_i {{T^2_{i.}}\over{b}}-CT$$

$$SS_{B} = \sum_i \sum_j (\bar{x_{.j}} - \bar{\bar{x}})^2 = \sum_i {{T^2_{.j}}\over{a}}-CT$$

$$SS_E = \sum_i \sum_j  (x_{ij}-\bar{x_{i.}} - \bar{x_{.j}} + \bar{\bar{x}})^2 = SS_T - SS_{A} - SS_B$$

<br>

#### 분산분석표

| 요인 | 제곱합 | 자유도 | 평균제곱 | $F_0$ |
| - | - | - | - | - |
| $A$ | $SS_A$ | $a$-$1$ | $MS_A$ | $MS_A$/$MS_E$ |
| $B$ | $SS_B$ | $b$-$1$ | $MS_B$ | $MS_B$/$MS_E$ |
| $E$ | $SS_E$ | $(a$-$1)(b$-$1)$ | $MS_E$ | |
| |
| $T$ | $SS_T$ | $ab$-$1$ |   |  |

<br>

## 04.04. 혼합모형 (고정요인 & 랜덤요인)
- **요인이 둘 이상일 때 고려해야 할 사항 중 하나는 통계모형의 유형구분**
- **통제모형은 요인이 고정요인이면 고정모형, 랜덤요인이면 랜덤모형, 고정모형과 랜덤요인이 같이 있으면 혼합모형**

<br>

### 분산분석 (반복 없는 경우)
#### 요인 $A$가 고정요인이고, $B$가 랜덤요인일 때,
- **$\mu$ : 전체 모평균**
- **$\alpha_i$ : 인자 $A$의 효과로서 $\sum \alpha_i = 0$**
- **$\beta_j$ : 인자 $B$의 효과로서 서로 독립이고 N(0, $\sigma^2_B$)($\sum \beta_j \neq 0$)**
- **$\epsilon_{ij} \sim N(0, \sigma^2_E)$**

$$x_{ij}=\mu + \alpha_i + \beta_j + \epsilon_{ij}$$



- 혼합모형의 경우

    **인자 $B$는 랜덤요인이므로, 각 수준에서 모평균의 추정은 의미가 없고, 인자 $B$의 산포 $\sigma^2_B$를 추정**
    
    $$\hat{\sigma^2_B} = {{MS_B - MS_E}\over{a}}\;\;\;인자\;A의\;수준수$$
    
    **인자 $A$는 모수인자이므로, 각 수준에서 모평균을 추정**
    
<br>

### 분산분석 (반복 있는 경우)
- 반복이 있는 혼합모형의 경우, 두 인자간 상호작용효과도 파악 가능

#### 요인 $A$가 고정요인이고, $B$가 랜덤요인일 때,
- **$\mu$ : 전체 모평균**
- **$\alpha_i$ : 인자 $A$의 효과로서 $\sum \alpha_i = 0$**
- **$(\alpha \beta)_{ij}$ : 인자 $A$와 $B$의 교호작용효과로서 서로 독립인 $N(0, \sigma^2_{A\times B})$을 따름**

    $\sum_i (\alpha \beta)_{ij}=0,\;\;\;\sum_j(\alpha \beta)_{ij} \neq 0$

$$x_{ijk}=\mu + \alpha_i + \beta_j + (\alpha \beta)_{ij} + \epsilon_{ijk}$$


<br>

* 혼합모형이나 고정모형은 변동 $SS_T, SS_A, SS_B, SS_{A\times B}, SS_E$를 구하는 것이나 자유도를 구하는 데 차이가 없음

    **그러나 평균제곱 (MS)의 기대값에 차이가 존재 $\rightarrow$ 분산분석표에서 F검정시 차이점 존재**

    **A가 고정요인이고 B가 랜덤요인인 경우**
    
    $$E(MS_A) = \sigma^2_E = r \sigma^2_{A\times B} + br\sigma^2_A$$
    
    $$E(MS_B) = \sigma^2_E + ar\sigma^2_B$$
    
    $$E(MS_{A\times B}) = \sigma^2_E + r\sigma^2_{A\times B}$$
    
    $$E(MS_E) = \sigma^2_E$$
    
    **$\sigma^2_A = \sum_i \alpha^2_i / (a-1)$이고, $\alpha_i$들간의 산포도**
    
    **$\sigma^2_B, \sigma^2_{A\times B}, \sigma^2_E$는 확률변수 $\beta_j,\;(\alpha \beta)_{ij},\;\epsilon_{ijk}$의 분산을 의미**
    

<br>

#### 귀무가설
- $H_0 : \alpha_1 = \dots = \alpha_a = 0$  vs $H_1:\;not\;H_0$

    $\rightarrow$ $H_0 : \sigma^2_A=0$ vs $H_1:\sigma^2_A>0$
    
<br>

#### 검정통계량

- **고정모형과 같이 F비를 $MS_A / MS_E$로 놓는다면**

    **${{MS_A}\over{MS_E}}$의 평균 ${{E(MS_A)}\over{E(MS_E)}}$는**
    
    **$${{E(MS_A)}\over{E(MS_E)}}={{\sigma^2_E + r\sigma^2_{A\times B}+br\sigma^2_A}\over{\sigma^2_E}}$$가 되고, 귀무가설이 옳아서 $\sigma^2_A=0$이 되더라도, $\sigma^2_{A\times B}$이 큰 값을 갖는다면 F비의 값은 큰 값을 가지게 될 것**
    
    **다시 말해, $MS_A / MS_E$를 F비로 설정하면, F비가 큰 값이 나왔을 때, 그것이 $\sigma^2_A$ 때문인지, $\sigma^2_{A\times B}$ 때문인지 알 수 없음**
    
    <br>
    
    #### 따라서  F비를 $MS_A / MS_{A\times B}$로 설정
    
    **$${{E(MS_A)}\over{E(MS_{A\times B})}} = {{\sigma^2_E + r\sigma^2_{A\times B}+br\sigma^2_A}\over{\sigma^2_E + r\sigma^2_{A\times B}}}$$가 되고, 만약 귀무가설이 옳아서 $\sigma^2_A > 0$이 되면, F비의 값이 큰 값을 가지게 될 것**
    
* 귀무가설 하에서, $F_0 = MS_A / MS_{A\times B}$는 자유도 ($\phi_A, \phi_{A\times B}$)인 F분포를 따르며

    기각역은 $F_0 = MS_A / MS_B > F(\phi_A, \phi_{A\times B};\;\;\alpha)$

<br>

- **고정모형과의 차이점은, 요인 $A$가 특성값에 유의한 영향을 미치는지 검정할 때 $F$비에서 $MS_A$를 $MS_E$로 나누는 것이 아니라**

    **$MS_A$를 $MS_{A\times B}$로 나누는 것**
    


<br>

#### 분산분석표

| 요인 | 제곱합 | 자유도 | 평균제곱 | $F_0$ |
| - | - | - | - | - |
| $A$ | $SS_A$ | $a$-$1$ | $MS_A$ | $MS_A$/$MS_{A\times B}$ |
| $B$ | $SS_B$ | $b$-$1$ | $MS_B$ | $MS_B$/$MS_E$ |
| $A\times$$B$ | $SS$$_{A\times\;B}$ | $(a$-$1)(b$-$1)$ | $MS$$_{A\times B}$ | $MS$$_{A\times B}$/$MS_E$ |
| $E$ | $SS_E$ | $ab(r$-$1)$ | $MS_E$ | |
| |
| $T$ | $SS_T$ | $ab$-$1$ |   |  |

<br>

#### 화학제품의 불순물에 대한 영향조사
- 로트 B를 랜덤하게 3로트 선택 (랜덤)

    **오차항을 풀링할 것인가의 여부는 다음 사항으로 고려하여 결정**
    
    <br>
    
    1. **상호작용항의 유의확률이 0.25보다 큼 $\rightarrow$ 변수 제거**
    2. **0.25와 0.1 사이 $\rightarrow$ : 기술적으로 검토해서 변수 제거 여부 결정**
    3. **0.1보다 작음 $\rightarrow$ 변수를 제거하지 않음**

## 04.05. 평균제곱의 기댓값을 구하는 방법
- 이원배치법으 모형이 고정모형인가 혼합모형인가에 따라서, 고정요인의 주효과에 대한 유의성 검정통계량이 달라지므로 분산분석표도 달라짐
    
    따라서, 가설검정에 필요한 검정통계량을 알려면 **평균제곱의 기댓값**을 구해야 함

<br>

* **고정요인의 경우 수준변화에 따른 효과들 간의 산포의 측도는 효과의 제곱합을 자유도로 나눈 것으로 표시**

    A, B가 모두 고정요인인 경우, A의 산포의 측도 $\sigma^2_A$와 상호작용의 산포의 측도 $\sigma^2_{A\times B}$는
    
    $$\sigma^2_A={{\sum_i \alpha^2_i}\over{a-1}}$$
    
    $$\sigma^2_{A\times B}={{\sum_i \sum_j (\alpha \beta)^2_{ij}}\over{(a-1)(b-1)}}$$
    
    
<br>

- **A가 랜덤요인이면 주효과 $\alpha_i$는 확률변수이고, $\alpha_i$의 분산은 $\alpha^2_A$**

    **또한 상호작용에 1개 이상의 랜덤요인이 포함되어 있으면 상호작용을 확률변수로 간주**
    
<br>

### 이원배치법 혼합모형에서 평균제곱의 기댓값을 계산

$$x_{ijk}=\mu + \alpha_i + \beta_j + (\alpha \beta)_{ij} + \epsilon_{ijk}$$

<br>

#### 1. 
- **행에 순서대로 $\alpha_i, \beta_j, (\alpha \beta)_{ij}, \epsilon_{ij}k$를 넣음**
    - 오차항 $\epsilon_{ij}k$의 의미는 요인 A와 B의 수준이 정해진 후에 반복이 이루어지므로 이와 같이 표시
    
        **즉, 반복을 나타내는 첨자 $k$가 의미를 갖기 위해서는 A와 B의 수준값인 $i,j$가 주어져야 함**
        
- **첫 째열에는 A의 수준수($a$), 요인의 종류 [고정요인 F / 랜덤요인 R], 요인에 사용되는 첨자 ($i$)를 기입**

    두 번째열에는 B, 세 번째 열에는 반복
    
<br>

| | $a$ <br> $F$ <br> $i$ | $b$ <br> $R$ <br> $j$ | $r$ <br> $R$ <br> $k$ |
| - | - | - | - |
| $\alpha_i$ | | | |
| $\beta_j$ | | | |
| $(\alpha \beta)_{ij}$ | | | |
| $\epsilon_{ij}k$ | | | |

    
<br>

#### 2. 
- **행에 있는 첨자 중 괄호 속에 있는 것이 있으면 그 첨자와 같은 열의 첨자를 찾아 교차 점에 1을 기입**

| | $a$ <br> $F$ <br> $i$ | $b$ <br> $R$ <br> $j$ | $r$ <br> $R$ <br> $k$ |
| - | - | - | - |
| $\alpha_i$ | | | |
| $\beta_j$ | | | |
| $(\alpha \beta)_{ij}$ | | | |
| $\epsilon_{ij}k$ | $1$ | $1$ | |
    
<br>

#### 3.
- **각 행별로 열의 첨자와 같은 곳이 있는 곳을 찾아, 고정요인이면 0, 랜덤요인이면 1 기입**

| | $a$ <br> $F$ <br> $i$ | $b$ <br> $R$ <br> $j$ | $r$ <br> $R$ <br> $k$ |
| - | - | - | - |
| $\alpha_i$ |$0$ | | |
| $\beta_j$ | | $1$| |
| $(\alpha \beta)_{ij}$ | $0$| $1$| |
| $\epsilon_{ij}k$ | $1$ | $1$ | $1$|

<br>

#### 4.
- **나머지 빈곳을 찾아 수준수 (또는 반복수) 기입**

| | $a$ <br> $F$ <br> $i$ | $b$ <br> $R$ <br> $j$ | $r$ <br> $R$ <br> $k$ |
| - | - | - | - |
| $\alpha_i$ |$0$ |$b$ | $r$|
| $\beta_j$ | $a$| $1$| $r$|
| $(\alpha \beta)_{ij}$ | $0$| $1$| $r$|
| $\epsilon_{ij}k$ | $1$ | $1$ | $1$|

<br>

#### 5.
- **각 행의 첨자를 포함하는 열을 제외하고, 각 행별로 첨자가 있는 행만 골라서 적힌 숫자를 선택**

    예 **첫 행 $\alpha_i$의 $i$첨자를 포함하는 열 제외 $\rightarrow$ $i$가 있는 행들의 값을 각 행별로 곱함 $\rightarrow$ $br, r, 1$을 얻게 됨**
    
    이 규칙을 각 행 별로 적용

<br>

- 마지막 행의 기댓값 인 $E(MS_E)=\sigma^2_E$

| | $a$ <br> $F$ <br> $i$ | $b$ <br> $R$ <br> $j$ | $r$ <br> $R$ <br> $k$ | $평균제곱의\;기댓값$|
| - | - | - | - | - |
| $\alpha_i$ |$0$ |$b$ | $r$| $\sigma^2_E+r\sigma^2$$_{AXB}+br\sigma^2_A$|
| $\beta_j$ | $a$| $1$| $r$| $\sigma^2$$_E$$+ar\sigma^2$$_B$| 
| $(\alpha \beta)_{ij}$ | $0$| $1$| $r$|$\sigma^2$$_E+r\sigma^2$$_{AXB}$ | 
| $\epsilon_{ij}k$ | $1$ | $1$ | $1$| $\sigma^2$$_E$|
