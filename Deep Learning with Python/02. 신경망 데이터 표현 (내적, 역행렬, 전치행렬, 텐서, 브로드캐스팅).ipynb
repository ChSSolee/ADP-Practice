## 텐서(Tensor)
- **일반적으로 텐서는 3차원 이상을 다룰 때 표현하는 방식**이지만, 여기서는 어떠한 데이터를 표현할 때, 그 값 모두를 텐서라고 부르기로 함

      a = np.array([1, 2])
      b = np.array([[1, 2],
                    [3, 4]])
      c = np.array([10]) 
      d = np.array(3)  
  
    a, b, c, d 모두 텐서라고 지칭할 수 있음
  
- **랭크(rank): 텐서의 축을 나타내고, 넘파이(numpy)의 `ndim`으로 구할 수 있음**

      a = np.array([1, 2])
        >> ndim = 1  
      b = np.array([[1, 2],
                    [3, 4]])
        >> ndim = 2
      d = np.array(3)
        >> ndim = 0
      
    위의 예시에서 알 수 있듯이 대괄호( [ ] )의 개수가 곧 랭크(축)의 값
    
      
- 크기(shape): 텐서의 각 축을 따라 얼마나 많은 차원이 있는지를 나타내며, 파이썬의 튜플(tuple) 형태

### 벡터의 합
- 같은 형상(shape)일 때, 각 원소별로 계산

```
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
c = a + b

print(c)
print(c.shape)
print(np.ndim(c))
```

<br>

### 벡터의 곱
-  $A = (x_1, \ x_2, \ x3, \ ..., \ x_n)$  
  $B = (y_1, \ y_2, \ y3, \ ..., \ y_n)$ 일 때,
- **원소곱**
  - 같은 형상(shape)일 때, 각 원소별로 계산   
   
    $$A \times B = (x_1, \ x_2, \ x_3, \ ..., \ x_n) \times (y_1, \ y_2, \ y_3, \ ..., \ y_n) \\   
    \qquad = (x_1y_1, \ x_2y_2, \ x_3y_3, \ ... \ ,  \ x_ny_n)  $$
    
- **벡터곱(product, dot)**
  - 두 1차원 벡터가 있을 때 **각각의 성분끼리의 곱을 모두 더하는 계산**  
  
$$A \bullet B \Rightarrow A \times B^T = (x_1, \ x_2, \ x_3, \ ..., \ x_n)     \begin{pmatrix} y_1 \\ y_2 \\ y_3 \\ ... \\ y_n  \end{pmatrix}$$
    
    
$$\qquad = (x_1y_1 + \ x_2y_2 + \ x_3y_3 + \ ... + \ x_ny_n)$$

<br>

#### `np.dot(A, B)`


```
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
c = a * b

print(c)
print(c.shape)
print(np.ndim(c))

print("\n---------------------------\n")

x = np.array([1, 2, 0])
y = np.array([0, 2, 1])
z = np.dot(x, y)
print(z)
print(z.shape)
print(np.ndim(z))
```

## 2차원 텐서(행렬)
- **2차원 텐서는 행렬로 생각할 수 있음**

![matrix](https://upload.wikimedia.org/wikipedia/commons/thumb/4/4d/Matrix_ko.svg/570px-Matrix_ko.svg.png)

```
matrix = np.array([[1, 2, 3],\
                   [4, 5, 6]])

print(matrix)
print(matrix.shape)
print(np.ndim(matrix))

print("\n---------------------------\n")

matrix2 = np.array([[1, 2, 3, 4]])
print(matrix2)
print(matrix2.shape)
print(np.ndim(matrix2))
```

<br>

### 행렬 원소곱
- 같은 형상(shape)일 때 덧셈, 곱셈과 같은 연산은 원소별로 진행


```
A = np.array([[1, 2], [3, 4]])
B = np.array([[10, 10], [10, 10]])

print("행렬 A\n", A)
print("행렬 B\n", B)
print("A * B\n", A * B)
```

<br>

### 행렬 점곱(내적, product)
- 1차원 벡터와 마찬가지로 앞 행렬의 열과 뒤 행렬의 행의 수가 같아야함

```
M = np.array([[1, 2], [3, 4]])
N = np.array([[2, 3, 4], [2, 3, 4]])

print("행렬 M\n", M)
print("행렬 N\n", N)
L = np.dot(M, N)

print("행렬 L\n", L)
print(L.shape)
print(np.ndim(L))
```

<br>

### 역행렬
- **어떤 행렬 A가 있을 때, 곱해서 단위행렬(E)를 만드는 행렬 B가 존재한다면, 행렬 B는 A의 역행렬**
  - 단위행렬(E)를 만드는 행렬이 없다변, 그 행렬의 역행렬은 존재하지 않음

<br>

#### `np.linalg.inv(A)`

```
A = np.array([[1, 2,], [3, 4]])
print("행렬 A: ",A, "\n")

B = np.linalg.inv(A)
print("행렬 A의 역행렬:", B, "\n")

print("행렬 A와 A역행렬의 내적 = 단위행렬(E)\n", np.dot(A, B))
```

<br>

## 텐서 크기 변환
- `reshape`로 텐서의 크기 변환 가능
- **변환 전의 원소의 개수와 변환 이후의 텐서의 개수가 같아야** 함

<br>

#### `행렬.reshape()`

```
A = np.array([[1, 2, 3], [4, 5, 6]])
print("행렬 A\n", A)
print("A.shape:", A.shape, end='\n\n')
print("-----------------")

A = A.reshape(6)
print("행렬 A\n", A)
print("A.shape:", A.shape, end='\n\n')
```

```
B = np.array([[[2, 3, 4], [2, 3, 4]],
              [[1, 1, 1], [1, 1, 1]]])

print("행렬 B\n", B)
print("B.shape", B.shape)
print("------------------")

B = B.reshape(3, 4)
print("행렬 B\n", B)
print("B.shape", B.shape)
```

<br>

- **-1 을 통해 자동으로 형상을 지정 가능**  
  원소의 갯수에 맞게 넘파이가 자동으로 형상을 지정  
      (2, 2, 3) -> (3, -1)     (O)
                -> (2, 1, 6)   (O)
                -> (2, -1, -1) (X)
                -> (2, 5, -1)  (X) 

```
B = np.array([[[2, 3, 4], [2, 3, 4]],
              [[1, 1, 1], [1, 1, 1]]])

print("행렬 B\n", B)
print("B.shape", B.shape)
print("------------------")

B = B.reshape(4, -1)
print("행렬 B\n", B)
print("B.shape", B.shape)
```
