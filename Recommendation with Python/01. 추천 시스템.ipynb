# 01. 추천 시스템
- 추천 시스템이란 사용자의 과거 행동 데이터나 다른 데이터를 바탕으로 사용자에게 필요한 정보나 제품을 골라서 제시해 주는 시스템
- 전통적인 비즈니스에서 추천은 고객들을 취향이 비슷한 집단 (세그먼트)로 나누고, 각 집단에 잘 맞는 제품이나 서비스를 골라 주는 것
    - 집단의 크기를 극단적으로 줄이면, 각 집단이 한 사람이 되며
    
        $\rightarrow$ 각 사용자별로 맞춤형 추천과 서비스를 제공하는 것을 개인화 (Personalization)
        
- **사용자에게 맞춤형 추천을 제공하기 위해서는 각 사용자가 어떤 아이템을 좋아하는지 선호도를 예측하는 것이 필수적**

    $\rightarrow$ 선호도 예측 결과를 아이템을 추천하는 형태로 사용하거나, 다른 형태의 서비스에 활용 가능

<br>

- 추천의 기술
> 협업 필터링
>
> 내용 기반 필터링
>
> 지식 기반 필터링
>
> 하이브리드 : 2개 이상의 기술 혼합

## 01.01. 주요 추천 알고리즘

<br>

### 협업 필터링 (Collaborative Filtering : CF)
- **구매 소비한 제품에 대한 각 소비자의 평가를 받아서 평가 패턴이 비슷한 소비자를 한 집단으로 보고,**
    
    **그 집단에 속한 소비자들의 취향을 활용**
    
    - **A라는 소비자와 제품평가 패턴이 비슷한 사람들을 골라서,** 

        **이 사람들이 공통적으로 좋아하는 제품 중에서 A가 아직 접하지 않은 제품을 골라내서 제시하는 기술**
        
- **사람들의 취향이 뚜렷이 구분되는 제품을 추천할 떄 더욱 정확함**

- **소비자들의 평가 정보를 구하기 어려울 수도 있다는 단점 존재**
    - 제품을 구입한 후에 평가 정보를 요청하면 이를 제공하는 사람이 많지 않음
    
    $\rightarrow$ 이러한 한계를 극복하기 위해서 간접적인 정보를 활용 (장바구니, 클릭스트림)
    
- 협업 필터링의 대표적인 예 : 아마존의 제품 추천, 넷플릭스 영화 추천 시스템

<br>

### 내용 기반 필터링 (Content-Based Filtering : CB)
- **제품의 내용을 분석해서 추천하는 기술**
- **소비자가 소비하는 제품 중 텍스트 정보가 많은 제품을 분석하여 추천할 때 많이 이용**
- 텍스트 중에서 형태소를 추려내고, 이 중에서도 핵심 키워드가 어떤 것인지를 분석하는 기술 필요
    
    $\rightarrow$ **추려낸 키워드를 활용해서 소비자의 관심사항에 대한 분석이 가능**

<br>

### 지식 기반 필터링 (Knowledge-Based Filtering : KB)
- **협업 필터링과 내용 기반 필터링의 공통적인 단점은 전체적인 그림이 없다는 것**

    **어떤 제품을 좋아할지에 대한 분석은 가능하지만, '왜 그 제품을 좋아할 것인가'에 대해서는 답을 해줄 수 없음**
    
- **지식 기반 필터링은 특정 분야 전문가의 도움을 받아서 그 분야에 대한 전체적인 지식구조를 만들어서 이를 활용**
    - 전문가는 이들 각각에 대해서 그 밑에 어떤 것들이 있고, 다시 그 밑에는 또 어떤 것이 있는지 등을 제시
    - 작성된 체계도를 바탕으로 소비자가 구매한, 혹은 관심 있는 제품과 관련이 있는 제품을 이 체계도에서 찾아서 제시할 수 있음
- 지식을 기반으로 한 필터링의 장점은 전체적인 구조를 알 수 있다는 것이고,
    
    지식이 중요한 분야에서 효과가 좋음
- **각 분야에 대해서 전문가가 필요하고, 모든 세세한 분야에 대해서 다 적용할 수 없음**

<br>

### 딥러닝 (Deep Learning : DL) 추천 기술
- AI 알고리즘 중에 현재 가장많이 사용되는 딥러닝 방법을 사용하는 형식
- 딥러닝의 입력으로 다양한 사용자와 아이템의 특정값을 사용하고,
    
    이를 통해 얻은 출력은 각 아이템에 대한 각 사용자의 예상 선호도로 사용하는 것이 보통
    
    $\rightarrow$ 각 사용자에 대해서 많은 아템의 예상 선호도를 계산할 수 있고, 이 중에서 예상 선호도가 높은 아이템을 추천
    
- **다양한 입력 변수를 사용할 수 있다는 장점 (이미지 인식 등에서 탁월한 성능)**

<br>

### 하이브리드 (Hybrid) 기술
- 실제로 추천 시스템을 구축할 때는 한 가지만 사용하는 경우는 거의 없으며,
    
    대부분 두 가지 이상의 기술을 혼합해서 사용하는 하이브리드 형태로 주로 사용
- 아마존의 추천 시스템에서도 협업 필터링이 추가 되지만, 그 외의 다양한 기술이 같이 사용
- 하나의 기술을 사용할 때 보다, 복수의 기술을 사용하는 경우에 정확도가 향상되고 결합되는 기술의 종류가 많아질수록 더 큰 향상이 이루어짐



<br>

## 추천 시스템 적용 사례

<br>

### 넷플릭스의 사례
- 넷플릭스는 영화를 본 고객이 그 영화에 대한 평가를 하면

    이를 바탕으로 협업 필터링 기술을 사용해서 그 고객이 좋아할만한 영화를 추천해 주는 'CineMatch'라는 시스템을 가동
    
    $\rightarrow$ 고객이 넷플릭스에 로그인을 하는 화면에는 이 시스템이 추천한 영화로 채워짐
    
    $\rightarrow$ 넷플릭스 웹사이트 자체가 각 고객의 취향에 따라 맞춤화, 더 나아가서 개인화 됨
- 넷플릭스에서 소비하는 영화 중 양 80%가 이 추천에 의한 것

<br>

### 아마존의 사례
- 아마존은 협업 필터링을 기반으로, 다양한 추천 기수을 결합한 하이브리드 기술이 적용
- 아마존은 제품에 대한 평가 데이터를 다양한 방법으로 수집 (제품 평점, 후기, 장바구니, 세션 정보)
    
    $\rightarrow$ 고객의 선호도를 계산할 때 가중치로서 사용

- 아마존은 사용자 수도 더 많기에 복잡하고 어려운 추천 시스템을 운용
1. 아마존의 시스템은 처리하는 신호 정보의 종류가 다양함
2. 아마존의 제품수와 사용자 수가 넷플릭스보다 더 많은
3. 취급하는 제품 카테고리의 수가 많음
